```python
from sklearn.datasets import load_boston
boston = load_boston()
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.DataFrame(boston.target)
```

<h1>Create dataframe from X, y</h1>

```python
# create dataframe from X, y for easier plot handling
dataframe = pd.concat([X, y], axis=1)
dataframe.head()
```

<h1>Evaluate Data Analysis</h1>

```python
# Linear Regression Model
lin_x = np.c_[dataframe["LSTAT"]]
lin_y = np.c_[y]
model = LinearRegression()
model.fit(lin_x, lin_y)

# predict y from the data
x_new = np.linspace(lin_x.min(), lin_x.max(), 200)
y_new = model.predict(x_new[:, np.newaxis])


plt.plot(X['LSTAT'], y.values, "b.")
plt.plot(x_new, y_new, "r-")
plt.xlabel("$x_1$ - % Lower Status of The Population", fontsize=18)
plt.ylabel("$y$ - Housing Price", rotation=90, fontsize=18)
# plt.axis([-3, 3, 0, 10])
# plt.savefig("images/scatter_lin_reg_plot.png", bbox_inches='tight')
plt.show()
```

<div style="font-size:14px; border:1px solid black; padding:10px">
    <center><h1>Standard liner regression does not fit the data properly.</h1></center>
<ul>
<li>Linear models are typically considered for data that scatters in a single direction (correlate).</li><br>    
<li>In many cases, data will be more complex than a straight line, calling for a more nuanced approach.</li><br>
<li>Linear models can be used to fit nonlinear data.</li><br>
<li>One method is to add new features that contains the powers of the original features.</li><br>
<li>A linear model can then be trained on this new extended data, and this technique is called <font color="red"><strong><em>Polynomial regression</em></strong></font>.</li><br>    
</ul>
</div>

<hr style="border-top: 3px solid Black;">

<h1>Generating polynomial features using sklearn <code>PolynomialFeatures()</code></h1>

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(np.c_[dataframe["LSTAT"]])
```

<h1>Linear Regression on Transformed Data</h1>

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
image = os.path.join(path, 'images/')
Image(filename=image)
```

<h1>Plot Data with Linear Regression Curve that used Transformed Data</h1>

```python
X_new=np.linspace(0, 40, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)
plt.plot(X['LSTAT'], y.values, "b.")
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper right", fontsize=14)
# plt.axis([-3, 3, 0, 10])
# plt.savefig("images/polynomimal_plot_linreg.png", bbox_inches='tight')
plt.show()
```

    <li>The prediction curve now more closely fits the shape of the data.</li><br>

<h1>Evaluate Learning Curve</h1>

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))

    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)   
    plt.xlabel("Training set size", fontsize=14) 
    plt.ylabel("RMSE", fontsize=14) 
```

<h1>Learning Curve for Simple Linear Regression</h1>

```python
plt.figure(figsize=(15,8))
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, lin_x, lin_y)
plt.axis([0, 100, 0, 20])  
plt.show()   
```

<div style="font-size:16px; border:1px solid black; padding:10px">
        <center><h3>Learning Curve for Simple Linear Regression.</h3></center>
<ul><em><font color="red">Training curve</font></em>
    <li>The error (RMSE) is 0 when the training is a few instances, which means the model can perfectly fit the data.</li><br>      
    <li>As new training instances were added, the model is no longer able ot fit the training data, especially as the data is not linear and the data is noise, especially at the two extreme of the data distribution.</li><br>
    <li>The training data error goes up and plateanus at around 5.0 at 15-20 instances, and no additional data alters the average RMSE. The model can no longer learn from the training.</li><br>
</ul>
<ul><em><font color="red">Validation curve</font></em>
    <li>When the model is trained with a few training instances, the model cannot generalize well so the RMSE remains high.</li><br>
    <li>As the training set size increases, the model learns from the data, and the RMSE goes down.</li><br>
    <li>Since the model is based on a simple linear regression (straight line), it cannot do a good job to model the actual data, and the error plateaus at a high RMSE above 7.5</li><br>    
    <li>The training and leanring curves remain close, but the distance between them suggest the model complexity could be improved.</li><br>      
</ul>
</div>

<h1>Learning Curve for Polynomial Linear Regression</h1>

```python
from sklearn.pipeline import Pipeline
plt.figure(figsize=(15,8))
polynomial_regression = Pipeline([
        ("poly_features", PolynomialFeatures(degree=2, include_bias=False)),
        ("lin_reg", LinearRegression()),
    ])

plot_learning_curves(polynomial_regression, lin_x, lin_y)
plt.axis([0, 100, 0, 20]) 
plt.show()
```
