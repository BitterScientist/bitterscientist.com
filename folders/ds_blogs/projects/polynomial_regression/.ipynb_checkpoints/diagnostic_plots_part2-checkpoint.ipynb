<h1>Residual Histogram</h1>

<div style="font-size:16px; border:1px solid black; padding:10px">
<font color="blue">Explained</font>
<ul>
<li>A <em>residual</em> is the difference between the predicted value and the actual value.</li>    
<li>A histogram of the Residuals can be used to check whether the variance is normally distributed.</li>
<li>A symmetric bell-shaped histogram that is evenly distributed around zero indicates that the normality assumption is likely to be true.</li>
<li>If the histogram's residual is not symmetric, it may indicate that the random error is not normally distributed.</li>    
<li>To plot a residual hitogram, you can use seaborns distplot:<br><strong><code>seaborn.distplot(a, bins=None, hist=True, kde=True, rug=False, fit=None, hist_kws=None, kde_kws=None, rug_kws=None, fit_kws=None, color=None, vertical=False, norm_hist=False, axlabel=None, label=None, ax=None)</code></strong></li>
<li>Additional information on distplot: <a href="https://seaborn.pydata.org/generated/seaborn.distplot.html">Documentation</a></li>
</ul>
</div>

```python
sns.distplot(model_residuals, bins=20, kde=True, rug=True).set_title('Residual Histogram')
```

<div style="font-size:16px; border:1px solid black; padding:10px">
<font color="blue">Observations:</font>
<ul>
<li>The histogram appears heavily left-skewed.</li>
<li>The histogram does not have the ideal bell-shape for normality</li>
</ul>
<font color="blue">Preliminary Conclusion:</font>
<ul>
<li>The data may not be suited for a simple normal equation, linear regression.</li>
</ul>    
</div>
