<h1>Import Dependencies</h1>

```python
# Scikit-Learn â‰¥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"
# Linear regresison module
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Math module
import numpy as np

# to make this notebook's output stable across runs
np.random.seed(42)

# Data Visualization
import matplotlib.pyplot as plt

# Formats plots and uses seaborn theme
plt.style.use('seaborn')
plt.rc('font', size=14)
plt.rc('figure', titlesize=18)
plt.rc('axes', labelsize=15)
plt.rc('axes', titlesize=18)

# Read data
%store -r lin_x
%store -r lin_y
```

<hr style="border-top: 3px solid Black;">

<h1>Evaluate Learning Curve</h1>

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))

    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)   
    plt.xlabel("Training set size", fontsize=14) 
    plt.ylabel("RMSE", fontsize=14) 
```

<h1>Learning Curve for Simple Linear Regression</h1>

```python
plt.figure(figsize=(15,8))
lin_reg = LinearRegression()
plot_learning_curves(lin_reg, lin_x, lin_y)
plt.axis([0, 100, 0, 20])  
plt.show()   
```

<div style="font-size:16px; border:1px solid black; padding:10px">
        <center><h3>Learning Curve for Simple Linear Regression.</h3></center>
<ul><em><font color="red">Training curve</font></em>
    <li>The error (RMSE) is 0 when the training is a few instances, which means the model can perfectly fit the data.</li><br>      
    <li>As new training instances were added, the model is no longer able ot fit the training data, especially as the data is not linear and the data is noise, especially at the two extreme of the data distribution.</li><br>
    <li>The training data error goes up and plateanus at around 5.0 at 15-20 instances, and no additional data alters the average RMSE. The model can no longer learn from the training.</li><br>
</ul>
<ul><em><font color="red">Validation curve</font></em>
    <li>When the model is trained with a few training instances, the model cannot generalize well so the RMSE remains high.</li><br>
    <li>As the training set size increases, the model learns from the data, and the RMSE goes down.</li><br>
    <li>Since the model is based on a simple linear regression (straight line), it cannot do a good job to model the actual data, and the error plateaus at a high RMSE above 7.5</li><br>    
    <li>The training and leanring curves remain close, but the distance between them suggest the model complexity could be improved.</li><br>      
</ul>
</div>

<h1>Learning Curve for Polynomial Linear Regression</h1>

```python
from sklearn.pipeline import Pipeline
plt.figure(figsize=(15,8))
polynomial_regression = Pipeline([
        ("poly_features", PolynomialFeatures(degree=2, include_bias=False)),
        ("lin_reg", LinearRegression()),
    ])

plot_learning_curves(polynomial_regression, lin_x, lin_y)
plt.axis([0, 100, 0, 20]) 
plt.show()
```
