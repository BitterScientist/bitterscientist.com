<h1>Import Dependencies</h1>

```python
# Python â‰¥3.5 is required
import sys
assert sys.version_info >= (3, 5)
import os
path = os.getcwd()

# Common imports
import numpy as np
import pandas as pd
import re

import sqlite3
from sqlite3 import Error

# Reading an excel file using Python
import xlrd

# to make this notebook's output stable across runs
np.random.seed(42)
```

<h1>Traverse Files System</h1>

```python
files = os.listdir(path) # get current path
data_folder = 'static/data' # data subfolder
sub_dir = os.path.join(path, data_folder) #  create relative path to data folder
subfolder_files = os.listdir(sub_dir) # get all file names in the subfolder
data_files = [] # empty list
for data_file in subfolder_files: # iterate through subfolder
    if data_file.endswith(".xlsx"): # identify xlsx files
        data_files.append(os.path.join(data_folder, data_file)) # save xlsx filepath to data_files list
```

<h1>Get specific worksheets from HIT workbooks, save as dataframes, and save individual files as csv files</h1>

```python
frames = []
for file in data_files:
    xls = pd.ExcelFile(file)
    sheets = xls.sheet_names
    for sheet in sheets:
        if re.search('All Projects', sheet) != None or re.search('ALL', sheet) != None:
            df_temp = pd.read_excel(file, sheet_name=f'{sheet}')
            df_temp["Year"] = re.findall(r"data/(.*?)\-",file)[0]
            df_temp.to_csv(f"{file[:-5]}.csv", index=False)
            frames.append(df_temp)
        else:
            pass
```

<h1>Get CSV file information for concatenation step</h1>

```python
csv_file_path = [] # empty list
csv_file_name = []
for data_file in subfolder_files: # iterate through subfolder
    if data_file.endswith(".csv"): # identify xlsx files
        csv_file_path.append(os.path.join(data_folder, data_file)) # save xlsx filepath to data_files list
        csv_file_name.append(data_file)
```

```python
# concatenate files
df_frames = pd.concat(frames, join="inner")
df_frames.to_csv("static/data/data_files.csv", index=False)
```

<h1>How many Intances are there per year</h1>

```python
engine.execute("SELECT Year, COUNT(Year) AS Instances FROM data GROUP BY Year").fetchall()
```

<h1>How many beds are there per year</h1>

```python
 engine.execute("SELECT Year, SUM(CAST(`Total Beds` AS Integer)) AS `Total Beds` FROM data GROUP BY Year").fetchall()
```

<h1>How many Distinct Organizations Are there per year</h1>

```python
 engine.execute("SELECT Year, COUNT( DISTINCT `Organization Name`) AS `Provider Count` FROM data GROUP BY Year").fetchall()
```

<h1>What Are the project types and their total counts by year</h1>

```python
 engine.execute("SELECT Year, `Proj. Type`, COUNT(`Proj. Type`) AS `Provider Count` FROM data GROUP BY Year, `Proj. Type` ORDER BY Year, `Proj. Type`;").fetchall()
```

<h1>Are the number of ES project types increasing year over year?</h1>

```python
 engine.execute("""SELECT Year, COUNT(`Proj. Type`)
                 FROM data 
                 WHERE `Proj. Type` = 'ES'
                 GROUP BY Year;""").fetchall()
```

<h1>Has the number of beds with children and project type ES increased year over year</h1>

```python
# Run query
query0 = """SELECT Year, SUM(CAST(`Beds HH w/ Children` AS Integer)) AS 'Total Beds with Children'
                 FROM data 
                 WHERE `Proj. Type` = 'ES' AND CAST(`Beds HH w/ Children` AS Integer) > 0
                 GROUP BY Year;"""
view_data0 = pd.read_sql(query0, engine)
# Displaying subset of data
view_data0
```

<h1>What are the average number of beds with children by year for Project type ES</h1>

```python
 engine.execute("""SELECT Year, AVG(CAST(`Beds HH w/ Children` AS Integer)) AS 'Avg Beds with Children'
                 FROM data 
                 WHERE `Proj. Type` = 'ES'
                 GROUP BY Year;""").fetchall()
```

<h1>Is there a year where the median number of beds with children significantly different?</h1>

```python
# Run query
query2 = """SELECT Year, "Organization Name", CAST(`Beds HH w/ Children` AS Integer) AS 'Beds with Children'
                 FROM data 
                 WHERE `Proj. Type` = 'ES' AND CAST(`Beds HH w/ Children` AS Integer) > 0;"""
view_data2 = pd.read_sql(query2, engine)
# Displaying subset of data
view_data2.head(5)
```

```python
# generate a boxplot to see the data distribution by treatments. Using boxplot, we can 
# easily detect the differences between different treatments
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(8, 6), dpi=300)
ax = sns.boxplot(x="Year", y='Beds with Children', data=view_data2, color='#99c2a2')
# ax = sns.swarmplot(x="Year", y="Beds with Children", data=view_data2, color='#7d0013')
plt.show()
```

```python
data_20 = view_data2[view_data2['Year']=='2020']['Beds with Children']
data_19 = view_data2[view_data2['Year']=='2019']['Beds with Children']
data_18 = view_data2[view_data2['Year']=='2018']['Beds with Children']
data_17 = view_data2[view_data2['Year']=='2017']['Beds with Children']
data_16 = view_data2[view_data2['Year']=='2016']['Beds with Children']
```

```python
import scipy.stats as stats
# stats f_oneway functions takes the groups as input and returns ANOVA F and p value
fvalue, pvalue = stats.kruskal(data_20, data_19, data_18, data_17, data_16)
print(fvalue, pvalue)
```

- Post-hoc test: Dunn test for multiple comparisons of groups
- If the Kruskalâ€“Wallis test is significant, a post-hoc analysis can be performed to determine which groups differ from each other group. 
- Probably the most popular post-hoc test for the Kruskalâ€“Wallis test is the Dunn test.
- holm: step-down method using Bonferroni adjustments
