# Import dependencies

```python
import numpy as np

# to make this notebook's output stable across runs
np.random.seed(42)

# To visualize data
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# ML specific

# To get MNIST dataset
from sklearn.datasets import fetch_openml
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score
from sklearn.base import BaseEstimator
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import precision_score, recall_score
```

# Get dataset

```python
mnist = fetch_openml('mnist_784', version=1)
mnist.keys()
```

### Dataset shape

```python
X, y = mnist["data"], mnist["target"]
```

```python
X.shape
```

This dataset has 70,000 instances and 784 features

```python
y.shape
```

### Cast target column from string to integer

```python
y = y.astype(np.uint8)
```

# Display an image from dataset

```python
def plot_digit(data):
    image = data.reshape(28, 28)
    plt.imshow(image, cmap = mpl.cm.binary,
               interpolation="nearest")
    plt.axis("off")
```

```python
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = mpl.cm.binary, **options)
    plt.axis("off")
```

```python
plt.figure(figsize=(9,9))
example_images = X[100:200]
plot_digits(example_images, images_per_row=10)
# plt.savefig("static/images/fig1.png", bbox_inches='tight')
plt.show()
```

### Split data

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

### Train Binary Classifier

```python
y_train_7 = (y_train == 7)
y_test_7 = (y_test == 7)
```

### Percent of dataset that is 7

```python
percent_7 = round(sum(y_train_7)/len(y_train_7)*100,2)
print(f'The percent of images that are "7" is: {percent_7}%')
```

```python
sgd_clf = SGDClassifier(max_iter=1000, tol=1e-3, random_state=42)
sgd_clf.fit(X_train, y_train_7)
```

### Measure Accuracy of model using Cross-Validation

```python
cross_val_score(sgd_clf, X_train, y_train_7, cv=3, scoring="accuracy")
```

The accuracy for this model is above 93% for the cross-validation folds tested.

### Measure Accuracy of model at Predicting Non-7 numbers

```python
class ClassifierNot7(BaseEstimator):
    def fit(self, X, y=None):
        pass
    def predict(self, X):
        return np.zeros((len(X), 1), dtype=bool)
```

```python
never_7_clf = ClassifierNot7()
cross_val_score(never_7_clf, X_train, y_train_7, cv=3, scoring="accuracy")
```

### Use cross_val_predict to return a prediction for each instance in the training set

```python
# Create a set of predictions on training dataset
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_7, cv=3)
```

# Precision and Recall

```python
precision_score(y_train_7, y_train_pred)
```

```python
recall_score(y_train_7, y_train_pred)
```

# Obtain Array of Predicted Values using cross_val_predict()

```python
y_scores = cross_val_predict(sgd_clf, X_train, y_train_7, cv=3, method="decision_function")
```

# Plot ROC Curve

```python
fpr, tpr, thresholds = roc_curve(y_train_7, y_scores)
```

### ROC Curve function

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') # dashed diagonal
    plt.axis([0, 1, 0, 1])                                    
    plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16) 
    plt.ylabel('True Positive Rate (Recall)', fontsize=16)    
    plt.grid(True)
```

```python
plt.figure(figsize=(8, 6))                        
plot_roc_curve(fpr, tpr)
plt.plot([20.837e-3, 20.837e-3], [0., 0.8901], "r:")
plt.plot([0.0, 20.837e-3], [.8901, .8901], "r:")
plt.plot([20.837e-3], [0.8901], "ro")              
plt.savefig("static/images/roc_curve.png", bbox_inches='tight')
plt.show()
```

```python
# Calculate score
roc_auc_score(y_train_7, y_scores)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier(n_estimators=100, random_state=42)
y_probas_forest = cross_val_predict(forest_clf, X_train, y_train_7, cv=3,
                                    method="predict_proba")
```

```python
y_scores_forest = y_probas_forest[:, 1] # score = proba of positive class
fpr_forest, tpr_forest, thresholds_forest = roc_curve(y_train_7,y_scores_forest)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, "b:", linewidth=2, label="SGD")
plot_roc_curve(fpr_forest, tpr_forest, "Random Forest")
# Line markers for SGD
plt.plot([20.837e-3, 20.837e-3], [0., 0.8901], "r:")
plt.plot([0.0, 20.837e-3], [.8901, .8901], "r:")
plt.plot([20.837e-3], [0.8901], "ro")
# Line markers for random forest
plt.plot([4.837e-3, 4.837e-3], [0., 0.9487], "r:")
plt.plot([4.837e-3], [0.9487], "ro")
plt.grid(True)
plt.legend(loc="lower right", fontsize=16)
plt.savefig("static/images/roc_curve_comparison_plot.png", bbox_inches='tight')
plt.show()
```

```python
roc_auc_score(y_train_7, y_scores_forest)
```

```python
y_train_pred_forest = cross_val_predict(forest_clf, X_train, y_train_7, cv=3)
precision_score(y_train_7, y_train_pred_forest)
```

```python
recall_score(y_train_7, y_train_pred_forest)
```
