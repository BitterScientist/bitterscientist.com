# Import dependencies

```python
import numpy as np

# to make this notebook's output stable across runs
np.random.seed(42)

# To visualize data
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# ML specific

# To get MNIST dataset
from sklearn.datasets import fetch_openml
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score
from sklearn.base import BaseEstimator
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import precision_score, recall_score, f1_score, precision_recall_curve
from mlxtend.plotting import plot_confusion_matrix
```

# Get dataset

```python
mnist = fetch_openml('mnist_784', version=1)
mnist.keys()
```

### Dataset shape

```python
X, y = mnist["data"], mnist["target"]
```

```python
X.shape
```

This dataset has 70,000 instances and 784 features

```python
y.shape
```

### Cast target column from string to integer

```python
y = y.astype(np.uint8)
```

# Display an image from dataset

```python
def plot_digit(data):
    image = data.reshape(28, 28)
    plt.imshow(image, cmap = mpl.cm.binary,
               interpolation="nearest")
    plt.axis("off")
```

```python
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(size,size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    n_empty = n_rows * images_per_row - len(instances)
    images.append(np.zeros((size, size * n_empty)))
    for row in range(n_rows):
        rimages = images[row * images_per_row : (row + 1) * images_per_row]
        row_images.append(np.concatenate(rimages, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = mpl.cm.binary, **options)
    plt.axis("off")
```

```python
plt.figure(figsize=(9,9))
example_images = X[100:200]
plot_digits(example_images, images_per_row=10)
plt.savefig("static/images/fig1.png", bbox_inches='tight')
plt.show()
```

### Split data

```python
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
```

### Train Binary Classifier

```python
y_train_7 = (y_train == 7)
y_test_7 = (y_test == 7)
```

### Percent of dataset that is 7

```python
percent_7 = round(sum(y_train_7)/len(y_train_7)*100,2)
print(f'The percent of images that are "7" is: {percent_7}%')
```

```python
sgd_clf = SGDClassifier(max_iter=1000, tol=1e-3, random_state=42)
sgd_clf.fit(X_train, y_train_7)
```

### Measure Accuracy of model using Cross-Validation

```python
cross_val_score(sgd_clf, X_train, y_train_7, cv=3, scoring="accuracy")
```

The accuracy for this model is above 93% for the cross-validation folds tested.

### Measure Accuracy of model at Predicting Non-7 numbers

```python
class ClassifierNot7(BaseEstimator):
    def fit(self, X, y=None):
        pass
    def predict(self, X):
        return np.zeros((len(X), 1), dtype=bool)
```

```python
never_7_clf = ClassifierNot7()
cross_val_score(never_7_clf, X_train, y_train_7, cv=3, scoring="accuracy")
```

The accuracy of classifying non 7 is above 89%. So if you guess that an image is not 7, you would be right 89% of the time.

# Confusion Matrix
 - cross_val_predict()
 - confusion_matrix()

### Use cross_val_predict to return a prediction for each instance in the training set

```python
# Create a set of predictions on training dataset
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_7, cv=3)
```

```python
cm = confusion_matrix(y_train_7, y_train_pred)
print(cm)
```

### Visualize confusion matrix

```python
fig, ax = plot_confusion_matrix(conf_mat=cm,
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True,
                               figsize=(8,8))

plt.savefig("static/images/fig3.png", bbox_inches='tight')
plt.show()
```

### Compare Confusion Matrix with Perfect Model

```python
y_train_perfect_prediction = y_train_7
perfect_cm = confusion_matrix(y_train_7, y_train_perfect_prediction)
```

```python
fig, ax = plot_confusion_matrix(conf_mat=perfect_cm,
                                show_absolute=True,
                                show_normed=True,
                                colorbar=True,
                               figsize=(8,8))

plt.savefig("static/images/fig4.png", bbox_inches='tight')
plt.show()
```

# Precision/Recall Trade-Off

### Get the scores for each instance in the training set using <font color="red">cross_val_predict()</font>

```python
y_scores = cross_val_predict(sgd_clf, X_train, y_train_7, cv=3, method="decision_function")
```

### Get the precision and recall for all possible tresholds using: <font color="red">precision_recall_curve()</font>

```python
precisions, recalls, thresholds = precision_recall_curve(y_train_7, y_scores)
```

### Get the threshold for 90% precision

```python
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]
```

```python
threshold_90_precision
```

```python
y_train_pred_90 = (y_scores >= threshold_90_precision)
```

```python
precision_score(y_train_7, y_train_pred_90)
```

```python
recall_score(y_train_7, y_train_pred_90)
```

# Visualize Precision,  Recall, and Treshold

```python
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.legend(loc="center right", fontsize=16) 
    plt.xlabel("Threshold", fontsize=16)        
    plt.grid(True)                              
    plt.axis([-50000, 50000, 0, 1])             
```

```python
plt.figure(figsize=(12, 8))                      
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)    
plt.plot([2288, 2288], [0.83, 0.9], "r:")         
plt.plot([-50000, 2288], [0.9, 0.9], "r:")      
plt.plot([-50000, 2288], [0.83, 0.83], "r:")

# Adds red points
plt.plot([2288], [0.9], "ro")                   
plt.plot([2288], [0.83], "ro")                
plt.legend(loc='best')
plt.savefig("static/images/fig5.png", bbox_inches='tight')
plt.show()
```

### Precision versus Recall Plot

```python
def plot_precision_vs_recall(precisions, recalls):
    plt.plot(recalls, precisions, "b-", linewidth=2)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1, 0, 1])
    plt.grid(True)

plt.figure(figsize=(8, 6))
plot_precision_vs_recall(precisions, recalls)
# Horizontal line
plt.plot([0.0, 0.83], [0.9, 0.9], "r:")
# Add Horizontal redpoint
plt.plot([0.83], [0.9], "ro")
# Add Vertical redpoint
plt.plot([0.83, 0.83], [0., 0.9], "r:")
plt.savefig("static/images/fig6.png", bbox_inches='tight')
plt.show()
```
