        <li>Support Vector Machine algorithm is a powerful supervised Machine Learning Models capable of performing

        <li>Outlier detection</li>
            </ul>
        </li><br>
        <li>SVM is one of the most popular machine learning models and is well suited for classification of complex datasets that are small and medium sized.</li><br>
        <li>Like logistic regression, support vector machines were designed for binary classification and do not natively support classification tasks with more than two classes.</li><br>
        <li>In this algorithm, data is plotted in n-dimensional space (where n is number of features you have). Then, we perform classification by finding a decision boundary (aka hyperplane in 3d+) that best differentiates the two classes.</li><br>     
        <li><mark style="background-color: yellow;">The objective of the support vector machine algorithm is to find a hyperplane in an N-dimensional space(N â€” the number of features) that distinctly classifies the data points.</mark></li><br>
        <li>Decision boundaries are used in SVM to help sort data into categories for classification.</li><br>     
        <li>Although linear SVM classification is generally efficient and works in many cases, most datasets are not linearly separable.</li><br>

                <li>Polynomial Features</li>
                <li>Polynomial Kernel</li>
                <li>Similarity Features</li>
                <li>Gaussian RBF Kernel</li>  
            </ul>
        </li><br>
        <li>SVM is popular because of the Kernel functions, which helps solve nonlinearity, especially for complex datasets.</li><br>        
        <li>The method used to handle nonlinearity depends on the size of the dataset, it's complexity, and the computational resources available.</li><br>
        <li>Although SVM is commonly used for classification, it can also be used for regression.</li><br>
        <li>Regression is a statistical modeling technique that is used to estimate the relationship between two continuous variables.</li><br>  
        <li>Regression aims to predict the value of a dependent variable based on the value of an independent variable.</li><br>

<center><h1><br><font color="blue">Introduction to Suppor Vector Machine (SVM) Regression.</font></h1></center>
    <hr>
    <ul><h4>What is SVM Regression and How is it Different from Linear Regression.</h4>
        <li>Support vectors machine algorithms are often used for classification problems, but can also be used in regression.</li><br>        
        <li>SVM methods used to predict the value of a dependent variable based on the value of an independent variable are referred to as SVM Regression.</li><br> 
        <li>SVR provides a flexible way to define how much error is acceptable in the regression model and will find an appropriate line (or hyperplane in higher dimensions) to fit the data.</li><br>
        <li>Unlike Linear Regression, SVM Regression minimizes the L2-norm of the coefficient vector and not the squared error.</li><br>        
        <li>The weights of each feature in SVM regression reflect their true contribution, unlike Linear regression, gives the biggest weight to features that are best correlated to the dependent variable.</li><br>     
    </ul>
</div>

<div id="objective3" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Mathematic Intuition of Suppor Vector Machine (SVM) Regression.</font></h1></center>
    <hr>
    <ul><strong>&epsilon; the margin of error in SVM Regression</strong>:
        <li>The objective function of SVM Regression is to minimize the L2-norm of the coefficient vectors:<br><br>
            <strong>MIN &#189;||w||<sup>2</sup></strong>.</li><br>        
        <li>The error term is handled in the models constraints, as expressed:
            <br><br>
            <strong>|y<sub>i</sub> - w<sub>i</sub>x<sub>i</sub>| &#8804; &epsilon; </strong><br><br>
            where <strong>y<sub>i</sub></strong> is the target<br>
            <strong>w<sub>i</sub></strong> is the coefficient<br>
            <strong>x<sub>i</sub></strong> is the predictor (feature)
        </li><br> 
        <li>This is illustrated in <a href="#example1">example 1</a>.</li><br>
        <li>The maximum error <strong>&epsilon; (epsilon)</strong> is the constraint, which is the absolute error less than or equal to the specified margin, and is illustrated in <a href="#example1">example 1</a>.</li><br>
        <li>The <strong>&epsilon; (epsilon)</strong> can be tuned to gain a desired accuracy in the model.</li><br>        
        <li>The red line in <a href="#example1">example 1</a> is the best fit line and the black lines represent the margin of error <strong>&epsilon; (epsilon)</strong>.</li><br> 
    </ul>
    <hr>
    <ul><strong>&zeta;(zeta)</strong>:
        <li>The algorithm does not always work for all data points, as some points may fall outside the margins, as illustrated in <a href="#example1">example 1</a> where you have blue data points above or below the black lines that represent the margin of error <strong>&epsilon; (epsilon)</strong>.</li><br>
        <li>To account for the errors that are larger than <strong>&epsilon; (epsilon)</strong>, we use a slack variable called <strong>&zeta;(zeta)</strong>.</li><br>
        <li><strong>&zeta;(zeta)</strong> is the deviation from the margin, and we we would like to minimize these as much as possible, so we can add these to the objective function.</li><br>           
        <li>Minimize function that takes <strong>&zeta;(zeta)</strong> into account is defined as:
            <br><br>
             <strong><em>MIN &#189;||w||<sup>2</sup> + C * &Sigma;<sup>n</sup><sub>i=1</sub>|&zeta;<sub>i</sub>| </em></strong>.
        </li><br>
        <li>The contraints is also updated to account for &zeta;:
            <br><br>
            <strong>|y<sub>i</sub> - w<sub>i</sub>x<sub>i</sub>| &#8804; &epsilon; + &zeta;<sub>i</sub>| </strong>
        </li><br>
    </ul>
    <hr>
    <ul><strong><code>C</code> hyperparameter</strong>:
        <li>We now have an additional hyperparameter, <strong><code>C</code></strong>, that is used to adjust the tolerance for points outside of the margins &epsilon;</li><br>
        <li>As <strong><code>C</code> &uarr;:</strong> tolerance for points outside of &epsilon; also &uarr;</li><br>
        <li>As <strong><code>C</code> &darr;:</strong> tolerance for points outside of &epsilon; also &darr;:</li><br>
        <li>Finding the best value for the <strong><code>C</code> hyperparameter</strong> is important to obtain the best model.</li><br>
        <li>One method commonly used to identify the optimal <strong><code>C</code> hyperparameter</strong> is through a grid search over <code>C</code> to obtain a scoring metric <em>% within Episolon</em> score.</li><br>           <li><em>% within Episolon</em> measures how many of the total points within the test set falls within our margen of error.</li><br>
        <li>A second method commonly used to identify the optimal <strong><code>C</code> hyperparameter</strong> is through monitoring the <em>Mean Absolute Error (MAE)</em> score.</li><br>
        <li>The <em>Mean Absolute Error (MAE)</em> score generally decreases as <code>C</code> increases.</li><br>
        <li>You can then find the value of <code>C</code> that will maximizes the prediction within a target margin of error with these methods.</li><br>
    </ul>
</div>

<h2>Obtain Array of Support Vectors for Plotting Purposes</h2>

 - Custom function takes the Trained model, Target Array, and Features data.
 - the 

```python
def find_support_vectors(svm_reg, X, y):
    y_pred = svm_reg.predict(X)
    off_margin = (np.abs(y - y_pred) >= svm_reg.epsilon)
    return np.argwhere(off_margin)
```

<div class="alert alert-block alert-warning" style="font-size:11px; border:0.5px solid black; padding:5px">
<center><h1>Code Explained:</h1></center>
    <li><strong>Function Output:</strong> determines the support vectors and returns an array of indices of support vector of shape (n_SV,).</li>
    <li><strong>Purpose:</strong> will be used to identify and mark support vectors when plotted.</li>    
    <ol><strong>Code Line</strong>
        <li><code>def find_support_vectors(svm_reg, X, y)</code>: Starts and names a function with three parameters</li>
        <ul><strong>parameters</strong>
            <li><code>svm_reg</code>: the trained model</li>
            <li><code>X</code>: the feature variables array</li>
            <li><code>y</code>: the target variable array</li>            
        </ul>
        <li><code>y_pred = svm_reg.predict(X)</code>: perform regression on samples in X.</li>
        <li><code>off_margin = (np.abs(y - y_pred) >= svm_reg.epsilon)</code>:
            <ul><strong>Steps</strong>
                <li><code>np.abs(y - y_pred)</code>: The difference between the target value and the predicted y value</li>
                <li><code>svm_reg.epsilon</code>: is the &epsilon; value that was put in the paramater.</li>
                <li><code>(np.abs(y - y_pred) >= svm_reg.epsilon)</code>: If the difference between the actual target and the predicted is greater than the error epsilon, then save that as True, else False. This will save an array with boolean values.</li>            
            </ul>
        </li>
        <li><code>return np.argwhere(off_margin)</code>: Uses <a href="https://numpy.org/doc/stable/reference/generated/numpy.argwhere.html">numpy.argwhere</a>, which is method that returns an array (N, a.ndim) that contains indices where the difference of the target value and the predicted value is greater than the epsilon value.</li>
    </ol>
</div>

```python
# Get support vector data points
svm_reg1.support_ = find_support_vectors(svm_reg1, X, y)
svm_reg2.support_ = find_support_vectors(svm_reg2, X, y)
```

<div class="alert alert-block alert-warning" style="font-size:11px; border:0.5px solid black; padding:5px">
<center><h1>Code Explained:</h1></center>
    <li><strong>Purpose:</strong> Evokes the <code>find_support_vectors</code> function, and save the array of indices.</li>    
    <ol><strong>Code Line</strong>
        <li><code>svm_reg1.support_ = find_support_vectors(svm_reg1, X, y)</code>: indices of support vectors using the svm regression model trained wih &epsilon; = 1.5</li>
        <li><code>svm_reg1.support_ = find_support_vectors(svm_reg1, X, y)</code>: indices of support vectors using the svm regression model trained wih &epsilon; = 0.5</li>
    </ol>
</div>

```python
def plot_svm_regression(svm_reg, X, y, axes):
    x1s = np.linspace(axes[0], axes[1], 100).reshape(100, 1)
    y_pred = svm_reg.predict(x1s)
    plt.plot(x1s, y_pred, "k-", linewidth=2, label=r"$\hat{y}$")
    plt.plot(x1s, y_pred + svm_reg.epsilon, "k--")
    plt.plot(x1s, y_pred - svm_reg.epsilon, "k--")
    plt.scatter(X[svm_reg.support_], y[svm_reg.support_], s=180, facecolors='#FFAAAA')
    plt.plot(X, y, "bo")
    plt.xlabel(r"$x_1$", fontsize=18)
    plt.legend(loc="upper left", fontsize=18)
    plt.axis(axes)
```

        <li><code>plt.scatter(X[svm_reg.support_], y[svm_reg.support_], s=180, facecolors='#FFAAAA')</code>: plots the support vectors as a scatter plot and adds a color</li>
        <li><code> plt.plot(X, y, "bo")</code>: plots the original data as "bo" blue circles</li>

```python
eps_x1 = 1
eps_y_pred = svm_reg1.predict([[eps_x1]])

fig, axes = plt.subplots(ncols=2, figsize=(9, 4), sharey=True)
plt.sca(axes[0])
plot_svm_regression(svm_reg1, X, y, [0, 2, 3, 11])
plt.title(r"$\epsilon = {}$".format(svm_reg1.epsilon), fontsize=18)
plt.ylabel(r"$y$", fontsize=18, rotation=0)

plt.annotate(
        '', xy=(eps_x1, eps_y_pred), xycoords='data',
        xytext=(eps_x1, eps_y_pred - svm_reg1.epsilon),
        textcoords='data', arrowprops={'arrowstyle': '<->', 'linewidth': 1.5}
    )
plt.text(0.91, 5.6, r"$\epsilon$", fontsize=20)
plt.sca(axes[1])
plot_svm_regression(svm_reg2, X, y, [0, 2, 3, 11])
plt.title(r"$\epsilon = {}$".format(svm_reg2.epsilon), fontsize=18)
plt.show()
```
