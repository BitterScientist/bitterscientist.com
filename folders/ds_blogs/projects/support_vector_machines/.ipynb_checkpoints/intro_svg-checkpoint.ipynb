<center><h1><br><font color="blue">Introduction to Support Vector Machines (SVM)</font></h1></center>
    <hr>
    <ul><h4></h4>
        <li>Support Vector Machine algorithm is a powerful supervised Machine Learning Models capable of performing
            <ul>
        <li>Linear classification</li>
        <li>Nonlinear classification</li>
        <li>Regression</li>
        <li>Outlier detection</li>
            </ul>
        </li><br>
        <li>SVM is one of the most popular machine learning models and are well suited for classification of complex datasets that are small and medium sized.</li><br>
        <li>Like logistic regression, support vector machines were designed for binary classification and do not natively support classification tasks with more than two classes.</li><br>
        <li>There is a way to carry out multi-classification problems using binary classification algorithms, but that requires the data be split into multiple binary classification datasets and then fit to a binary classification model.</li><br>
        <li>These methods are called One-vs-Rest and One-vs-One strategies.</li><br>
        <li>In this algorithm, data is plotted a point in n-dimensional space (where n is number of features you have). Then, we perform classification by finding a decision boundary (aka hyperplane) that best differentiates the two classes.</li><br>
        <li>SVM is popular because it uses Kernel function to help solve nonlinearity, and create decision boundaries.</li><br>
        <li>Decision boundaries are used in SVM to help sort data into two or more categories.</li><br>
        <li><mark style="background-color: yellow;">The objective of the support vector machine algorithm is to find a hyperplane in an N-dimensional space(N â€” the number of features) that distinctly classifies the data points. <a href="#image6">Image 6</a></mark></li><br>
        <li>To better understand the mathematics of SVM, the following topics will be briefly discussed:
            <ul>
                <li><a href="#objective2">Points in Space</a></li>
                <li><a href="#objective3">Decision Boundaries</a></li>
                <li><a href="#objective4">Measures of Distance</a></li>              
            </ul>
        </li>
    </ul>
</div>

      <center><h2><br><font color="blue">Decision Boundaries</font></h2></center>
    <ul>
        <li>Is a hypersurface that partitions the underlying vector space into two sets, one for each class . The classifier will classify all the points on one side of the decision boundary as belonging to one class and all those on the other side as belonging to the other class <a href="#image1">Image 1, the line in the left image</a>.</li><br>
        <li>This boundary can be thought of as a line in a 2 dimensional space <a href="#image2">Image 2</a>, and a hyperplane in higher dimensional space.</li><br>
        <li>A hyperplane is a higher-dimensional generalization of a lines and planes, and can be denoted as described in <a href="#equation3">Equation 3.</a></li><br> 
        <li>In D dimensional space, the hyperplane would always be D -1 operator.</li><br>
        <li>The mathematical equation for this decision boundary is called a hyperplane equation.</li>

<div id="objective4" style="font-size:16px; border:1px solid black; padding:10px">
      <center><h2><br><font color="blue">Measures of Distance</font></h2></center>
    <ul>
        <li>The next step is to determine the position of the decision boundary such that the data points are segragated as far as possible, thereby minimizing misclassification of data points.</li><br>
        <li>The distances of a point to any line (ax + by + c = 0) can be determined from  <a href="#equation4">Equation 4</a> and depicted in <a href="#image3">Image 3</a>.</li><br>
        <li>This same formula can be adjusted to calculate the distance of a hyperplane equation from a given point vector &Phi;(<em>x<sub style="vertical-align: sub; font-size: smaller;">0</sub></em>) using <a href="#equation5">Equation 5</a></li><br>
        <li>If a hyperplane is able to differentiate between two groups, this is known as <em>Perfect Separation</em> <a href="#image4">Image 4</a>.</li><br>
        <li>As illustrated in <a href="#image5">Image 5</a>, there are many potential hyperplanes that could be used to differentiate between group, but an optimal/best hyperplane should be selected, whereby it exists as close to the center where the distances is maximum from the closes points that would give the least number of misclassification errors.</li><br>
        <li>The goal of Support Vector Machine is to find this optimal distance that would serve as a decision boundary to
            <ol>
                <li>separate groups</li>
                <li>help classify new data</li>
            </ol>
            This is illustrated in <a href="#image6">Image 6</a></li><br> 
        <li>There are many possible hyperplanes that could be chosen, but the objective of SVM is to find a plane that has the maximum margin (the maximum distance between data points of both classes)</li><br>
        <li>Maximizing the margin distance provides reinforcement so that future data points can be accurately classified.</li><br>        
        <li>Data that falls on either side of the hyperplane can be attributed to a different class, and the dimension of the hyperplane depends on the number of features.</li><br>
        <li>If the number of input features is 2, then the hyperplane is a line. If the input features is 3, then the hyperplane is a two-dimensional plane <a href="#image1">Image 1</a>.  Input features greater than 3 are difficult to image.</li><br>      
        <li>Support vectors are the data points that are closer to the hyperplane and influence the position, orientation and margin size of the hyperplane, as illustrated in <a href="#image6">Image 6</a> and <a href="#image7">Image 7</a></li><br>
        <li>The support vectors are used to maximize the margin of the classifier.</li><br>
        <li>Removing support vectors will change the position of the hyperplane.</li><br>        
        <li><strong>Large Margin Intuition</strong>
            <ul>
                <li>In logistic regression, the output of the linear function falls within a range of <code>[0,1]</code> using a sigmoid function.</li>
                <li>If a data points value is greater than the <code>0.5</code> threshold, then the instance is labeled <code>1</code>, else it is labeled<code>0</code>.</li>
                <li>In SVM, we take the output of the linear function, and if the output is greater than 1, we identify it as one class, and if the output is -1, then we identify it as another class.  Thus the reinforcement range for SVM is <code>[-1,1]</code></li>
            </ul>
        </li>
    </ul>
</div>

<div id="objective5" style="font-size:16px; border:1px solid black; padding:10px">
<center><h1><br><font color="blue">Geometric Intuition: Hyperplanes, Margins, Zeta, C hyperparameter, and Support Vector Planes.</font></h1></center>
    <hr>
    <ul><h4>Hyperplane Sign</h4>
        <li>As described so far, the goal of SVM is to find the plane that best separates the data into separate classes.</li><br>
        <li>For binary classification, the hyperplane cases are either positive or negative.</li><br>
        <li>As illustrated in <a href="#image8">Image 8</a>, the boundaries of a hyperplane are either -1, or +1. The area in between is the margin.</li><br>
        <li>The vector <code>w</code> represents the direction that the decision boundary is moving as the hyperplane location is optimized.  The <code>w</code> vector can move in any direction.</li><br>
        <li>The support vectors are points that lie on either side of the hyperplane lines.</li><br>
    </ul>
    <hr>
    <ul><h4>Mathematical intution for maximizing the margin (Hard and Soft margin SVM).</h4>
        <li>The hyperplanes in <a href="#image8">Image 8</a> can be represented by two equations defined in <a href="#equation6">Equation 6</a>.</li><br>
        <li>To maximize the distance between these two hyperplanes, we first substract the negative hyperplane line to that of the positive hyperplane <a href="#equation7">Equation 7</a>.</li><br>
        <li>We then take the length of the vector w, which is defined in equation <a href="#equation8">Equation 8</a> and divide this to both sides of <a href="#equation7">Equation 7</a>.</li><br>
        <li>By normalizing Equation 7 by 8, we get <a href="#equation9">Equation 9</a>, which the left side of the equation describes the distance between the positive and negative hyperplane. This is the margin we want to maximize.</li><br>
        <li>Thus the optimization function for SVM would be described in <a href="#equation10">Equation 10</a>, and the condition implies that the points are perfectly linearly seperable as has been illustrated in <a href="#image8">Image 8</a> (and all other images so far).</li><br>
        <li>This approach is called the <strong>Hard-Margin SVM</strong> method, that is defined in <a href="#equation11">Equation 11.</a></li><br>
        <li>The Hard-Margin method minimizes the expression <strong>&#189;w<sup style="vertical-align: sup; font-size: smaller;">T</sup>w</strong>, because this is equal to <strong>&#189;*&#x2016;w&#x2016;</strong>, which is differentiable. Optimization algorithms work better when a function is differentiable.</li><br>
        <li>The problem with this approach is that in the real world, the dataset is rarely perfectly separable, as illustrated in <a href="#image9">Image 9</a>.</li><br>
        <li>Often, data from one class is located within the margins of the hyperplane, or on the side of the opposite class.  This is not perfectly separable data.</li><br>    
        <li>A common approach address these real work cases is using <strong>Soft Margin SVM</strong> described in <a href="#equation12">Equation 12</a>.</li><br>
    </ul>
    <hr>
    <ul><h4>Understanding Zeta in Soft Margin SVM.</h4>
        <li><a href="#equation12">Equation 12</a> introduces a slack variable called <code>Zeta</code> that is also illustrated in <a href="#image9">Image 9</a>.</li><br>
        <li><mark style="background-color: bisque; color: black;">The Zeta variable represents the distance of the misclassified data point from its actual plane.</mark></li><br>
        <li>If the zeta value for a data point is <code>(0,1]</code> then this is referred to as a <strong>margin violation</strong>.</li><br>
        <li>If the zeta value for a data point is <code>zeta > 1</code> then this is referred to as a <strong>misclassification</strong>.</li><br>  
        <li><a href="#image10">Image 10</a> illustrates zeta in action.</li>
    </ul>
<hr>
    <ul><h4>Understanding the Regularization term <code>C</code> in Soft Margin SVM.</h4>
        <li><a href="#equation12">Equation 12</a> also introduced a <code>C</code> parameter.</li><br>
        <li>The <code>C</code> parameter helps tune the model to avoid overfitting and underfitting.</li><br>   
        <li><mark>As <code>C</code> decreases, so does the tendency of the model to underfit. You approach a large margin and you allow the contraints to be easily ignored.</mark></li><br>
        <li><mark>As <code>C</code> increases, so does the tendency of the model to overfit. You approach a narrow margin and you make the contraints hard to ignore.</mark></li><br> 
        <li><mark>As <code>C</code> reaches infinity, all contraints are enforced, and you have a hard margin</mark></li><br>
        <li>There is only one <code>C</code> parameter.</li><br>
    </ul>
    <ul><h4>Understanding the Support Vector Planes Boundary.</h4>
        <li><a href="#equation13">Equation 13</a> describes the support vector planes using the constant value <code>k</code>.</li><br>
        <li><code>k</code> is either positive or negative, and are equally distant from the decision boundary.</li><br>
        <li><code>k</code> can take on any constant value as long as both planes are equally distant, although the <code>k = +-1 </code> because this simplifies the mathematical calculations.</li><br>    
    </ul>       
</div>

<center><h1><br><font color="blue">Optimization and Loss Function.</font></h1></center>
    <hr>
    <ul><h4>Optimization</h4>
        <li>Learning an SVM has been formulated as a constrained optimization problem over <strong><code>w</code></strong>and <strong><code>Î¾</code></strong>.</li><br>
        <li>This optimization problem is expressed in <a href="#equation14">Equation 14</a>.</li><br>
        <li>This learning problem then is equavilent to the unconstrained optimization problem over the <code>w</code>, as expressed in <a href="#equation15">Equation 15</a></li><br>
    </ul>
    <hr>
    <ul><h4>Loss Function</h4>
        <li>The loss function used in SVM is the <mark>hinge loss</mark>.</li><br>
        <li>Hings loss is a function whose value is non-zero until it reaches a certain point "z", and after that point, the hinge loss value is equal to zero, as illustrated in <a href="#image11">Image 11</a>.</li><br>
        <li>The loss term is in the loss function is the c hyperparameter and the zeta term, this is the second term in <a href="#equation15">Equation 15</a> and also highlighted in <a href="#image11">Image 11</a>.</li><br>
        <li>If the loss function is less than 1, then the point is misclassified, and contributes to the loss. This is illustrated in <a href="#image11">Image 11</a>, as all points that would be to the left of the dash line.</li><br>
        <li>If the loss function is equal to or greater than 1, then the point is correctly classified and does not contributes to the loss. This is illustrated in <a href="#image11">Image 11</a>, as all points that would be to the right of the dash line.</li><br>
        <li>Another way to think about the data points position contribute to the loss term in <a href="#equation16">Equation 16</a> and in <a href="#image12">Image 12</a>.</li><br>
        <li>Data points can be organized into three cases:
            <ol>
                <li>A data point is in the positive plane <strong>(w<sup style="vertical-align: sup; font-size: smaller;">T</sup>x + b) > 0</strong>, and has a <strong>positive y value</strong> => y * plane = 1 * 1 = positive classification.</li>
                <li>A data point is in the negative plane <strong>(w<sup style="vertical-align: sup; font-size: smaller;">T</sup>x + b) &#60; 0</strong>, and has a <strong>negative y value</strong>  => -y * -plane = -1 * -1 = positive classification.</li>
                <li>A data point is in the negative plane <strong>(w<sup style="vertical-align: sup; font-size: smaller;">T</sup>x + b) &#60; 0</strong>, and has a <strong>positive y value</strong> => y * -plane = 1 * -1 = negative (mis-) classification.</li>
            </ol>
        </li><br>
        <li>From the three cases described, only the third case would contribute to the loss and violate the margin constraint.</li><br>
        <li><a href="#equation16">Equation 16</a> provides another way to evaluate the loss term in the loss function using the cases described.</li><br>
        <li>Taken together, what we want to minimize is the zeta term (misclassifications) and we do so using <a href="#equation12">Equation 12</a>.</li>
    </ul>
