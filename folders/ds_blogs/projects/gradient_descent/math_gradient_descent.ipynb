<center><h1>The Objective (Big Picture) of Gradient Descent</h1></center>

 - Image you are are the top of this mountain, are blind, and want to reach the town at the base of this mountain. How would you do this?
 - One way is to use your feet, and tap in all directions around you, until you feel a direction with a slope that is descending.
 - You do this with each step, until you gradually reach the bottom.
 - Yes this will take you a very long time!
 - This is what Batch Gradient Descent seeks to do. Get you to the bottom of your cost function so that your trained model has the lowest rate of false predictions.

<ul>
<li><font color="blue">The theta 1</font> is the next step you would take as you trek down the moutain to reach the town at it's base.</li><br>
    <li><font color="blue">Theta zero</font> is the current position.</li><br>
    <li><font color="blue">Alpha</font> is how big of a step you take, a small step is conservative, which means it will take you longer to reach the base, but a big step can take you in the wrong direction all together. As explained later.</li><br>
    <li><font color="blue">the nabla</font> (upside down triangle) is the gradient/slope, so this means you will go down the direction with the steepest slope.</li><br>
    <li><strong>Why do we subtract the size of the step and the direction with the steepest slope to the current position?</strong></li><br><br>
    <li><strong>Answer:</strong> our current position on this mountain (or cost function) is presumably high, so we want to go down, so we subtract (or descent) down the mountain/cost function by <em>subtracting</em> the step we take, times the gradient from our current position</li>
</ul>
</div>

<h1>Cost Function Explained</h1>

<div style="font-size:18px; border:1px solid black; padding:10px">
<ul>
    <li>The <font color="red">Cost Function/Loss Function</font> evaluates the performance of the machine learning algorithm.</li><br>
    <li>The <font color="red">Loss Function</font> calculates the error for a single instance.</li><br>
    <li>The <font color="red">Cost Function</font> is the average of the loss function for the set of training instances.</li><br>
</ul>
<br>
For the formula below:
<ul>
    <li>The <font color="red">N</font> is the total number of instances in the dataset.</li><br>    
    <li>The <font color="red">Y^1</font> is the predicted value for a specific instance.</li><br>       
    <li>The <font color="red">Y</font> is the actual value for a specific instance.</li><br>     
</ul>        
</div>

<h1>How the cost function is minimized</h1>
<br>
<div style="font-size:18px; border:1px solid black; padding:10px">
<ul>
    <li>The Cost function is essentially of the <strong><code>Y = X^2</code></strong>, which when plotted, is a parabola (such as the gif below).</li><br>
    <li>The point in this parabola that has the lowest cost is the minimum.</li><br>
    <li>Gradient Descent helps fund this location.</li><br>
    <li>If we draw tangent lines as we move throuch each point in the parabola (each y value, aka the prediction value), we can get the gradient (or steepness of the slope).</li><br>
    <li>As the steepness approaches zero, we approach the lowest cost in our function, as illustrated in the tangent lines in the gif below.</li>
</ul>      
</div>
