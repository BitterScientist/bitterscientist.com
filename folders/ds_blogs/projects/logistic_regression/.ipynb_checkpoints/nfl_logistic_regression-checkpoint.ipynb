<div style="font-size:16px; border:1px solid black; padding:10px">
    <center><h1><strong>Part 1: Provide an Introduction to Logistic Regression and Model Assumptions.</strong></h1></center>
    <ul>
        <li><mark>Logistic regression is regression algorithm that can be used for classification problems.</mark></li><br>        
        <li>Logistic regression is also known as <em>Logit Regression</em></li><br>
<li>There are three major types of logistic regression:
    <ul>
        <li><em>Binomial Logistic Regression</em>: there are only two possible outcomes (yes/no, cat 1 or cat 2).</li>
        <li><em>Multinomial Logistic Regression</em>: there are three or more possible outcomes without order. Example: planet, start, comet</li>
        <li><em>Ordinal Logistic Regression</em>: there are three or more possible outcomes with ordering. Example: movie ratings from boring, okay, good, great.</li>         
    </ul>       
        <li>The focus of this post is the binomial logistic regression</li><br>
        <li><mark>Binomial Logistic regression is used to explain the relationship between a dependent binary variable and one or more continuous or categorical data (nominal, ordinal, interval, or ratio-level) independent variables.</mark></li><br>
        <li><strong>In other words: </strong>Binomial Logistic regression is used to estimate the probability that an instance belongs to a specific class.  Does the observation fall into one of two categories?</li><br>
        <li>A decision boundary is used to set a threshold for prediction making.  For binary classification, if an instance is less than 50%, it is classified in one category, and if is greater or equal to 50%, it is classified in the 2nd cagetory.</li><br>  
        <li>Decision boundaries can be linear or non-linear. Polynomial order can be increased to get a more complex decision boundary</li><br>
        <li>Example of logistic regression problems:
            <ul>
                <li>What is the probability that email is spam? (spam/ not spam)</li>
                <li>What is the probability that transaction is fraud? (yes/no)</li>
            <li>What is the probability that this tissue sample is cancerous? (yes/no)</li> 
            </ul></li><br>
        <li>Instances greater than 50% are labeled positive class, "1" .</li><br>
        <li>Instances less than 50% are labeled negative class, "0".</li><br>      
        <li><strong>Major Assumptions</strong></li><br>
        <ol>
        <li>Dependent variable should be dichotomous in nature (yes/no, present/absent, male/female etc).  If your dependent variable is continuous, you need to carry out a multiple regression analysis instead.  
            <br><font color="red">The dependent variable in this study is dichotomous (wide receiver or not wide receiver).</font></li><br>
        <li>You have one or more independent variables, which can be either continuous variables (i.e., an interval or ratio variable) or nominal variables.
            <br><font color="red">The independent variables for this study are both continuous variables (player weight in lbs and player height in inches).</font>
            </li><br>
        <li>There should be independence of observations.
            <br><font color="red">Each instance in this dataset are individual players. Each instance is independent.</font>            
            </li><br>
        <li>The categories of the dichotomous dependent variable and all your nominal independent variables should be mutually exclusive and exhaustive.
            <br><font color="red">There are no nominal independent variables in this study.</font>
            </li><br>
        <li>There should be a bare minimum of 15 cases per independent variable (although some recommend as high as 50 cases per independent variable).
            <br><font color="red">There are over 2652 observations, well over the minimum for this method.</font>
            </li><br>        
        <li>There needs to be a linear relationship between any of the continous independent variables and the logit transformation of the dependent variable.  The Box-Tidwell (1967) procedure can be used to test for linearity.
            <br><font color="red">Box-Tindell test was conducted using SPSS as described by <a href="https://statistics.laerd.com/premium/spss/blr/binomial-logistic-regression-in-spss-6.php#dealing-with-nonlinearity">Laerd Statistics</a></font>
        </li><br>
        <li>Your data must not show multicollinearity.
            <br><font color="red">A Variable Inflation Factor test was conducted on the weight and height features and the result suggest high multicollinearity. To resolve this issue, this binary logistic regression will only use 1 of these two features, the weight of players.</font>
        </li><br>
        <li>There should be no significant outliers, high leverage points or highly influential points.
            <br><font color="red">Boxplot and z-score analysis were conducted on the key independent variables and no evidence to suggest outliers exist in either of these features.</font>
        </li><br>            
        </ol>
    </ul>
</div>

<h2>Test for Multicollinearity Using Variable Inflation Factors (VIF)</h2>

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
X = nfl_df.iloc[:,3:5]
calc_vif(X)
```

<div style="font-size:16px; border:1px solid black; padding:10px">
    <center><h1><strong>Comments on VIF:</strong></h1></center>
    <ul>
        <li>The VIF is 1 divided by (1 - R^2).</li><br>
        <li>The VIF starts at 1 and there is no upper limit.</li><br>
        <li>If VIF is 1, then no correlations between independent variables.</li><br>
        <li>If VIF exceeds 5, it indicates multicollinearity between the independent variables.</li><br>
        <li>In this case, the VIF for both variables are higher than 5 and suggest multicollinearity, as a result, only one variable will serve as the independent feature. <strong>weight</strong></li>
    </ul>
</div>

<h2>Encoding Target Variable</h2>

```python
# Create target (Y) for players that are wide receivers
nfl_df["target"] = np.where(nfl_df['Position'] == 'WR', 1, 0)
```

### Prepare Data for plotting and linear regression

 - np.c_ method translates slice objects to concatenation along the second axis.
 - Stack 1-D arrays as columns into a 2-D array.
 - Documentation as of 5/16/20: https://numpy.org/devdocs/reference/generated/numpy.c_.html
 - This is important for matrix calculations

```python
# Assign data to dependent and independent variables
X = np.c_[nfl_df["Weight_lbs"]]
y = np.c_[nfl_df["target"]]
```

<h2>Train Logistic Regression Model</h2>

```python
from sklearn.linear_model import LogisticRegression
# Split and Train Data
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=0)

logreg = LogisticRegression(solver="lbfgs", random_state=42)
#log_reg.fit(X, np.ravel(y))
logreg.fit(X_train, np.ravel(y_train))
```

<h2>Visualize the Model's Estimated Probabilities</h2>

```python
plt.figure(figsize=(10, 5))
X_new = np.linspace(0, X.max(), 1000).reshape(-1, 1)
y_proba = logreg.predict_proba(X_new)

plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Wide Receiver")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Wide Receiver")
plt.legend()
plt.xlabel("Player Weight (lbs)")
plt.ylabel("Probability")
plt.savefig("static/images/model_estimated_prob.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
# make predictions based on weight and height
DeSean_Jackson = [177]
Marshal_Yanda = [305]
two_tests = logreg.predict([DeSean_Jackson, Marshal_Yanda])
print(f'The examples tests the weight of NFL players\nMarris Harrison is predicted: {two_tests[0]}\nMarshal Yanda is predicted {two_tests[1]}')
```

<h2>Evaluate Model Using Confusion Matrix</h2>

```python
# create an array that contains the predictions for the test set
y_pred = logreg.predict(X_test)
```

```python
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)

class_names=[0,1] # name  of classes
fig, ax = plt.subplots(figsize=(10,8))
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.savefig("static/images/confusion_matrix.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

<div style="font-size:16px; border:1px solid black; padding:10px">
    <center><h1><strong>Comments:</strong></h1></center>
    <ul>
        <li>The Accuracy tells us how often is the model correct: (True positive + True negative)/100</li><br>     
        <li>In this model the accuracy is 88%, which is largely due to the models prediction of false negatives but looking at the confusion matrix, we see that it did not correctly classify 75 of the true positives</li><br>
        <li>The precision score tells us how often the model's prediction is correct (True Positives/Total Predictions).</li><br>
        <li>The recall score tells us how good was the model at capturing all the real positive values in the dataset (True Positive/(True Positive + False Negative)).</li><br>
        <li>The recall score in this example if very low at 12%, which suggest this is not a good model at classifying wide receivers using weight as the only independent feature.</li><br>    
    </ul>
</div>

<h2>Evaluate Model Using Reciever of Operating Characteristics (ROC) Curve</h2>

```python
fig, ax = plt.subplots(figsize=(10,8))
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="Data 1, auc="+str(auc.round(4)))
plt.legend(loc="upper left")
plt.xlabel("Specificity (1 - True Negative Rate)")
plt.ylabel("True Positive Rate (Recall)")
plt.savefig("static/images/roc_curve.png", bbox_inches='tight', dpi=300)
plt.show()
```

<div style="font-size:16px; border:1px solid black; padding:10px">
    <center><h1><strong>Comments:</strong></h1></center>
    <ul>
        <li>The ROC is a plot of the true positive rate against the false positive rate.</li><br>
        <li>The plot shows the tradeoff between sensitivity and specificity.</li><br>
        <li>An ACU score of 1 represents a perfect classifer, while a curve of 0.5 represents a useless classifier.</li><br>
        <li>In this model, the ACU score is 0.83, which is a decent classifer.</li><br>        
    </ul>

    <center><h1><strong>Demonstrate How to Use Decision Boundaries in Logistic Regressions.</strong></h1></center>
    <ul>
        <li>A fundamental property of logestic regression is the decision bondary.</li><br>
        <li>The decision boundary creates a cutoff for when one class begins, and the other begins.</li><br>
        <li>Decisions boundaries can be linear, non-linear, and polynomial decision boundaries can be developed to more complex models.</li><br>
        <li>Two decision boundaries are used blow.<li>
        <li>A decision boundary using the <a href="#image8"><code>predict_proba()</code></a> method.</li><br>
        <li>A decision boundary displaying two features <a href="#image9">petal width and petal length</a>.</li><br>
    </ul>
</div>

<div id="image8">
</div>

```python
X_new = np.linspace(0, X.max(), 1000).reshape(-1, 1)
y_proba = logreg.predict_proba(X_new)
decision_boundary =  X_new[y_proba[:, 1] >= 0.5][0]

plt.figure(figsize=(12, 6))
plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")
plt.plot([decision_boundary, decision_boundary], [-1, 2], "k:", linewidth=2)
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Wide Receiver")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Wide Receiver")
plt.text(decision_boundary+25, 0.10, "Decision\nboundary", fontsize=14, color="k", ha="center")
plt.arrow(decision_boundary, 0.08, -0.3, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')
plt.arrow(decision_boundary, 0.92, 0.3, 0, head_width=0.05, head_length=0.1, fc='g', ec='g')
plt.xlabel("Player Weight (lbs)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, X.max(), -0.02, 1.02])
plt.savefig("static/images/decision_boundaries_probability.png", bbox_inches='tight', dpi=300)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

X = nfl_df[['Height_in', 'Weight_lbs']].values
y = (nfl_df["target"] == 1).astype(np.int)

logreg = LogisticRegression(solver="lbfgs", C=10**10, random_state=42)
logreg.fit(X, y)

x0, x1 = np.meshgrid(
        np.linspace(nfl_df.Height_in.min(), nfl_df.Height_in.max(), 500).reshape(-1, 1), #xaxis
        np.linspace(nfl_df.Weight_lbs.min(), nfl_df.Weight_lbs.max(), 200).reshape(-1, 1), #yaxis
    )
X_new = np.c_[x0.ravel(), x1.ravel()]

y_proba = logreg.predict_proba(X_new)

plt.figure(figsize=(15, 8))
plt.plot(X[y==0, 0], X[y==0, 1], "bs")
plt.plot(X[y==1, 0], X[y==1, 1], "g^")

zz = y_proba[:, 1].reshape(x0.shape)
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)


left_right = np.array([nfl_df.Height_in.min(), nfl_df.Height_in.max()])
boundary = -(logreg.coef_[0][0] * left_right + logreg.intercept_[0]) / logreg.coef_[0][1]

plt.clabel(contour, inline=1, fontsize=12)
plt.plot(left_right, boundary, "k--", linewidth=3)
plt.text(65, 350, "Not Wide Receiver", fontsize=14, color="b", ha="center")
plt.text(80, 150, "Wide Receiver", fontsize=14, color="g", ha="center")
plt.xlabel("Height (in)", fontsize=14)
plt.ylabel("Weight (lbs)", fontsize=14)
plt.axis([nfl_df.Height_in.min()-1, nfl_df.Height_in.max()+1, nfl_df.Weight_lbs.min()-15, nfl_df.Weight_lbs.max()+5])
plt.savefig("static/images/decision_boundaries_2features.png", bbox_inches='tight', dpi=300)
plt.show()
```
