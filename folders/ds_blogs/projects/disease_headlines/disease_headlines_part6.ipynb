# Visualize US Clusters

```python
# Make an empty map
figure = folium.Figure(width=1000, height=500)
mapp = folium.Map(location=[35, -100], tiles="OpenStreetMap", zoom_start=4).add_to(figure)
 
# I can add marker one by one on the map
for i in range(0,len(us_df)):
    folium.CircleMarker([us_df.iloc[i]['Latitude'], us_df.iloc[i]['Longitude']], popup=us_df.iloc[i]['City'],
                        radius=3,
                        color=cluster_colors[us_df.iloc[i]['Cluster']], fill=True, 
                        fill_color=cluster_colors[us_df.iloc[i]['Cluster']]
                 ).add_to(mapp)

# Show map
mapp
```

### Analyze the content within the largest US clusters

```python
# Uses the same functions defined earlier
us_groups = us_df.groupby('Cluster')
us_sorted_groups = sorted(us_groups, key=lambda x: len(x[1]),
                         reverse=True)
for _, group in us_sorted_groups[:5]:
    sorted_group = sort_by_centrality(group)
    for headline in sorted_group.Headline.values[:5]:
        print(headline)
    print('\n')
```
