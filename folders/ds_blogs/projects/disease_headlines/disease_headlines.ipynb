# Function to great regexes for matching

```python
def data_to_regex(name):
    # ASCII transliterations of Unicode text
    decoded_name = unidecode.unidecode(name)
    # If the name is not already in ASCII characters, then it will convert it
    if name != decoded_name:
        regex = fr'\b({name}|{decoded_name})\b'
    # if it is already in ASCII characters, then it
    # returns it as a regular expression pattern objects, which can be used for pattern matching
    # flags=re.IGNORECASE ignores case    
    else:
        regex = fr'\b{name}\b'
    return re.compile(regex, flags=re.IGNORECASE)
```

### Function that will compare each headline and will either return the name of the country or city, or if it is not present, it will return none

```python
# Function takes in a string, and a dictionary to compare
def get_data_in_text(text, dictionary):
    # iterate through each key and value pair in the dictionary (name of the country or city, and the regex object)
    for regex, name in sorted(dictionary.items(),
                              key=lambda x: x[1]):
    # each individual regex object is used to see if it matches with the text
    # if it is found, then it will return the text
        if regex.search(text):
            return name
    # if it is not found, then None will be returned
    return None
```

# Fix Erroneaous City Output

 - Create a function that returns a set containing all the cities listed in a headline.
 - Apply function to city column and create a new column that contains all the cities found in a given headline
 - Create a new column that contains the count of the number of cities counted in a headline
 - Determine the number of headlines with more than one city

```python
# The headlines will be a parameter
def get_cities_in_headline(headline):
    # Create an empty set (sets do not contain duplicates)
    cities_in_headline = set()
    # Iterate through each city_to_name dictionary that was created earlier
    for regex, name in city_to_name.items():
    # Use regex.search() to find all instances in the headline that match        
        match = regex.search(headline)
    # if there is a match (True)   
        if match:
                # then it will check if the first index character is uppercase      
            if headline[match.start()].isupper():
                # If it is uppercase, it will add the name to the set
                cities_in_headline.add(name)
    #  Return the set
    return list(cities_in_headline)
```

```python
# Apply the get_cities_in_headline(headline) to the df and create a new column
df['Cities'] = df['Headline'].apply(get_cities_in_headline)
```

```python
# Then create a new column that counts the number of cities listed in each headline
df['Num_cities'] = df['Cities'].apply(len)
```

# Determine the number of optimal clusters for the coordinates using the elbow curve method

- Estimate an appropriate value for K using a technique known as the Elbow method
- The Elbow method relies on the calculated inertia of the data (the sum of the squared distances between each point and its closest K-means center). 
- Inertia estimates total dispersion around our K computed means
- This value is used as an indicator to determine if K is too high or too low. 
- If k = 1, the inertia will be very high, so will the dispersion, because a lot of the points will be far from the mean.
- If k -> len(dataset), the inertia will be very low, the disperion will approach zero, but no real 'clusters', just a scatter plot
- Inertia can be plotted against the number of clusters
- Inertia is automatically computed for each KMeans object. 
- This calculation can be accessed using the <code>_inertia</code> variable
- The point at which the curve resembles a bent elbow is considered an optimal cluster number
- Each kink in the curve represents a potential centroid in the data, but the most 'elbow' like point is considered the most optimal

```python
coordinates = df[['Latitude', 'Longitude']].values
# Will test for 10 potential clusters
k_values = range(1, 10)
inertia_values = []
for k in k_values:
    inertia_values.append(KMeans(k).fit(coordinates).inertia_)


# Plot elbow line
plt.plot(range(1, 10), inertia_values)
plt.xlabel('K')
plt.ylabel('Inertia')
plt.title('The Elbow Method')

# Add a marker around potential cluster number
x_marker = 3
y_marker = inertia_values[2]
plt.scatter(x_marker, y_marker, s=100, color='r')
plt.show()
```

# DBSCAN: A Clustering Algorithm for Grouping Data Based on Spatial Density

 - DBSCAN clustering data composed of curving and dense shapes. 
 - DBSCAN does not require an approximation of the cluster 
 - DBSCAN can filter random outliers located in sparse regions of space.
 - DBSCAN cannot be reapplied to new data, a new DBCAN would need to be applied to the union of the old and new data.
 - DBSCAN does not depend on the mean

```python
# Function that defines a NumPy-based great-circle metric
def great_circle_distance(coord1, coord2, radius=3956):
    if np.array_equal(coord1, coord2):
        return 0.0

    coord1, coord2 = np.radians(coord1), np.radians(coord2)
    delta_x, delta_y = coord2 - coord1
    haversin = m.sin(delta_x / 2) ** 2 + np.product([m.cos(coord1[0]),
                                                   m.cos(coord2[0]),
                                                   m.sin(delta_y / 2) ** 2])
    return  2 * radius * m.asin(haversin ** 0.5)
```

## DBSCAN assumptions

 - min_samples: The global cities cluster around at least 3 cities
 - eps: Clusters are on average no more than 250 miles apar

### Use DBSCAN to cluster data

```python
# Assign great_circle_distance to variable
metric = great_circle_distance
# Set eps and min_samples parameters and invoke DBSCAN
dbscan = DBSCAN(eps=250, min_samples=3, metric=metric)
# Apply instantiated DBSCAN to coodinates dataframe and 
# save output to a new column in df dataframe
df['Cluster'] = dbscan.fit_predict(coordinates)
```

# Compare US data to the rest of the world

- re-cluster US cities by modifying epsilon parameter

```python
#  This function will map a country code to a city
def get_country_code(city_name):
    city = max(gc.get_cities_by_name(city_name),
               key=lambda x: list(x.values())[0]['population'])
    return list(city.values())[0]['countrycode']

# Apply the country code a new column in the entire dataframe
# This country code will then be used to filter out US
df['Country_code'] = df.City.apply(get_country_code)
```

```python
# Separate US data from other countries
us_df = df[df.Country_code == 'US']
df_not_us = df[df.Country_code != 'US']
```

### Re-clustering function

```python
# Re-clustering function that takes in a dataframe and the eps: 
# The eps defines the distance between clusters
def re_cluster(input_df, eps):
    input_coord = input_df[['Latitude', 'Longitude']].values
    dbscan = DBSCAN(eps=eps, min_samples=3,
                    metric=great_circle_distance)
    clusters = dbscan.fit_predict(input_coord)
    input_df = input_df.assign(Cluster=clusters)
    return input_df[input_df.Cluster > -1]
```

```python
# Apply function to US filtered data and all other countries
# Reduce the disntace (eps) for the US so to get a tighter density
df_not_us = re_cluster(df_not_us, 250)
us_df = re_cluster(us_df, 125)
```

### Extract the headlines that represent the most central location in the cluster

 - This assumes that these would be the most represenative headlines
 - Mitigates having to read each headline individually

### Function to get the estimated center of a cluster

```python
# Computing cluster centrality
# This function essentially just gives us the average coordinate per cluster
# This function takes a grouped dataframe as a parameter
# It then outputs the dataframe with a new column containing 
# the estimated distance to the central coordinate
def compute_centrality(group):
    group_coords = group[['Latitude', 'Longitude']].values
    center = group_coords.mean(axis=0)
    distance_to_center = [great_circle_distance(center, coord)
                          for coord in group_coords]
    group['Distance_to_center'] = distance_to_center
```

### Function that prints out 5 headlines that are near the center of the cluster

```python
# Finding the central headlines in largest cluster
# Function takes in a grouped dataframe
#  Applies the function that computes the estimated center coordinates
# It then sorts the dataframe by the distance to the center
# Returns new dataframe
def sort_by_centrality(group):
    compute_centrality(group)
    return group.sort_values(by=['Distance_to_center'], ascending=True)

# This code applies these functions on the largest_group data that 
# was extracted earlier.
# It then prints out the 5 headlines closest to the estimated center of the
# Cluster
largest_group = sort_by_centrality(largest_group)
for headline in largest_group.Headline.values[:5]:
    print(headline)
```

#  Determine if Zika Headlines Cluster Along Tropical Regions

### Function that takes in all the headlines and returns only those that have Zika

```python
# Function takes in headlines and then returns the count of headlines with Zika
def count_zika_mentions(headlines):
#   Regular expression will match word "Zika" in a headline, and will be be case-insensitive.
    zika_regex = re.compile(r'\bzika\b',
                            flags=re.IGNORECASE)
    zika_count = 0
    for headline in headlines:
        if zika_regex.search(headline):
            zika_count += 1

    return zika_count
```

### Visualize Clusters that contained Zika in >50% of the headlines

```python
figure = folium.Figure(width=1000, height=500)
mapp = folium.Map(location=[25, -100], tiles="OpenStreetMap", zoom_start=2, max_bounds=True).add_to(figure)

# Use a for loop to iterate over both US and global clusters.
for _, group in sorted_groups + us_sorted_groups:
    headlines = group.Headline.values
    zika_count = count_zika_mentions(headlines)
#   Conditional statement will filter for clusters where Zika is mentioned in more than 50% headlines.
    if float(zika_count) / len(headlines) > 0.5:
        for i in range(0,len(group)):
            folium.CircleMarker([group.iloc[i]['Latitude'], group.iloc[i]['Longitude']], popup=group.iloc[i]['City'],
                                radius=3,
                                color=cluster_colors[group.iloc[i]['Cluster']], fill=True, 
                                fill_color=cluster_colors[group.iloc[i]['Cluster']]
                         ).add_to(mapp)

# Show map
mapp
```
