### Jupyter Notebook Step 4: Exploratory data analysis

```python
order = springer.events.value_counts().index

sns.set(rc={'figure.figsize':(25,15)})
sns.set(font_scale=1.5)
result_events = sns.countplot(x='events', hue = "batter", data=springer, order=order)
result_events.set_xticklabels(result_events.get_xticklabels(), rotation=45)
fig = result_events.get_figure()
```
