<div id="objective2" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">$\chi^2$ Distribution</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics</u></strong>
        <li>Suppose we have a <em>k</em> independent standard normal varibles where each variable $Z_1$ .. $Z_k$ has a ~ N(0,1), then the sampling distribution for the same varibles squared $Z_1^2$ .+. $Z_k^2$ is called a chi-sqaured distribution, with k degress of freedom (df).</li><br>
        <li>The degrees of freedom determine the shape of the curve and as a result, how the probabilities and quantiles are obtained.</li><br>
        <li>The $\chi^2$ is denoted with this degrees of freedom in its subscript $\chi^2_k$</li><br>
        <li>The shape of the $\chi^2$ distribution is determined solely by its degrees of freedom.</li><br>
        <li>$\chi^2$ values only take on positive values.</li><br>
        <li>The distribution is right-skewed and becomes less skewed as the degrees of freedom increases.</li><br>
        <li>The mean is <em>k</em></li><br>
        <li>The mode is <em>k - 2</em></li><br>
        <li>The median is approximately $k(1 - \frac{2}{9k})^3$</li><br>
        <li>The mean, median and mode are close to one another with then values of k is large.</li><br>
        <li>The variance of the distribution is 2k</li><br>
        <li>The standard deviation is $\sqrt{2k}$</li><br>
        <li>The quantiles for the $\chi^2$ distribution is denoted in the subscript $\chi^2_k,p$ and corresponds to the quantile that cuts off an area or probability of the <em>p</em> in the right tail.</li><br>
        <li>The quantity can be found in R command <code>qchisq(p, k, lower.tail=FALSE)</code></li><br>
    </ul>
</div>

## Quantiles from the ùúí2 distribution

## Example 1 $\chi^2_{10,0.25}$

 - the 10 is the degrees of freedom
 - the 0.25 is the percentile, but since we will only get the right side, then it represents the 75th percentile

```r
# examples for the 75th percentile
qchisq(.25, 10, lower.tail=FALSE)
```

```r
#create density curve
curve(dchisq(x, df = 10), from = 0, to = 40,
main = 'Chi-Square Distribution (df = 10)',
ylab = 'Density',
lwd = 2)

#create vector of x values
x_vector <- seq(12.5488613968894, 40)

#create vector of chi-square density values
p_vector <- dchisq(x_vector, df = 10)

#fill in portion of the density plot from 0 to 40
polygon(c(x_vector, rev(x_vector)), c(p_vector, rep(0, length(p_vector))),
        col = adjustcolor('red', alpha=0.3), border = NA)
```

## Example 2 $\chi^2_{30,0.01}$

 - the 30 is the degrees of freedom
 - the 0.01 is the percentile, but since we will only get the right side, then it represents the 99th percentile

```r
# examples for the 75th percentile
qchisq(.01, 30, lower.tail=FALSE)
```

## Tail Probabilities from the ùúí2 distribution

 - For a given $\chi^2$ value, what is the tail probability/area to the right or left<br>
 - For tail probabilities to the right: $P(\chi^2_k > x)$<br>
 - For tail probabilities to the left: $P(\chi^2_k < x)$<br> 
 - <code>pchisq(x, k, lower.tail=FALSE)</code><br>

## Example 3 $P(\chi^2_{15} > 20)$

 - the 15 is the degrees of freedom
 - the 20 is the given $\chi^2$ value
 - since $\chi^2$ is greater than the given $\chi^2$ value, this is a right tail probability

```r
pchisq(20, 15, lower.tail=FALSE)
```

```r
#create density curve
curve(dchisq(x, df = 15), from = 0, to = 40,
main = 'Chi-Square Distribution (df = 15)',
ylab = 'Density',
lwd = 2)

#create vector of x values
x_vector <- seq(20, 40)

#create vector of chi-square density values
p_vector <- dchisq(x_vector, df = 15)

#fill in portion of the density plot from 0 to 40
polygon(c(x_vector, rev(x_vector)), c(p_vector, rep(0, length(p_vector))),
        col = adjustcolor('red', alpha=0.3), border = NA)
legend(x=25, y=0.07, legend=c("area = 0.17193"), lty=1:2, col=c('red'), bg="white")
```

## Role of ùúí2 distribution

 - Involved in estimatees of variability in a population
 - Distribution related to sums of squares
 - Example, if we have a random independent samples drawn from $N(\mu,\sigma)$ with <em>n</em> random variables $X_1,\ X_2,\ ..\ X_n$, then:
 - $ Z_j = \frac{X_j - \mu}{\sigma}\ \sim\ N(0,1)$ for j = 1, 2, ... n
 - Since all the samples are independent, then the $ Z_j^2\ \ X_j^2$ for for j = 1, 2, ... n
 - then we can conclude that each squared standardized value of $X_j$ has a $\chi^2$ distribution with a single dree of freedom and the sum of all the $X_j^2$ random variables has the $X_n^2$ distribution
 - $\Sigma_{j=1}^n Z_{j=1}^2$ = $\frac{1}{\sigma^2}\Sigma (X_j - \mu)^2$ ~ $\chi^2_n$
 - Expression can be rearranged so that we have the sum of squares: $$\chi^2_{n-1}\ +\ \chi^2_1$$
 - <mark>The degrees of freedom is n - 1</mark>

<div id="objective3" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Student's t Distribution</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics</u></strong>
        <li>Distribution that resembles the normal distribution but is a bit shorter and fatter</li><br>
        <li>This distribution is used when the sample size is small and the population standard deviation is unknown.</li><br>
        <li>The students t distribution is denoted as:
        $$T\ =\ \frac{\bar{X} - \mu}{\frac{S}{\sqrt{n}}}$$ where S is the sample standard deviation.
        </li><br>
        <li>The shape of the Student's t Distribution has a single peak</li><br>        
        <li>The peak conincides with the mean and median, and are located at 0</li><br>
        <li>The distribution is symmetric around the peak</li><br>
        <li>The students t distribution has heavier tails than the normal distribution</li><br>
        <li>Plot below illustrates that as the df increases, the t distribution begins to resemble a normal distribution.</li><br>
    </ul>
</div>

```r
#create density plots
curve(dt(x, df=6), from=-4, to=4, col='blue') 
curve(dt(x, df=10), from=-4, to=4, col='red', add=TRUE)
curve(dt(x, df=30), from=-4, to=4, col='green', add=TRUE)
curve(dt(x, df=1000), from=-4, to=4, col='green', add=TRUE)


#add legend
legend(-4, .3, legend=c("df=6", "df=10", "df=30", "df=1000"),
       col=c("blue", "red", "green", "black"), lty=1, cex=1.2)
```

## Quantiles from Student‚Äôs t Distribution

 - $t_{k,p}$ where k is the degrees of freedom and p is the prescribed upper tail probability
 - <code>qt(p, k, lower.tail=FALSE)</code>
 - In R, this code will denote the quantile that cuts off the probability to the right of the <em>p</em>

## Example 4 $t_{15,0.1}$

 - the 15 is the degrees of freedom
 - the 0.1 is the 90th percentile

```r
qt(0.1, 15, lower.tail=FALSE)
```

## Tail Probabilities from the Student's t distribution

 - For tail probabilities to the right: $P(t_k > x)$<br>
 - For tail probabilities to the left: $P(t_k < x)$<br> 
 - <code>pt(x, k, lower.tail=FALSE)</code><br>

## Example 5 $P(t_{10} > 1.75)$

 - the 10 is the degrees of freedom
 - the 1.75 is the t value

```r
pt(1.75, 10, lower.tail=FALSE)
```

<div id="objective4" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Snedecor‚Äôs F Distribution</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics on the F distribution</u></strong>
        <li>The F distribution is the ratio of two independent random variables with $\chi^2$ distributions drawn from a normal population</li><br>
        <li>The ratio is F = $\frac{\frac{W}{df_w}}{\frac{V}{df_v}}$</li><br>
        <li>Unlike the$\chi^2$  and t distribution, the F distribution has two degrees of freedom, expressed as $F_{df_w, df_v}$</li><br>
    </ul>
</div>

```r
##### Using R library functions for the F distribution

#### probability density function
n1 = 10
n2 = 12
r1 = n1-1
r2 = n2-2
F = 2.5

F_density = df(F, r1, r2)
F_density = round(F_density, digits=2)
print(paste("probability density for F = ", F, " and degree of freedom = ", r1," and ", r2, " is ", F_density))

### Generating the curve of F distribution probability density

x = seq(0,5,0.1)

r1 = 10
r2 = 12

string = "P(F,r1=10, r2=12)"

curve(df(x, r1, r2), xlim=c(0,4), xlab="F", ylab=string, lwd=1.5, cex.lab=1.2, col="blue", main="F distribution", font.lab=2)
```

## Quantiles from F Distribution

 - $F_{df_w, df_v}$ where df_w is the degrees of freedom for the numerator and df_v is the degree of freedom of the denominator
 - <code>qf(p, df1, df2, lower.tail=FALSE)</code>
 - In R, this code will denote the quantile that cuts off the probability to the right of the <em>p</em>

## Example 6 $F_{5,25, 0.1}$

 - the 5 is the degrees of freedom for the numerator
 - the 25 is the degrees of freedom for the denominator
 - the 0.1 is the 90th percentile

```r
qf(0.1, 5, 25, lower.tail=FALSE)
```

## Tail Probabilities from the F distribution

 - For tail probabilities to the right: $P(F_{df_w, df_v} > x)$<br>
 - For tail probabilities to the left: $P(F_{df_w, df_v} < x)$<br> 
 - <code>pf(f, dfw, dfv, lower.tail=FALSE)</code><br>

## Example 7 $P(F_{5,25} > 1.66)$

 - the 5 is the degrees of freedom for the numerator
 - the 25 is the degrees of freedom for the denominator
 - the 1.66 is the F value

```r
pf(1.66, 5, 25, lower.tail=FALSE)
```

## Example 8 $P(F_{6,13, 0.01})$

 - the 6 is the degrees of freedom for the numerator
 - the 13 is the degrees of freedom for the denominator
 - the 0.01 is the p value at the 99th percentile

```r
qf(0.01, 6, 13, lower.tail=FALSE)
```
