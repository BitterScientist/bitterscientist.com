<div id="objective1" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Primer on Statistical Distributions</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>What are statistical distributions?</u></strong>
        <li>A statistical distribution is the distribution that describes the relative number of times each possible outcome will occur in a number of trials.</li><br>    
        <li>Statistical distributions provide a parameterized mathematical function that can be used to calculate the probability of any instance or observation from a sample data.</li><br>
        <li>Statistical distributions are described by probability distributions functions, and two commonly probability functions is the <strong>Probability density function (PDF)</strong> and the <strong>Cumulative distribution function (CDF)</strong>.</li><br>
        <li>The <strong>Probability density function (PDF)</strong> gives the probability of a given value will occur.</li><br>
        <li>The <strong>Cumulative distribution function (CDF)</strong> describes the cumulative probability that a given value or any value smaller than it will occur.</li><br>
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>How are distributions mathematically defined?</u></strong>
        <li>Formally, a distribution is defined as a normalized measure, with the distribution of a random variable <strong>x</strong> is the measure \(P_{x} \) on \(\mathbb{S}' \) defined by setting:<br><br>
        \[P_{x} (A')= P\{s\in S: x(s) \in A' \} \newline \text{where (S,}\ \mathbb{S} \text{, P) is a probability space} \newline
        \text{(S,}\ \mathbb{S}\ \text{) is a measurable space} \newline
        \text{and}\ P\ \text{a measure on}\ \mathbb{S}\ \text{with } P(S)=1 \]
        </li><br>
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>Why are distributions important?</u></strong>
        <li>Probability distribution functions are a fundamental concept in statistics and machine learning.</li><br>        
        <li>They are used to calculate confidence intervals for parameters and the critical regions for hypothesis testing.</li><br>
        <li>Distributions are useful to determine a reasonable distribution model for univariate data <em>(data which consists of observations on only a single characteristic or attribute)</em>.</li><br>        
        <li>Statistical intervals and hypothesis test are based on specific distributions, and knowing which distribution best models the data can help verify that the selected test is justified.</li><br> 
        <li><mark>Probability distributions help us build model from our data, enabling us to obtain estimates of the probability that a certain event may occur, or estimate the variability of occurrence.</mark></li><br>       
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>Features of data that help characterize its distribution.</u></strong>
        <li>Before an analysis is performed on a given data set, it is important to characterize it by asking the following questions:
            <ol>
                <li>The data take on discrete or continuous values. Discrete data takes are numeric whole, concrete numbers with specific and fixed values. Continuous data includes complex numbers of varying data values.</li>
                <li>When the data is plotted, what features of the line describe the symmetry. If it is asymmetric, which direction is does it lie.</li>
                <li>Are there upper and lower limits on the data. Example, data does not go below zero, or does not exceed a specific value.</li>
                <li>What is the likelihood of observing extreme values in the distribution.</li>
            </ol>
        </li><br>
        <li>Probability distributions fall into two major categories, those that are  discrete or continuous.</li><br>        
        <li>Continuous probability distribution functions represent many commonly used models in statistics and machine learning, such as the Gaussian, chi-square and t distribution.</li><br>
        <li>The following sections will go into the mathematical intuition for these two categories, and the subsequent sections will survey major distributions in these two categories.</li><br>
    </ul>
     <hr style="border-top: 2px solid black;">
    <ul><strong><u>Discrete Distributions</u></strong>
        <li>Discrete distributions have values that take on numeric whole, concrete numbers with specific fixed values.</li><br>        
        <li>The mathematical definition of a discrete probability function p(x) has the following properties:
            <ol>
                <li>The probability of x can take on specific values:<br>
                \[P[X = x]\ =\ p(x)\ =\ p_{x} \]
                </li><br>
                <li>p(x) is non-negative for all real x values.</li><br>
                <li>The sum of p(x) over all possible values of x is 1.<br>
                \[\Sigma_{j}p_{j} = 1 \]<br>
                    where <strong>j</strong> is all possible values that x can have<br>
                    and \(p_{j} \) is the probability at \(x_{j} \)<br>
                </li><br>
            </ol>
        </li><br>
        <li>The result of properties 2 and 3 is that all probabilities of <strong>x</strong> are between <code>[0, 1]</code>.</li><br>        
        <li>A discrete probability function can take a discrete number of values.</li><br> 
        <li>In practice, discrete probability functions only take on integer values.</li><br>        
        <li>Each discrete value has a certain probability of occurring that is between 0 and 1.</li><br>
        <li>A discrete function that allows negative values or values greater than one are not a probability function.</li><br>        
        <li>Finally, the sum of all probabilities is one, means that at least one of the values has to occur.</li><br>   
        <li>Discrete probability functions are referred to as <strong>probability mass functions</strong></li><br>
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>Continuous Distributions</u></strong>
        <li>Continuous data includes complex numbers of varying data values.</li><br>     
        <li>The mathematical definition of a Continuous probability function f(x) has the following properties:
            <ol>
                <li>The probability that x is between point a and b is:<br><br>
                \[p[a \leq x \leq b] = \int_{a}^{b} f(x)dx  \]
                </li><br>
                <li>Values are non-negative for all real x.</li><br>
                <li>The integral of the probability function is one:<br><br>
                \[\int_{\infty}^{-\infty} f(x)dx = 1  \]
                </li><br>
            </ol>
        </li><br>
        <li>Since a continuous probability functions is defined over an infinite number of points over a continuous interval, the probability of a single point is always zero.</li><br>        
        <li>Probabilities are measured over intervals and not single points, and the area under the curve between these two distinct points define the probability for that interval.</li><br> 
        <li>The height of the probability function can be greater than 1.</li><br>        
        <li>The integral for continuous probability functions must be equal and is equivalent to the property for discrete distributions.</li><br>
        <li>Continuous probability functions are referred to as <strong>probability density functions</strong></li><br>         
    </ul>
</div>

<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics</u></strong>
        <li>Chi-Square distribution is a continuous probability distribution with <em>k</em> degrees of freedom..</li><br>
        <li>Chi-Square distribution is one of the most commonly used discrete distribution.</li><br>
        <li>Chi-Square distribution is the distribution of a sum of the squares of the k independent standard normal random variables.</li><br>
        <li>Chi-Square distribution is a special case of the gamma distribution, and is one of the most widely used probability distributions in inferential statistics.</li><br>
        <li>Two common statistical tests that relies on the Chi-Squared distribution are the <strong>Chi-square goodness of fit test</strong> and the <strong>Chi-square test of independence</strong>.</li><br>
        <li>The distribution is defined by the degrees of freedom, and will change the distribution, as the number of degrees of freedom increases, the distribution takes on a normal curve.</li><br>
        <li>Since the Chi-square distribution can be used wherever a normal distribution can be used for a hypothesis test.</li><br>
        <li>The Chi-square distribution is also used in these hypothesis tests
            <ul><strong>Examples</strong>
                <li>Chi-square test of independence in contingency tables</li>
                <li>Chi-square test of goodness of fit of observed data to hypothetical distributions</li>
                <li>Likelihood-ratio test for nested models</li>
                <li>Log-rank test in survival analysis</li>
                <li>Cochran–Mantel–Haenszel test for stratified contingency tables</li>
            </ul>
        </li><br>        
        <li>The Chi-square test is also a component of the definition of the t-distribution and the F-distribution used in t-tests, analysis of variance, and regression analysis.</li><br>
        <li>The Chi Square distribution is very important because many test statistics are approximately distributed as Chi Square.</li><br>   
    </ul>
</div>

<div id="objective3" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Mathematical Intuition for Chi-Square Distribution</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
     <ul><strong><u>Chi-square basics</u></strong>
         <li>If \(Z_{1}...Z_{k}\) are independent, standard normal variables, then the sum of their squares would be:
             \[Q = \Sigma_{i=1}^{k}Z_{i}^{2}\]
             </li><br>
         <li>Q is distributed according to the chi-square distribution with <em>k</em> degrees of freedom and is denoted as:<br>
         \(Q ~ \chi^{2}(k)\) or \(Q ~ \chi^{2}_{k}\)
         </li><br>
         <li>Note that the chi-square distribution only has one parameter: a positive integer <em>k</em> that specifies the number of degrees of freedom.</li><br>
         <li>The degrees of freedom are the number of random variables that are summed up \(Z_{i}\).</li><br>
    </ul>
        <hr style="border-top: 2px solid black;">
     <ul><strong><u>Mathematical Intuition for Chi-Square Distribution</u></strong>
         <li>\[X = \frac{m - N_{p}}{\sqrt{Npq}} \]
             where <em>m</em> is the number of observed successes in N trials and the probability of success is <em>p</em>. Therefore <em>q = 1 - p</em>.
         </li><br>
         <li>Squaring both sides of the equation gives us:
             \[X^{2} = \frac{(m - N_{p})^{2}}{Npq}  \]
         </li><br>
         <li>if we then use the following formulas
             \[N = N_{p} + N(1 - p)\],
             \[N = m + (N - m)\],
             and q = 1 - p
         </li><br>
         <li>\[X^{2} = \frac{(m - N_{p})^{2}}{Np} + \frac{(N - m - N_{q})^{2}}{Nq} \]</li><br>
         <li>This is then generalized to the following form
             \[X^{2} = \Sigma_{i=1}^{n}\frac{(0_{i} - E_{i})^{2}}{E_{i}}\]</li><br>
         <li>\(X^{2} \) is Pearson's cumulative test statistic</li>
         <li>\(O_{i} \) is the number of observation of type <em>i</em></li>
         <li>\(E_{i} = N_{pi} \) is the expected frequency of type <em>i</em> asserted by the null hypothesis that the fraction of type <em>i</em> in the population is \(p_{i} \)</li>
        <li>n is the number of cells in the table.</li>
    </ul>
            <hr style="border-top: 2px solid black;">
     <ul><strong><u>Probability Density Function</u></strong>
         <li>\[f(x) = \frac{e^{\frac{-x}{2}}x^{\frac{v}{2}-1}}{2^{\frac{v}{2}}\Gamma(\frac{v}{2})}\] for \(x \geq 0 \)</li>
         <li><em>v</em> is the shape parameter \(\Gamma\) is the gamma function</li><br>
         <li>The gamma function is then described as \[\Gamma(a) = \int_{0}^{\infty}t^{a-1}e^{-t}dt \]</li><br>
         <li>The chi-square distribution is treated as a standard distribution with no location or scale parameter, but it can be transformed with a location parameter \(\mu\) and a scale parameter \(\Sigma\).</li><br>
    </ul>
                <hr style="border-top: 2px solid black;">
         <ul><strong><u>Cumulative Distribution Function</u></strong>
         <li>\[f(x) = \frac{\gamma(\frac{v}{2}, \frac{x}{2})}{\Gamma(\frac{v}{2})} \] for \(x \geq 0 \)</li>
         <li>where \(\Gamma\) is the gamma function described in the previous example.</li><br>
         <li>and \(\gamma\) is the incomplete gamma function that is described as
             \[\gamma_{x}(a) = \int_{0}^{\infty}t^{a-1}e^{-t}dt  \]
             </li><br>
    </ul>
</div>

<div id="objective5" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Visualizing the Chi-Square Probability Density Function</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Why we plot the Probability Density Function (PDF)</u></strong>
        <li>The Probability density function gives the probability of a given value will occur.</li><br>     
        <li>Visualizing PDF gives a graphically portrayed off a given probability and the area under the curve will indicate the interval in which the variable will fall. </li><br>
        <li>The probability density plot is used to understand the data distribution for a given variable and we want to know the likelihood (or probability) of obtaining a range of values that the variable can assume.</li><br> 
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>How to plot PDF using python scipy (v1.6.3) and matplotlib.</u></strong>
        <li>The <a href="https://docs.scipy.org/doc/scipy/reference/reference/generated/scipy.stats.chi2.html">scipy \(\chi^{2} \) class</a> provides a chi-squared continous random variable.</li><br>        
        <li><code>scipy.stats.chi2</code> in this package is defined as:<br>
        \[f(x,k) = \frac{1}{2^{\frac{k}{2}}\Gamma(\frac{k}{2})}x^{\frac{}k{2-1}}exp(\frac{-x}{2} \] <br>
        \[for\ x\ > 0\ and\ k > 0\]<br>
            \[and\ k = degrees of freedom\]<br>
        </li><br>
        <li>This class has several methods that can be used on a A chi-squared continuous.</li><br>
        <li>To plot the probability mass function, we will use the percent point function <code>chi2.ppf(q, df, loc=0, scale=1)</code> and the probability density function <code>chi2.pdf(x, df, loc=0, scale=1)</code></li><br>
        <li><strong><code>chi2.ppf(q, df, loc=0, scale=1)</code>:</strong> takes the quartile and the degrees of freedom.  We obtained the ppf for the 1% and the 99% percentiles and then used the numpy method <code><a href="https://numpy.org/doc/stable/reference/generated/numpy.linspace.html">numpy.linspace</a></code> to return an array of 100 data points between the 1% and 99% and saves all this data into the <code>x</code> variable.</li><br>
        <li><strong><code>chi2.pdf(x, df)</code>:</strong> uses the array of 100 values between 0.1 and 0.99 of a and returns the chi-square probability distribution.</li><br>
        <li>The df (or k) will affect the overall shape of the curve, and as the df (k) increases, the curve begins to resemble a normal distribution.</li><br>    
        <li>In example below, we will generate four subplots, each subplot will have a different value for <em>p</em><br> 
        <li>Note that as the probability for success increases, the binomial probability mass function will shift to the right, and vice versa.<br>
    </ul>
</div>

```python
from scipy.stats import chi2
import matplotlib.pyplot as plt
import numpy as np
```

```python
degrees_of_freedom = [1, 3, 6, 10] # the degrees of freedom
fig = plt.figure(figsize=(12,8))
for df, num in zip(degrees_of_freedom, range(1,5)):
    
    ax = fig.add_subplot(2,2, num)
    x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)
    ax.plot(x, chi2.pdf(x, df), 'r-', lw=5, alpha=0.6, label='chi2 pdf')
    rv = chi2(df)
    vals = chi2.ppf([0.001, 0.5, 0.999], df)
    r = chi2.rvs(df, size=1000)
    ax.hist(r, density=True, histtype='stepfilled', alpha=0.1)
    ax.legend(loc='best', frameon=False)
    ax.set_ylabel('Probability Mass', fontsize=14)
    ax.set_xlabel('x', fontsize=14)
    ax.set_title(f"Chi-Square PDF (k={df}, n={len(x)})")

plt.tight_layout()
plt.show()
```

<div id="objective6" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Visualizing the Chi-Square Cumulative Density Function</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Why we plot the Chi-Square Cumulative Density Function</u></strong>
        <li>The Cumulative distribution function (CDF) describes the cumulative probability that a given value or any value smaller than it will occur.</li><br>        
        <li>The chi-square cumulative distribution function lets one obtain the probability of observing less than or equal to <em>x</em> successes in <em>n</em> trials, with the probability <em>p</em> of success on a single trial.</li><br>       
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>The mathematics of CDF</u></strong>
         <li>\[f(x) = \frac{\gamma(\frac{v}{2}, \frac{x}{2})}{\Gamma(\frac{v}{2})} \] for \(x \geq 0 \)</li>
         <li>where \(\Gamma\) is the gamma function described in the previous example.</li><br>
         <li>and \(\gamma\) is the incomplete gamma function that is described as
             \[\gamma_{x}(a) = \int_{0}^{\infty}t^{a-1}e^{-t}dt  \]
             </li><br>
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>How to plot CDF using python scipy (v1.6.3) and matplotlib.</u></strong>
        <li>The <a href="https://docs.scipy.org/doc/scipy/reference/reference/generated/scipy.stats.chi2.html">scipy chi2 class</a> provides a chi-squared continuous random variable.</li><br>        
        <li><code>scipy.stats.binom(*args, **kwds)</code></li><br>
        <li>This class has several methods that can be used on a binomial discrete random variable.</li><br>
        <li>To plot the cumulative distribution function, we will use <code>cdf(x, df)</code></li><br>
        <li><strong><code>x</code>:</strong> array of values</li><br>
        <li><strong><code>df</code>:</strong> degrees of freedom</li><br>
        <li>In example below, we will generate four subplots, each subplot will have a different value for the degrees of freedom<em>df</em><br> 
    </ul>
</div>

```python
degrees_of_freedom = [1, 3, 6, 10] # the degrees of freedom
fig = plt.figure(figsize=(12,8))
for df, num in zip(degrees_of_freedom, range(1,5)):
    
    ax = fig.add_subplot(2,2, num)
    x = np.linspace(chi2.ppf(0.01, df), chi2.ppf(0.99, df), 100)
    ax.plot(x, chi2.cdf(x, df), 'r-', lw=5, alpha=0.6, label='chi2 cdf')
    ax.legend(loc='best', frameon=False)
    ax.set_ylabel('Probability', fontsize=14)
    ax.set_xlabel('x', fontsize=14)
    ax.set_xlim(0,10)
    ax.set_title(f"Chi-Square PDF (k={df}, n={len(x)})")

plt.tight_layout()
plt.show()
```
