<div id="objective1" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">How to Resolve Missing Values.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Strategy 1: Removing Missing Values</u></strong>
        <li>Missing values are data points in the sample that are absent.</li><br>
        <li>Missing values may occur due to several reasons, including errors during the data collection process, non-response (such as in surveys), or values were not determine at the time of the model building.</li><br>
        <li>It is important to understand why the data is missing to avoid future studies from having the issue issue.</li><br>
        <li>A predictor or feature can be removed from the model if it contains a high percent of missing values.  Although be aware that removing a feature from a small dataset can have big problems downstream.</li><br>
    </ul>
<hr style="border-top: 2px solid black;">
    <ul><strong><u>Strategy 2: Keep or Replace Missing Values</u></strong>
        <li>If your downstream application is a predictive model such as a tree-based technique, you can keep the missing values.</li><br>
        <li>Methods for replacing values that do not rely on predictive modeling include: using the average value for that feature or using the last known value).</li><br>
        <li>Use an imputation approach to replace missing values with predicted values using an additional layer of modeling.</li><br>        
        <li>K-nearest neighbors (KNN) is one predictive model used to replace the missing value.</li><br>
        <li>In KNN, the new sample is imputed by finding the closest set of values, calculating the average of those values and then replacing the missing values with this calculated field.</li><br>
        <li>The number of neighbors is the tuning parameters for how many "closest" points to use for the averaging.</li><br>        
    </ul>
</div>

<div id="objective2" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Removing Predictors.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Remove Predictors</u></strong>
        <li>The advantage of removing a predictor is that it reduces the computational time and complexity, creates a model easier to interpret, and a reduced number of predictors may improve the model performance and stability.</li><br>
        <li>Categorical features can be removed if the feature has zero variance or near-zero variance.</li><br>
        <li>Zero variance predictor (feature) contains only a single unique value. An example would be a column of data that for students that has the label undergraduate, and all the students sampled are undergraduate. Theoretically, you could have had a graduate student in the sample, but this is not the case in this data. So there will be zero variance and this feature would not add anything to the model.</li><br>
        <li>Zero variance features would create problems for linear regression models.</li><br>
        <li>Near-Zero variance predictors (Features) is a vector (column) of data that have only a handful of unique values that occur with very low frequencies.</li><br>
        <li>An example of near-zero variance would be a vector of data for undergraduate students major: [235 english, 4 math, 1 computer science, 1 art].  In this example, 98% of the data (students) are english majors, and the other majors have a small influence on the model.  If any resampling was used, it is highly possible that only unique value will be returned for major (english) and would resemble a Zero-Variance predictor.</li><br>
        <li>You have a near-zero variance predictor if:
            <ul>
                <li>The fraction of unique values is low (less than 10%), such as 4 unique 'majors' in 239 students samples is 4/239 = 0.16%.</li><br>
                <li>If the most common value divided by the second most common value is over 20. Example, 235 english majors divided by 4 math majors = 58.75.</li><br> 
            </ul>  
        </li><br>
        <li>The <code>caret</code> package contains a function called <code>nearZeroVar</code> that will return the categorical columns that have new-zero variance so that you can omit them from your dataset before you perform your predictive model.</li><br>
        <li>Note that continuous variables cannot be near-zero variance predictors since continuous values have more unique values that categorical data.</li><br>        
    </ul>   
</div>

<div id="objective3" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Dealing with Between-Predictor Correlations.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Multicollinearity</u></strong>
        <li>Collinearity occurs when a pair of predictor variables have a substantial correlation with each other.</li><br>
        <li>The problem with collinearity is that it becomes difficult to determine with features are important in predictive modeling, so it is common practice to remove highly correlated features.</li><br>
        <li>A correlation plot is commonly used to diagnose multicollinearity.</li><br>
        <li>The correlation plot is a 2-D symmetric matrix, with the top and bottom diagonals showing identical information.</li><br>
        <li>The correlation plot groups clusters of predictor variables that have collinearity.</li><br>
        <li>Principle component analysis can be used to identify the extent of the problem, such as if the first principal component accounts for a large percentage of the variance, it implies there is at least one group of predictors that represents the same information.</li><br>
    </ul>
<hr style="border-top: 2px solid black;">
    <ul><strong><u>How to address multicollinearity</u></strong>
        <li>Remove the minimal number of predictors that ensures all the pairwise correlations are below a specified threshold.</li><br>
        <li>First calculate the correlation matrix</li><br>
        <li>Determine the two predictors that are associated with the largest pairwise correlations.</li><br>
        <li>Determine the average correlations between these two predictors and remove the predictor with the lower average correlation.</li><br>
        <li>Repeat this process until no absolute correlations above the given threshold exist.</li><br>        
        <li>The <code>findCorrelation</code> function in <code>caret</code> package performs these steps for you..</li><br>                
    </ul>  
</div>

# Install packages and Load AppliedPredictiveModeling

 - AppliedPredictiveModeling is a package that contains various datasets commonly used for training purposes.

```r
# install packages
install.packages(c("caret", "corrplot", "e1071", "lattice", "AppliedPredictiveModeling"))
# Load the library
library(AppliedPredictiveModeling)

# Load and inspect the segmented data and print out the dimensions, 2019 rows, and 199 columns
data(segmentationOriginal) 

# segment data for training set
segData <- subset(segmentationOriginal, Case == "Train") 

# save class, cell and case into separate vectors
cellID <- segData$Cell;cellID[0:5]
class <- segData$Class;class[0:5]
case <- segData$Case;case[0:5]

# Now remove the columns
segData <- segData[, -(1:3)]
statusColNum <- grep("Status", names(segData))   

# count number of columns that need to be removed
length(statusColNum) 
segDataCont <- segData[, -statusColNum]

# transform data using center, scaling on predictor variables
YY = scale(segDataCont)

# Evaluate sample skewness statistic for each predictor using functions in the e1071 package
library(e1071)  

# Compute skewness for all numeric predictors using apply function
skewValues <- apply(segDataCont, 2, skewness)

# Import MASS package to determine transformation to apply to data
library(caret)
AvgIntenCh3Trans <- BoxCoxTrans(segDataCont$AvgIntenCh3)  

# Transformed
predict(AvgIntenCh3Trans, head(segDataCont$AvgIntenCh3))
x1=predict(AvgIntenCh3Trans, segDataCont$AvgIntenCh3)
skewness(x1)

# Transform entire numeric predictor dataset using the caret function preProcess
xx1 <- preProcess(segDataCont, method = c("BoxCox"))

# Apply the transformations:
transformed <- predict(xx1, segDataCont)
```

## Filter for near-zero variance predictors using the nearZeroVar function and this process should not be performed on continuous predictors. 

```r
## for all categorical predictors, need to recall the data
segDataCat<-segData[, statusColNum] 

# prints out the dimensions of the categorical features/predictors
dim(segDataCat)

# apply nearZeroVariance to categorical data and will return the predictor indexes
nearZeroVar(segDataCat)

# prints out the number of zeros in a specific index as an example
# the columns 33, 37, and 38 should be removed
length(which(segDataCat[,33]==0))
```

# Use the cor function to filter for between-predictor correlation and visualize this using the corrplot library

```r
correlations <- cor(segDataCont)
dim(correlations)
correlations[1:4, 1:4]

library(corrplot)
corrplot(correlations, order = "hclust")
```

## Use the findCorrelation function to filter based on correlations. The function will apply the algorithm discussed in the previous section.  We provide a given threshold of pairwise correlations, and the function will return the column index for the predictors that should be deleted

```r
# find the high correlation features at the 0.85 cutoff
highCorr <- findCorrelation(correlations, cutoff = .85)

# how many features are highly correlated
length(highCorr)

# print out these features
highCorr

# create new dataset that filters out these high correlation features
filteredSegData <- segDataCont[, -highCorr]

# print out the new number of features that lack Multicollinearity
length(filteredSegData)
```
