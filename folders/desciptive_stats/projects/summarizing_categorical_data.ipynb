<div id="objective1" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Introduction to Categorical Data.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics:</u></strong>
        <li><strong>Descriptive statistics</strong> condenses an entire dataset to a set of numbers that describe properties of the entire dataset.</li><br>
        <li>There are two types of descriptive statistics, 
        <ol>
            <li>Numeric summaries</li>
            <li>Graphical summaries</li>
        </ol>
        </li><br>
        <li><strong>Numeric summaries</strong> are either a function of a sample that results in a number, or a number that describes a characteristic of the dataset (mean, median, mode, range, correlation, percentile).</li><br>
        <li><strong>Graphical summaries</strong> are visualizations obtained from the data that are used to display common numeric summaries.</li><br>
        <li><strong>Marginal summaries</strong> is a summary of single variable data.</li><br>
        <li><strong>Joint summaries</strong> is a summary of two or more variables.</li><br>          
    </ul>
</div>

<hr style="border-top: 5px solid black;">

<div id="objective2" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Marginal Summaries.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Numeric Marginal Summaries:</u></strong>
        <li>Is a numerical summary of a single categorical variable that is presented as either a frequnecy table or relative frequency table.</li><br>
        <li>The sample frequency is the count of the number of sample observations taking a given value.</li><br>
        <li>The sample relative frequency is the frequency divded by the sample size.</li><br>
        <li>The sample relative proportion $\hat{p}$ is also the frequency divded by the sample size.</li><br>
        <li>The relative frequenccy can be converted to a percentage.</li><br>       
    </ul>
    <hr>
    <ul><strong><u>Graphical Marginal Summaries:</u></strong>
        <li>A bar chart consists of bars whose heights are the observed frequency (or relative or percentage).</li><br>        
        <li>A pie chart is a circle with slices whose area are in the same proportion as the observed frequencies.</li><br>        
    </ul>    
</div>

# Sorting Levels by Frequency in R

```r
eye.tab <- sort(table(eye), decreasing=TRUE) ; eye.tab
```

# Change Level Labels in R

```r
# create a new vector with changes labels
eye <- factor(eye)
levels(eye) <- c("Blue", "Brown", "Green", "Hazel")
levels(eye)

# confirm changes by printing frequency table in descending order
eye.tab <- sort(table(eye), decreasing=TRUE)
eye.tab
```

# Pie Chart in R using <code>pie(feature.tab)</code>

```r
pie(eye.tab, main="Eye Colors")
```

# Bar Plot in R using <code>barplot(feature.tab, main="title")</code>

```r
barplot(eye.tab, main="Student Eye Colors", xlab="Eye Colors", ylab="Frequency")
```

<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics:</u></strong>
        <li>Summarize relationship between two categorical variables with numbers by using a contingency table.</li><br>
        <li>A contingency table has rows that correspond to a level on one variable and columns that represent another variable and its levels.</li><br>       
    </ul>
</div>

# Contingency Table in R using <code>table(feature1, feature2)</code>

```r
table(hand,eye)
```

# Contingency Table with Marginal Summaries in R

```r
addmargins(table(hand,eye))
```

# Contingency Table with Relative Frequencies

```r
prop.table(table(hand,eye))
```

# Contingency Table with Relative Frequencies and Marginal Summaries

```r
addmargins(prop.table(table(hand,eye)))
```

# Contingency Table with Relative Frequencies and Marginal Summaries As Percentages

```r
addmargins(100*prop.table(table(hand,eye)))
```

# Stacked bar chart

```r
barplot(table(hand,eye), legend.text=TRUE, main="Eye Color and Writing Hand Preference", xlab="Eye Color", ylab="Frequency")
```

# Grouped Bar Chart

```r
hand.parking.tab <- table(hand,parking)

barplot(hand.parking.tab, legend.text=TRUE, beside=TRUE, xlab="Class Status", ylab="Frequency", main="Parking Satisfaction by Left and Righ Handedness", args.legend=list(title="Parking Satisfaction"))
```
