# Step 1: Add subquery in the <code><font color="red">SELECT</font></code> Clause
 - Average Goals: Extract the average home and away goals
 - Average Goal for each stage: Calculate the average home and away goals by stage
 - Filter both queries for the 2015/2016 season.
 - A filter in the select clause will return a column with a scalar value!
 - <b>Most of the work will be done in the <code>SELECT</code> clause.

```python
sql_view2 = """SELECT 
    m.stage,
    ROUND(AVG(m.home_team_goal + m.away_team_goal),2) AS avg_goals,
    ROUND((SELECT AVG(home_team_goal + away_team_goal) 
           FROM Match 
           WHERE season = '2015/2016'),2) AS overall
FROM Match AS m
WHERE season = '2015/2016'
GROUP BY m.stage;"""
```

```python
# Run query
view_data2 = pd.read_sql(sql_view2, conn)
# Displaying data
view_data2.head()
```

### <font color="blue"><code>SELECT</code> Clause Explained:</font><br>

<strong>The <code>SELECT</code> clause.</strong><br>
 - Will return 3 columns of records
 - Contains a subquery<br>
 - The subquery will return a scalar value (meaning it will be the same value in the column)

<code>SELECT 
    m.stage,
    ROUND(AVG(m.home_team_goal + m.away_team_goal),2) AS avg_goals,
    ROUND((SELECT AVG(home_team_goal + away_team_goal) 
           FROM Match 
           WHERE season = '2015/2016'),2) AS overall</code>
<br><br>
The <code>SELECT</code><br> Clause:
This will instruct the query to return the following columns<br>
  - Column 1: <code>m.stage</code><br>
     - The stage column from the Match table
  - Column 2: <code>ROUND(AVG(m.home_team_goal + m.away_team_goal),2) AS avg_goals</code>
     - This is a calculated field, and will return the rounded average of total goals scores from the match column
     - <code>AS avg_goals</code>
        - This will alias the records so that the column has the title "avg_goals"
        
  - Column 3: <code>(ROUND((SELECT AVG(home_team_goal + away_team_goal) FROM Match WHERE season = '2015/2016'),2) AS overall</code>
     - This is a subquery that will return a column with the overall average goals scores
     - The value will be the same for all rows in the column
     - The results from this subquery will be alias as <code><b>overall_avg</b></code> and will be the title of the column 

### <font color="blue"><code>FROM</code> Clause Explained:</font><br>

<strong>The <code>FROM</code> clause.</strong><br>
<code>FROM Match AS m</code>
<br>
<code>FROM</code><br>
Is the source of the data that will be used<br>
 - The dataset will be obtained from the Match table as is aliased as <code><b>m</b></code> for ease

### <font color="blue"><code>WHERE</code> Clause Explained:</font><br>

<strong>The <code>WHERE</code> clause.</strong><br>
<code>WHERE season = '2015/2016'</code>
<br>
<code>WHERE</code><br>
This will instruct the query to filter the rows of data if they meet a specific criteria<br>
 - The data will be filtered for records where the season is 2015/2016

## Taken Together

This code will use the <code>SELECT</code> clause to display the records provided by the <code>FROM</code> clause.
 - The <code>FROM</code> clause will prepare the data that will be eventualy displayed from the Match table.

    - This includes the three columns: the stage and the avg_goals the overall average goals using a the subquery
    - The column labels are given here with the aliases for two of the columns
 - The <code>WHERE</code> clause will instruct which records are to be displayed.
    - This will limit the records to those where the season is 2015/2016.

## Visualize average goals for the 38 stages and identify those stages that are greater than the overall goals for the 2015/2016 season.

```python
# This is the stages array
stages = view_data2.stage.values
# These are the average goal per stage
avg_goals = view_data2.avg_goals.values
# This is an array that contains the overall average number of goals scored for the entire season
overall =  view_data2.overall.values
```

# Step 2: Add subquery in <code><font color="red">FROM</font></code> Clause

This step will turn the main query into a subquery and filter out a list of stages where the average goals in a stage is greater than the overall average goals in a match.<br>
 - <code>FROM</code> clause subquery: Calculate the average goals for each stage in the FROM clause subquery.
 - <code>WHERE</code> clause subquery: Calculate the overall average goals scored
 - Filter the main query for stages where the average goals > than the overall average.
 - <code>SELECT</code> clause: Select the stage and avg_goals columns from the s subquery into the main query.

```python
sql_view3 = """
SELECT 
    subquery.stage,
    ROUND(subquery.avg_goals,2) AS avg_goals
FROM 
    (
    SELECT
        stage,
        AVG(home_team_goal + away_team_goal) AS avg_goals
    FROM Match
    WHERE season = '2015/2016'
    GROUP BY stage) AS subquery
WHERE 
    subquery.avg_goals > (SELECT AVG(home_team_goal + away_team_goal) 
                    FROM match WHERE season = '2015/2016');"""
```

```python
# Run query
view_data3 = pd.read_sql(sql_view3, conn)
# Displaying data
view_data3
```

# Step 3: Add a subquery in the <code><font color="red">SELECT</font></code> Clause

This step will create a query that will compare the average number of goals scored in each stage to the total.<br>
It will do the same as in step 2, <b>BUT</b> will add an additional column with the overall average goals scored in the 2015/2016 season.
 - <code>SELECT</code> clause subquery: returns a column "overall_avg" that yields the average goals scored in the 2015/2016 season.
 - <code>FROM</code> clause subquery: calculates the average goals scored in each stage during the 2015/2016 season.
 - Finally, filter the main query for stages where the average goals > than the overall average.

```python
sql_view4 = """
SELECT 
    subquery.stage,
    ROUND(subquery.avg_goals,2) AS avg_goal,
    (
    SELECT AVG(home_team_goal + away_team_goal) 
    FROM Match 
    WHERE season = '2015/2016'
    ) AS overall_avg
FROM 
    (
    SELECT
        stage,
        AVG(home_team_goal + away_team_goal) AS avg_goals
    FROM Match
    WHERE season = '2015/2016'
    GROUP BY stage) AS subquery
WHERE 
    subquery.avg_goals > (SELECT AVG(home_team_goal + away_team_goal) 
                    FROM match WHERE season = '2015/2016');
"""                 
```

```python
# Run query
view_data4 = pd.read_sql(sql_view4, conn)
# Displaying data
view_data4
```
