# Method 1: Modify Strings using Python string.lstrip()

<code>string.lstrip(s[, chars])</code>
<br>
 - Return a copy of the string with leading characters removed. If chars is omitted or None, whitespace characters are removed. If given and not None, chars must be a string; the characters in the string will be stripped from the beginning of the string this method is called on.
 - <font color="red">The string does not have to be identical.</font>
 - You can chain string.lstrip() methods <code>string.lstrip('string1').lstrip('string2')</code>
 - Documentation: https://docs.python.org/2/library/string.html

```python
method1 = []
for s in column_headers:
    method1.append(s.lstrip(del_string1).lstrip(del_string2))
```

```python
# Evaluate Edits
method1
```

### <font color="red">Notice that using string.lstrip() removes more of the header string than wanted. This is because string.strip() does not match exactly, just the beginning characters in a string.<font>

# Method 2: Modify Strings using Python string.replace()

<code>string.replace(s, old, new[, maxreplace])</code>
<br>
 - Return a copy of string s with all occurrences of substring old replaced by new. If the optional argument maxreplace is given, the first maxreplace occurrences are replaced.
 - You can chain commands <code>string.replace(s, old, new[, maxreplace]).replace(s, old, new[, maxreplace])</code>
 - Documentation: https://docs.python.org/2/library/string.html

```python
# Used a list comprehension to loop through the column_headers array and modify each string.
method2 = list(map(lambda x: x.replace(del_string1,'').replace(del_string2,''), column_headers))
```

```python
method2
```

<font color="blue">Notice that this method maintains all the header values needed.</font>

# Update Dataframe Columns

```python
# Rename header columns
df.columns = method2
```
