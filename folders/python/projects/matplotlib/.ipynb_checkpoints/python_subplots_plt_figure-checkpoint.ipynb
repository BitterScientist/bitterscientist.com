# Creating subplots using plt.figure()
 - <font color="blue">matplotlib.pyplot.figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True, FigureClass, clear=False, **kwargs)</font>
 - Documentation: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html

## Why use plt.figure()
<br>
1 <strong>Gives you handle to a figure. </strong> This will allow you to store this instance in a variable and use it later on. <br>
&emsp;&emsp;Example: <br>
&emsp;&emsp;fig = plt.figure()<br>
&emsp;&emsp;#... other code<br>
&emsp;&emsp;fig.autofmt_xdate()<br>
<br>
2 <strong>Set figure parameters. </strong> You can set some parameters for the figure by supplying them as arguments to plt.figure.<br>
&emsp;&emsp;Example: <br>
&emsp;&emsp;plt.figure(figsize=(10,10), dpi=200): <br>
<br>   
3 <strong>Create multiple figures. </strong> You can create several figures in the same script.
<br>
&emsp;&emsp;Example:
&emsp;&emsp;plt.figure()      # create a figure<br>
&emsp;&emsp;plt.plot([2,2,1]) <br>
&emsp;&emsp;plt.figure()      # create another figure <br>
&emsp;&emsp;plt.plot([2,2,2]) # successive commands are plotted to the new figure<br>
<br>

## Key plt.figure Options
 - <strong>num :</strong> integer or string, optional, default: None
 - <strong>figsize :</strong> (float, float), optional, default: None. Gurantees the figure will be a certain size when saved to disk.
 - <strong>dpi :</strong> integer, optional, default: None. Resolution of the figure.

```python
# This will create a new figure
# In IPython an empty window will appear, but in jn, nothing will be shown
fig = plt.figure()
```

## Add subplot(s) using <font color="red">add_subplots(nrows, ncols, index)</font>
 - nrows: Number of rows that will contain subplots in the figure
 - ncols: Number of columns that will contain subplots in the figure
 - index: Label of the returned index<br>
<font color="red">The number of subplots is nrows * ncols</font>

### Example: Create a 2x2 figure with 3 plots

```python
# This will create a 2x2 matrix of subplots  and the index = 1 will refer to the position of the subplot
fig = plt.figure()
# To create a subplot using this method, you need to create a unique axes instance variable and assign the 
# subplot to that axes
ax1 = fig.add_subplot(2, 2, 1) #This will create a plot on the top left grid
ax2 = fig.add_subplot(2, 2, 2) #This will create a plot on the top right grid
ax3 = fig.add_subplot(2, 2, 3) #This will create a plot on the bottom left grid
plt.show()
```

### What if I added a plt.plot() command when trying to add data to this figure?
 - It will add the plot to the last subplot in figure

```python
# This will create a 2x2 matrix of subplots  and the index = 1 will refer to the position of the subplot
fig = plt.figure()
# To create a subplot using this method, you need to create a unique axes instance variable and assign the 
# subplot to that axes
ax1 = fig.add_subplot(2, 2, 1) #This will create a plot on the top left grid
ax2 = fig.add_subplot(2, 2, 2) #This will create a plot on the top right grid
ax3 = fig.add_subplot(2, 2, 3) #This will create a plot on the bottom left grid
plt.plot(np.random.randn(50).cumsum(), 'k--') # Added to last subplot
plt.show()
```

### How to add data to a specific subplot in a figure
 - You need to specify the AxesSubplot object that you will add the data!!!!
 - In the example below, instead of using plt.plot(), we used <font color="red"><strong>ax#.plot()</strong></font>

```python
# This will create a 2x2 matrix of subplots  and the index = 1 will refer to the position of the subplot
fig = plt.figure()
# To create a subplot using this method, you need to create a unique axes instance variable and assign the 
# subplot to that axes
ax1 = fig.add_subplot(2, 2, 1) #This will create a plot on the top left grid
ax2 = fig.add_subplot(2, 2, 2) #This will create a plot on the top right grid
ax3 = fig.add_subplot(2, 2, 3) #This will create a plot on the bottom left grid
ax1.hist(np.random.rand(500), bins=30, color='red', alpha=0.4) # This will add data to the 1st AxesObject Instance!
ax2.plot(np.random.randn(50).cumsum(), 'k--') # This will add the data to the 2nd AxesObject Instance!
# Note no data was added to the 3rd AxesObject
plt.show()
```

### Plot HIV Prevalence Data
 - Step 1: Select four random countries
 - Step 2: Transpose dataframe so that you have countries in columns, and the years as the rows
 - Step 3: Relabel header (they will contain the index from the row from transposition)

```python
# Get four random countries
four_countries = df.sample(n=4)
```

```python
# Evalute country selection
four_countries.head()
```

```python
# Transform data for plotting purposes (Transpose)
four_countries = four_countries.T
```

```python
# Get the name of the countries
names = four_countries.iloc[0]
# Relabel the transposed column names
four_countries.columns = names
# Drop the row that contained the contry names and update dataframe
four_countries = four_countries.drop(four_countries.index[0])
```

```python
# Confirm transformation
four_countries.head()
```

```python
# Reset index so years are in a column
four_countries = four_countries.reset_index()
# Rename column
four_countries = four_countries.rename(columns={'index':'Year'})
```

```python
# Confirm name change
four_countries.head()
```

## Add Subplots with HIV Prevalence Data

```python
# Variables
years = four_countries["Year"].values
# A list of countries taken from header row, and excludes 'Year'
# We need this variable because the code above will randomly select four countries, and we 
# Have no way of knowing which countries yearly prevalence we will need to plot
country_names = four_countries.columns[1:]
```

```python
# This will create a 2x2 matrix of subplots  and the index = 1 will refer to the position of the subplot
# We used the figsize=(width, height) to set the dimensions of the figure
fig = plt.figure(figsize=(15, 10))
# To create a subplot using this method, you need to create a unique axes instance variable and assign the 
# subplot to that axes
ax1 = fig.add_subplot(2, 2, 1) #This will create a plot on the top left grid
ax2 = fig.add_subplot(2, 2, 2) #This will create a plot on the top right grid
ax3 = fig.add_subplot(2, 2, 3) #This will create a plot on the bottom left grid
ax4 = fig.add_subplot(2, 2, 4) #This will create a plot on the bottom left grid

# Add the data
ax1.plot(years, four_countries[country_names[0]]) # Add data to the 1st AxesObject Instance!
ax2.plot(years, four_countries[country_names[1]]) # Add data to the 2bd AxesObject Instance!
ax3.plot(years, four_countries[country_names[2]]) # Add data to the 2bd AxesObject Instance!
ax4.plot(years, four_countries[country_names[3]]) # Add data to the 2bd AxesObject Instance!

plt.show()
```

# Customize Plot!
 - Add Title and Axis labels
 - Format axis labels

### Add subplot axes and titles labels
 - x-axis label: set_xlabel('label')
 - y-axis label: set_ylabel('label')
 - title: .set_title("label")

```python
# This will create a 2x2 matrix of subplots  and the index = 1 will refer to the position of the subplot
# We used the figsize=(width, height) to set the dimensions of the figure
fig = plt.figure(figsize=(15, 10))
# To create a subplot using this method, you need to create a unique axes instance variable and assign the 
# subplot to that axes
ax1 = fig.add_subplot(2, 2, 1) #This will create a plot on the top left grid
ax2 = fig.add_subplot(2, 2, 2) #This will create a plot on the top right grid
ax3 = fig.add_subplot(2, 2, 3) #This will create a plot on the bottom left grid
ax4 = fig.add_subplot(2, 2, 4) #This will create a plot on the bottom left grid

# Subplot Nbr 1
ax1.plot(years, four_countries[country_names[0]]) 
ax1.set_title(country_names[0])
ax1.set_xlabel('Year') # Add the x-axis label
ax1.set_ylabel('Prevalence of HIV, total (% of population ages 15-49)') # Add the y-axis label
               
# Subplot Nbr 2               
ax2.plot(years, four_countries[country_names[1]]) 
ax2.set_title(country_names[1])
ax2.set_xlabel('Year') # Add the x-axis label
ax2.set_ylabel('Prevalence of HIV, total (% of population ages 15-49)') # Add the y-axis label

# Subplot Nbr 3
ax3.plot(years, four_countries[country_names[2]]) 
ax3.set_title(country_names[2])
ax3.set_xlabel('Year') # Add the x-axis label
ax3.set_ylabel('Prevalence of HIV, total (% of population ages 15-49)') # Add the y-axis label
               
# Subplot Nbr 4               
ax4.plot(years, four_countries[country_names[3]]) 
ax4.set_title(country_names[3])
ax4.set_xlabel('Year') # Add the x-axis label
ax4.set_ylabel('Prevalence of HIV, total (% of population ages 15-49)') # Add the y-axis label


plt.show()
```

### Customize X-ticks and spaceing between Subplots
 - Rotate X-tick Labels using set_xticklabels(X-axis variable, rotation=rotation angle)
 - Add space to subplots in figure using fig.subplots_adjust(hspace=0-1, wspace=0-1)
 - Adjust label string so that it is in two lines! just add '\n' where you want the lines to break

```python
# This will create a 2x2 matrix of subplots  and the index = 1 will refer to the position of the subplot
# We used the figsize=(width, height) to set the dimensions of the figure
fig = plt.figure(figsize=(15, 10))
fig.subplots_adjust(hspace=.5, wspace=0.2)
# To create a subplot using this method, you need to create a unique axes instance variable and assign the 
# subplot to that axes
ax1 = fig.add_subplot(2, 2, 1) #This will create a plot on the top left grid
ax2 = fig.add_subplot(2, 2, 2) #This will create a plot on the top right grid
ax3 = fig.add_subplot(2, 2, 3) #This will create a plot on the bottom left grid
ax4 = fig.add_subplot(2, 2, 4) #This will create a plot on the bottom left grid

# Subplot Nbr 1
ax1.plot(years, four_countries[country_names[0]]) 
ax1.set_title(country_names[0])
ax1.set_xlabel('Year') # Add the x-axis label
ax1.set_ylabel('Prevalence of HIV, \n total (% of population ages 15-49)') # Add the y-axis label
ax1.set_xticklabels(years, rotation=45)#we use this method to rotate the x-tick labels so they are visible!

# Subplot Nbr 2               
ax2.plot(years, four_countries[country_names[1]]) 
ax2.set_title(country_names[1])
ax2.set_xlabel('Year') # Add the x-axis label
ax2.set_ylabel('Prevalence of HIV, \n total (% of population ages 15-49)') # Add the y-axis label
ax2.set_xticklabels(years, rotation=45)

# Subplot Nbr 3
ax3.plot(years, four_countries[country_names[2]]) 
ax3.set_title(country_names[2])
ax3.set_xlabel('Year') # Add the x-axis label
ax3.set_ylabel('Prevalence of HIV, \n total (% of population ages 15-49)') # Add the y-axis label
ax3.set_xticklabels(years, rotation=45)

# Subplot Nbr 4               
ax4.plot(years, four_countries[country_names[3]]) 
ax4.set_title(country_names[3])
ax4.set_xlabel('Year') # Add the x-axis label
ax4.set_ylabel('Prevalence of HIV, \n total (% of population ages 15-49)') # Add the y-axis label
ax4.set_xticklabels(years, rotation=45)

plt.show()
```

# Add multiple subplots to a figure using less code!
 - Use a for loop

```python
# Variables Again
years = four_countries["Year"].values
# A list of countries taken from header row, and excludes 'Year'
# We need this variable because the code above will randomly select four countries, and we 
# Have no way of knowing which countries yearly prevalence we will need to plot
country_names = four_countries.columns[1:]
```

### Code first, explanation afterwards :)

```python
# Test function
fig = plt.figure(figsize=(15,10))
for country, num in zip(country_names, range(1,5)):
    
    ax = fig.add_subplot(2,2, num)
    ax.plot(years, four_countries[country])
    ax.set_ylabel('Prevalence of HIV, \n total (% of population ages 15-49)')
    ax.set_xlabel('Years')
    ax.set_xticklabels(years, rotation=45)#we use this method to rotate the x-tick labels so they are visible!
    ax.set_title(country)
    
plt.tight_layout()
plt.show()
```

# How does this work?!

fig = plt.figure(figsize=(15,10))
 - This code will create a figure object and set the size 15 width by 10 height
<hr>
for country, num in zip(country_names, range(1,5)):
 - the country_names is a list of the countries we will plot
 - since we are using a for loop, country is an index of the country_names array that as we iterate through the for loop, will change country
 - num is also a counter, although it is counting the index of the subplot
 - range(1,5) indicates how many subplots, range has to be >0, and the last number is not inclusive
 - zip(country_names, range(1,5) is getting the country index and range index for us that will be used inside the for loop
<hr>
ax = fig.add_subplot(2,2, num)
 - notice that here we are creating two rows and 2 columns, and the last position is 'num', as we move through the for loop, we will start at num = 1, so the first subplot will be first row, first column
<hr>
ax.plot(years, four_countries[country])
 - This code will add the specific country data to the plot in the current loop
<hr>
ax.set_ylabel('Prevalence of HIV, \n total (% of population ages 15-49)')<br>
ax.set_xlabel('Years')
 - This code adds the labels to the subplot in the current loop
<hr>
ax.set_xticklabels(years, rotation=45)
 - This code will rotate the x-tick labels so they are visible!
<hr>
ax.set_title(country)
 - This code will add the title of the current subplot in the current loop
<hr>
When we complete the first loop, we then iterate to the next country, and num in zip(country_name, range(1,5))<br>
