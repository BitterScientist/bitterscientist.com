# Load Data

```python
# Only select these columns
usecols=['Date Time', 'Value']
# Read CSV file
central_df = pd.read_csv('static/data/central.csv', usecols=usecols)
nwcoast_df = pd.read_csv('static/data/northwestcoast.csv',usecols = usecols)
```

# Transform Data

```python
# Change datatypes
central_df["Date Time"] = central_df["Date Time"].astype('datetime64[ns]')
nwcoast_df["Date Time"] = nwcoast_df["Date Time"].astype('datetime64[ns]')
```

### Group data by month using pd.Grouper()

```python
# This will downsample the data so that instead of having daily values
# We will get the mean value per month, so we are going from 4188 records
# to 7 records! This was done for plotting simplicity.
month_central_df = central_df.groupby([pd.Grouper(key='Date Time',freq='M')]).agg({'Value':'mean'})
month_nwcoast_df = nwcoast_df.groupby([pd.Grouper(key='Date Time',freq='M')]).agg({'Value':'mean'})
```

# The object-oriented (OO) interface

The OO interface allows users to prepare figures with many subplots.

Two instances are used in this method: 
 - 1) axes.Axes - represents an individual plot <font color="red">do not confuse with "axis"</font>, which refers to the x/y axis of a plot.
 - 2) figure.Figure: is the final image that may contain 1 or more Axes.
 - the figure is the canvas, and the axes instances are the images on the canvas.

### Generate an instance of figure.Figure and axes.Axes

```python
# The anatomy of a 
# Creates an instance of figure ang axes
fig, ax = plt.subplots()
plt.show() #This displays all active figure objects in code block.
```

### Create a simple plot with one axes instance
 - matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
 - Documentation 
 <a href="https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html">matplotlib.pyplot.plot()</a>
 - This Plot y versus x as lines and/or markers.
 - plot(x, y)       # plot x and y using default line style

```python
# the x-axis labels = month_central_df.index.values
# the y-axis values = month_central_df['Value']
fig, ax = plt.subplots()
ax.plot(month_central_df.index.values, month_central_df['Value'])
plt.show()
```

### Create a simple plot with more than one axes instance
 - Here we create a single axes instance, and we are created 'layers' of line plots.
 - the first is for central dataset and the other is the data from the north west coast

```python
# we add two lines to the ax instance
fig, ax = plt.subplots()
ax.plot(month_central_df.index.values, month_central_df['Value'])
ax.plot(month_nwcoast_df.index.values, month_nwcoast_df['Value'])
plt.show()
```

```python
# we markes to ax instances
fig, ax = plt.subplots()
ax.plot(month_central_df.index.values, month_central_df['Value'], marker=".")
ax.plot(month_nwcoast_df.index.values, month_nwcoast_df['Value'], marker="o")
plt.show()
```

### Adding or removing 'linestyles'
<strong>Linestyles </strong><a href="https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html">documentation</a><br>
 - can be defined using:
 - strings: "solid", "dotted", "dashed" or "dashdot"
 - symbols: "--", "."
 - to remove linestyle="None"

```python
# create dashdot with string
fig, ax = plt.subplots()
ax.plot(month_nwcoast_df.index.values, month_nwcoast_df['Value'], linestyle="dashdot", marker="o")
plt.show()
```

```python
# create dashdot with symbols
fig, ax = plt.subplots()
ax.plot(month_nwcoast_df.index.values, month_nwcoast_df['Value'], linestyle='-.', marker="o")
plt.show()
```

```python
# remove linestyle
fig, ax = plt.subplots()
ax.plot(month_nwcoast_df.index.values, month_nwcoast_df['Value'], linestyle='None', marker="o")
plt.show()
```

### color selection
<strong>Colors </strong><a href="https://matplotlib.org/api/colors_api.html">documentation</a><br>
 - Some formats to specify a color:
 - GB or RGBA tuple of float values in [0, 1] (e.g., (0.1, 0.2, 0.5) or (0.1, 0.2, 0.5, 0.3))
 - Hex RGB or RGBA string (e.g., '#0f0f0f' or '#0f0f0f80'; case-insensitive)
 - character representation {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}
 - string representation {'blue', 'green', 'red', 'cyan'...}

```python
# change color using rgb
fig, ax = plt.subplots()
ax.plot(month_central_df.index.values, month_central_df['Value'], linestyle="--", 
        marker="v", color="#0f0f0f80")
plt.show()
```

```python
# change color using string
fig, ax = plt.subplots()
ax.plot(month_central_df.index.values, month_central_df['Value'], linestyle="--", 
        marker="v", color="blue")
plt.show()
```

```python
# change color using character
fig, ax = plt.subplots()
ax.plot(month_central_df.index.values, month_central_df['Value'], linestyle="--", 
        marker="v", color="r")
plt.show()
```

```python
# multiple lines with multiple colors
fig, ax = plt.subplots()
ax.plot(month_central_df.index.values, month_central_df['Value'], marker=".", color="r")
ax.plot(month_nwcoast_df.index.values, month_nwcoast_df['Value'], marker="o", color='b')
plt.show()
```

### customizing axes labels
<strong>x axis label using .set_xlabel() </strong><a href="https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html">documentation</a><br>
<strong>y axis label using .set_ylabel() </strong><a href="https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html#matplotlib.axes.Axes.set_ylabel">documentation</a><br>

```python
# multiple lines with multiple colors
fig, ax = plt.subplots()
ax.plot(month_central_df.index.values, month_central_df['Value'], marker=".", color="r")
ax.plot(month_nwcoast_df.index.values, month_nwcoast_df['Value'], marker="o", color='b')
ax.set_xlabel
plt.show()
```

### Save Figure
 - fig.savefig('name.extension', transparent=False, dpi=value, bbox_inches="tight")

```python
fig, ax = plt.subplots()
ax.plot(central_df["Date Time"], central_df['Value'])
fig.savefig('single_subplot.png', transparent=False, dpi=80, bbox_inches="tight")
plt.show()
```
