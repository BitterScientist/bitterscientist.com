# Valence

 - The polarity score is a float within the range [-1.0, 1.0]. 
 - The subjectivity is a float within the range [0.0, 1.0] where 0.0 is very objective and 1.0 is very subjective.

```python
def get_valence(dataframe):
    valence_data = []
    for feature in dataframe.columns:
        text = dataframe[feature].str.cat(sep=',')
        valence = TextBlob(text).sentiment
        valence_data.append([feature, valence, text])
    return valence_data
```

```python
valence_data = get_valence(result)
```

```python
valence_df = pd.DataFrame.from_records(valence_data)
```

```python
valence_df.columns = ["feature", "valence", "text"]
```

```python
valence_df[['polarity', 'subjectivity']] = pd.DataFrame(valence_df['valence'].tolist(), index=valence_df.index)
```

```python
valence_df.drop(['valence'], axis=1, inplace = True)
```

```python
valence_df
```

```python
valence_df.to_csv("data/valence_df.csv", index=False)
```
