<div id="objective1" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">F-Test comparing two population variances.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The basics</u></strong>
        <li>The F distribution is a derived distribution with two degrees of freedom $F_{df_1, df_2}$</li><br>
        <li>The F-test uses the F distribution to compare two population variances.</li><br>
        <li>Unlike the t-test that is robust as long as the sample size is large enough, the F test needs to be normally distributed.</li><br>        
        <li>
            <ul>Assumptions:
                <li>Two random samples from respective populations</li><br>
                <li>The sampels are independent of one another</li><br>                
                <li>The two populations are normally distributed with $N(\mu_1, \sigma_1)$ and $N(\mu_2, \sigma_2)$</li><br>
            </ul>                
        </li><br>
        <li>Hypothesis test:<br>
                $$\begin{aligned} H_0\colon\sigma_1^2/\sigma_2^2=1\quad\text{against}\quad&H_a\colon\sigma_1^2/\sigma_2^2\ne1\\ \text{or}\quad&H_a\colon\sigma_1^2/\sigma_2^2>1\\ \text{or}\quad&H_a\colon\sigma_1^2/\sigma_2^2<1 \end{aligned}$$ 
        </li><br>         
        <li>
            <ul>Hypothesis test values:<br>
                <li>Parameter of interest. $$\colon\sigma_1^2/\sigma_2^2$$</li><br>
                <li>Point estimate. $$\colons^2/s^2$$</li><br>
                <li>Standard error under $H_0$: $$\widehat{\mathsf{SE}}=\sqrt{\frac{s_p^2}{n_1}+\frac{s_p^2}{n_2}}=s_p\sqrt{\frac1{n_1}+\frac1{n_2}}$$</li><br>
                <li>Test statistic. $$f=\frac{s_1^2}{s_2^2}$$</li><br>
                <li>Null distribution: F distribution</li><br>
                <li>Degrees of freedom: $$df_{num}=n_1 -1$$ and $$df_{denom}=n_2 -1$$</li><br>
            </ul>                
        </li><br>  
        <li>
            <ul>Confidence intervals:<br><br>
                <li>Calculated as $100(1- \alpha)$</li><br>
                <li>Two-sides: $$\left(\frac{s_1^2/s_2^2}{f_{n_1-1,n_2-1,\alpha/2}}, \frac{s_1^2/s_2^2}{f_{n_1-1,n_2-1,1-\alpha/2}}\right)$$</li><br>
                <li>Upper one-sided: $$\left(0,\frac{s_1^2/s_2^2}{f_{n_1-1,n_2-1,1-\alpha}}\right)$$</li><br>
                <li>Lower one-sided: $$\left(\frac{s_1^2/s_2^2}{f_{n_1-1,n_2-1,\alpha}},\infty\right)$$</li><br>
            </ul>                
        </li><br>
        <li>R code<code>var.test</code></li><br>
    </ul>
</div>

<div id="objective2" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Comparing two population proportions.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics</u></strong>
        <li>Procedure for comparing two population proportion is similar to comparing two population means.</li><br>
        <li>The following procedure will compare two population proportions $p_1 - p_2$ by comparing the differenced between the two proportions $\hat{p_1} - \hat{p_2}$.</li><br>
        <li>
            <ul>Assumptions:
                <li>Random smamples from both populations, and each sampliing unit to be success or failure.</li><br>
                <li>The two samples are independent of one another.</li><br>
                <li>The two sample sizes are large enough to invoke the central limit theorem.</li><br>
                <li>Rule of thumb, $np \geq 5$ and $n(1-p) \geq 5$.</li><br>           
            </ul>                
        </li><br>
        <li>Hypothesis test:<br>
                $$\begin{aligned} H_0\colon p_1=p_2\quad\text{against}\quad&H_a\colon p_1\ne p_2\\ \text{or}\quad&H_a\colon p_1<p_2\\ \text{or}\quad&H_a\colon p_1>p_2 \end{aligned}$$      
        </li><br>  
        <li>
            <ul>Hypothesis test values:<br>
                <li>Parameter of interest. $p_1 = p_2$</li><br>
                <li>Point estimate. $\hat{p_1} - \hat{p_2}$</li><br>
                <li>Standard error under $H_0$: $$\sqrt{\hat p(1-\hat p)(\frac{1}{n_1}+\frac{1}{n_2})}$$</li><br>
                <li>Test statistic. $$z=\frac{\hat p_1-\hat p_2}{\sqrt{\hat p(1-\hat p)(\frac{1}{n_1}+\frac{1}{n_2})}}$$</li><br>
                <li>Null distribution: $\mathsf{N}(0,1)$ Distribution</li><br>
                <li>Degrees of freedom: None</li><br>
                <li>The pooled sample proportion: $$\hat p_1=\frac{x_1}{n_1},\quad \hat p_2=\frac{x_2}{n_2},\quad\hat p=\frac{x_1+x_2}{n_1+n_2}$$</li><br>
            </ul>                
        </li><br> 
        <li>
            <ul>Confidence intervals:<br><br>
                <li>Calculated as $100(1- \alpha)$</li><br>
                <li>Two-sides: $$\left(\hat p_1-\hat p_2\pm z_{\alpha/2}\sqrt{\hat p_1(1-\hat p_1)/n_1+\hat p_2(1-\hat p_2)/n_2}\right)$$</li><br>
                <li>Upper one-sided: $$\left(-1,\hat p_1-\hat p_2+ z_{\alpha}\sqrt{\hat p_1(1-\hat p_1)/n_1+\hat p_2(1-\hat p_2)/n_2}\right)$$</li><br>
                <li>Lower one-sided: $$\left(\hat p_1-\hat p_2- z_{\alpha}\sqrt{\hat p_1(1-\hat p_1)/n_1+\hat p_2(1-\hat p_2)/n_2},1\right)$$</li><br>
            </ul>                
        </li><br>         
        <li>R method: <code>prop.test(x, n, alt=TYPE, conf.level=CL, correct=FALSE)</code></li><br>
    </ul>
</div>

# Example 2:

 - Compare bike tire strength during 1 week of usage in the same area by the same bike messenger.
 - Brand A, 34 tires pop, 66 do not pop, total 100
 - Brand B, 25 tires pop, 75 do not pop, total 100
 - Test Brand A proportio of tires pop is gretear than the proportion for Brand B
 - Alpha = 0.05

```r
# z-test critical value
qnorm(0.05, lower.tail=FALSE)
```

```r
# proportions
p_a <- 34/100; p_a
p_b <- 25/100; p_b
sample_pooled <- (34+25)/(200); sample_pooled
se <- sqrt(sample_pooled*(1-sample_pooled)*(1/100 + 1/100));se
z_stat <- (p_a-p_b)/se; z_stat
p.value <- pnorm(z_stat, lower.tail=FALSE);p.value
```

## Manual Method: P-value is greater than alpha 0.05, do not reject $H_0$ No evidence to suggest that Brand B bike tires are more succeptible to tire tears than Brand A.

```r
prop.test(c(34,25), c(100,100), alt='g', conf.level=0.95, correct=FALSE)
```

## R Method: P-value is greater than alpha 0.05, do not reject $H_0$ No evidence to suggest that Brand B bike tires are more succeptible to tire tears than Brand A. Confidence interval between -0.01 and 1
