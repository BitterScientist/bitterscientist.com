<div id="objective1" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">A Primer on Hypothesis Testing.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Hypothesis Testing (test of significance):</u></strong>
        <li>A hypothesis is a proposed explanation for a phenomenon that is based on observations</li><br>
        <li>A hypothesis test is a formal process used to explain a phenomenon by comparing observed data with a claim.</li><br>
        <li>In statistical hypothesis testing, two claims are compared.  These claims are referred to as the <strong>Null Hypothesis</strong> and <strong>Alternative Hypothesis</strong>.</li><br> 
        <li>The <strong>Null Hypothesis</strong> is a statement that there is no relation between the phenomenon under investigation.</li><br>
        <li>The <strong>Alternative Hypothesis</strong> is a statement that provides an alternative to the null hypothesis, and states that there is some kind of relation.</li><br>
        <li>The results of a significance test are expressed in terms of probability. The probability measures how well the data and the claim agree.</li><br> 
        <li>Statistical hypothesis testing uses probabilities in making decisions or arriving at a conclusion.</li><br>
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>Sampling Limitation in Hypothesis Testing.</u></strong><br><br>
        <li>In most cases, researchers do not have access to measure the entire population, and must rely on a sample to make inferences about the properties of the population as a whole.</li><br>        
        <li>Statisticians have developed various test statistics (models) that use sampled data to aid in hypothesis testing.</li><br> 
        <li>Since statistical inferences are based on a sampling of the population, it may occasionally misrepresent the population enough to cause errors in the hypothesis test.</li><br>
        <li>These errors can result in misinterpretation of the results.</li><br>        
        <li><mark>The aim of this post is to discuss the types of errors in hypothesis testing, their causes, and how to manage them</mark>.</li><br>
    </ul>
</div>

<div id="objective2" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Types of Errors</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Probabilities in Hypothesis Testing</u></strong><br>
        <li>As mentioned in the previous section, hypothesis testing uses probabilities to measure how well a claim and the data agree.</li><br>        
        <li>Probabilities in hypothesis testing is used to help determine how "likely" or "unlikely" of observing an extreme value in the direction of the alternative hypothesis (assuming the null hypothesis is true).</li><br> 
        <li>Hypothesis testing does this by using a significance level \(\alpha\) for deciding whether the null hypothesis is rejected, and the alternative hypothesis is accepted.</li><br>
        <li>The level of significance level \(\alpha\) is determined in advance, and usually takes values of <code>{0.1, 0.5, and 0.01}</code>.</li><br>        
        <li>Hypothesis testing uses test statistics to arrive at a probability called a <strong>p-value</strong> that is often used to compare with the level of significance \(\alpha\).</li><br>        
        <li>If the  <strong>p-value</strong> is less than \(\alpha\), then there is evidence to suggest that the null hypothesis is false, and the alternative hypothesis is true.</li><br> 
        <li>If the  <strong>p-value</strong> is greater than \(\alpha\), then there is evidence to suggest that the null hypothesis is true, and the alternative hypothesis is false.</li><br>
        <li>Although this process seems straight forward, it is not fail proof, as described in the next section.</li><br> 
    </ul>
    <hr style="border-top: 2px solid black;">
        <ul><strong><u>Error Types</u></strong>
            <li>Statistical hypothesis testing is based on sampling data from a larger population.</li><br>
            <li>Sampling introduces random errors in our system and results in some level of uncertainty.</li><br>
            <li>Even with a statistically significant result with a <strong>p-value</strong> is less than \(\alpha\) does not imply 100% certainty.</li><br>
            <li>A <strong>p-value</strong> is a probability and there is always some chance of making an incorrect conclusion regarding accepting or rejecting a Null hypothesis.</li><br>
            <li>There are two common mistakes that result in hypothesis testing
                <ul><strong><u>Error Types</u></strong>
                    <li>Rejecting the Null Hypothesis when the Null Hypothesis was actually true.</li><br>        
                    <li>Accept a Null Hypothesis when it is actually false.</li><br>
                </ul>
            </li><br>
        <li>As illustrated in this graphic below, there are two errors, referred to as either type 1 or type 2.</li><br> 
        <li>The error occurs because the decision made from the statistical analysis do not align with the actual state.</li><br>        
        <li>The chances of committing these errors are inversely proportional.  Decreasing one type error rate increases the other type error rate and vice versa.</li><br> 
        <li>The following sections will discuss each error type in more detail.</li><br>
    </ul>

</div>

<div id="objective3" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Type 1 Error.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The \(\alpha\) error: when you Reject Null when Null is true!</u></strong>
        <li>Is known as a false positive error</li>
        <li>A false positive error is an error that occurs when the Null Hypothesis \(H_{0} \) is true but you reject the Null Hypothesis \(H_{0} \)</li><br>        
        <li>In practical terms, this error means you reported a significant result when in fact they occured by chance and are not really significant.</li><br> 
        <li>The probability of making a type 1 error is represented by the level of significance \(\alpha \). which is the p-value below which you decide to reject the null hypothesis.</li><br>
        <li>In other words, an \(\alpha \) of 0.05 indicates that you are willing to accept a 5% chance of being wrong if you reject the null hypothesis when in fact it is true.</li><br>        
        <li>You can reduce the risk of a type 1 error by lowering the \(\alpha \), such as using 0.01, which mean there is a 1% chance of making a type 1 error.</li><br>        
        <li>Recall that type 1 and type 2 errors are inversely proportional, so if you lower the \(\alpha \) level, you also increase the risk of the type 2 error \(\beta \). Beta \(\beta \) will be discussed in the next section.</li><br> 
        <li>A fundamental problem with making type 1 errors is making changes or setting interventions when they are unnecessary.</li><br>
    </ul>
</div>

<div id="objective4" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Type II Error.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The \(\beta\) error: when you Accept Null when Null is false!</u></strong>
        <li>Is known as the false negative error.</li><br>        
        <li>A false negative error is an error that occurs when the Null Hypothesis \(H_{0} \) is false but you accept the Null Hypothesis (H_{0} \).</li><br> 
        <li>In practical terms, the researcher concluded there is not a significant effect, when in actuality there is a significant result.</li><br>
        <li>In other words: there is evidence to reject the null hypothesis but you do not reject the null hypothesis.</li><br>        
        <li>The probability of making a type 2 error is called Beta \(\beta\).</li><br>        
        <li>This is related to the power of the statistical test, called <strong>power</strong>.</li><br> 
        <li>Power = 1 - \(\beta \).</li><br>
        <li>You can decrease the risk of committing a type 2 error by increasing your power, and this is often done by increasing the sample size.</li><br>
        <li>Increasing the sample size can help detect a difference if one truly exists.</li><br>
        <li>A fundamental problem with making type 2 errors is preserving the status quo when change is needed.</li><br>     
    </ul>
</div>

<div id="objective5" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Relationship between \(\alpha \) and \(\beta \) Error Types</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Whats the deal with \(\alpha \) and \(\beta \)?</u></strong>
        <li>When \(\alpha \) is increased from 0.05 to 0.1, you effective increase the type 1 error you are willing to tolerate from 5% to 10%.</li><br>              
        <li>As mentioned before, alpha and beta are inversely proportional, so if you increase type 1 error, you decrease type 2 error, and vice versa.</li><br>        
        <li>In the example below we want to test the hypothesis that the mean weight for a population is 163 lbs.
            <br>The \(H_{0}: \mu = 163 \), and the \(H_{A}: \mu \neq 163\).</li><br> 
        <li>We then plotted the results with two different levels of significance \(\alpha \).</li><br>
        <li>We see in the left plot that when we set \(\alpha \) at 0.05, we are allowing the probability of making a type 1 error of 5% (false positive, saying we have a significant result when in actuality we do not).</li><br>
        <li>On the right plot we set \(\alpha \) at 0.1, and we are allowing the probability of making a type 1 error at 10%.</li><br>
        <li>Notice that as you increase the \(\alpha \), the blue area decreases.</li><br>
        <li>The blue area corresponds to the \(\beta \) area, this is the probability of making a type 2 error (not rejecting the null when the null is false).</li><br>
        <li>In other words, when you increase the probability of rejecting the null hypothesis by increasing alpha , you will also decrease the chance of accepting a false null hypothesis (which is a beta error).</li><br>
        <li>This can be generalized, when you increase alpha, you decrease beta and vice versa.</li>
    </ul>
</div>

```python
# scientific calculations
import numpy as np

# Import Statistics library
import scipy.stats

# Data visualization
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
%matplotlib inline

# seed the random number generator so you get the same results
np.random.seed(10)

#Mean weight and sd for the distribution where the sample will be generated
gr1_mean = 190
gr1_sd = 7.2

# Generate a normal distribution for gr1 weights 
gr1_weight = scipy.stats.norm(gr1_mean, gr1_sd)

#Mean weight and sd for gr2s: True mean and std
gr2_mean = 163
gr2_sd = 7.2

# Generate a normal distribution for gr2 weights 
gr2_weight = scipy.stats.norm(gr2_mean, gr2_sd)

# to plot the PDFs
def evaluate_PDF(rv, x=4):
    '''Input: a random variable object, standard deviation
       output : x and y values for the normal distribution
       '''
    
    # Identify the mean and standard deviation of random variable 
    mean = rv.mean()
    std = rv.std()

    # Use numpy to calculate evenly spaced numbers over the specified interval (4 sd) and generate 100 samples.
    xs = np.linspace(mean - x*std, mean + x*std, 100)
    
    # Calculate the peak of normal distribution i.e. probability density. 
    ys = rv.pdf(xs)

    return xs, ys # Return calculated values
```

```python
fig, axes = plt.subplots(1,2, figsize =(20,8), sharey=True)

for alpha in [0.05,0.1]:
    idx = int(alpha/0.05)-1
    
    #alternative hypothesis plot
    gr1_xs, gr1_ys = evaluate_PDF(gr1_weight)
    axes[idx].plot(gr1_xs, gr1_ys, label='Alternative H', linewidth=4, color='black', linestyle="dashed") 

    #null hypothesis plot
    gr2_xs, gr2_ys = evaluate_PDF(gr2_weight)
    axes[idx].plot(gr2_xs, gr2_ys, label='Null H', linewidth=4, color='black')
    
    #critical values
    thresh_l = gr2_weight.ppf(alpha)
    thresh_r = gr2_weight.ppf(1-alpha)

    axes[idx].vlines(thresh_r,ymin=0,ymax=0.06, linestyles ='dashed')
    axes[idx].vlines(thresh_l,ymin=0,ymax=0.06, linestyles ='dashed')

    axes[idx].fill_betweenx(gr1_ys,x1=thresh_r,x2 = gr1_xs, where = gr1_xs < thresh_r, color='b', alpha=0.5)
    axes[idx].fill_betweenx(gr2_ys,x1=gr2_xs,x2=thresh_r, where = gr2_xs > thresh_r,color='r', alpha=0.5)
    axes[idx].fill_betweenx(gr2_ys,x1=gr2_xs,x2=thresh_l, where = gr2_xs < thresh_l,color='r', alpha=0.5)
    
    axes[idx].set_xlabel('weight (lb)')
    axes[idx].set_title(f'alpha = {alpha}')
    plt.legend([r"$H_{A}$", r"$H_{0}$"])
```
