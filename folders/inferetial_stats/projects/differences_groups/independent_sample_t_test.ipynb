<div id="objective1" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Introduction to independent-sample t-test.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Independent-sample t-test:</u></strong>
        <li>Is used to determine if a difference exists between the mean of two independent groups on a continuous numeric dependent variable.</li><br>
        <li>The t- statistic was introduced in 1908 by <a href="https://en.wikipedia.org/wiki/William_Sealy_Gosset">William Sealy Gosset</a>.</li><br>
        <li>The t-test can be used to determine if two sets of data are significantly different from each other.</li><br>
        <li>This test is also known as:
            <ul>
                <li>Independent t-test</li>
                <li>Independent-measures t-test</li>
                <li>Between-subjects t-test</li>
                <li>Unpaired t-test</li>
                <li>Student's t-test</li>
            </ul>
        </li><br>
        <li>The independent t-test is the most common form of the T test</li><br>
        <li>A t test is used to determine if there is a meaningful difference between groups.  It lets you know if those differences could have happened by chance.</li><br>
    </ul>
    <hr style="border-top: 2px dashed black;">    
    <ul><strong><u>Example</u></strong>   
        <li>
            <ul>
                <li><strong>Case</strong> Determine if the mean salaries (in \($\)) is different between individuals that live in two neighboring zip codes (US).</li>
                <li><strong>Zip Code Variable</strong> would be the independent variable.</li>
                <li><strong>Salary Variable</strong> would be the dependent variable.</li>
                <li><strong>Null Hypothesis \(H_{0} \):</strong> The mean salaries for both zip codes are the same.</li>
                <li><strong>Alternative Hypothesis \(H_{A} \):</strong> The mean salaries for both zip codes are different (two tailed test).</li>
            </ul>
        </li><br>
    </ul>
    <hr style="border-top: 2px dashed black;">
    <ol><strong><u>The three common case uses for independent-sample t-test Study Designs</u></strong>
        <li>Determine if there is a difference between the two independent groups.</li><br>
        <li>Determine if there are differences between an intervention.</li><br>
        <li>Determine if there is a difference in change score.</li><br>
    </ol>
    <hr style="border-top: 2px dashed black;">
    <ul><strong><u>Difference between the two independent groups.</u></strong>
        <li>Is a study design that measures a dependent variable in two independent groups.</li><br>
        <li>The goal is to determine if a mean difference exist in the dependent variable between the groups.</li><br>
        <li>The participants/instances are separated into the different groups, and they are separated based on some categorical characteristic that they do not share. Example, participants are separated on whether they tested positive for Covid-19, yes or no.  An individual cannot be placed in both groups. This ensures independence of observation and two independent groups.</li><br>
        <li>The independent-sample t-test would be used to calculate whether score on the dependent variable are different between the two groups.  If we use the Covid-19 example, we could ask is the number of unique people they interacted with on a daily basis is different.</li><br>
    </ul>
    <hr style="border-top: 2px dashed black;">
    <ul><strong><u>Difference between  an intervention.</u></strong>
        <li>In this study, participants are randomly assigned into one of two groups.</li><br>
        <li>Each group receives a different intervention, a good example is an A/B test. Participants are directed to one version of a website. Version A is the standard website, Version B is a new version that is tested.</li><br>
        <li>The dependent variable is measured at the end of the intervention, and not during or before. This is called a '</li><br>
        <li>The goal of this test is to determine if there is a difference between the two groups after an intervention.</li><br>
        <li>The length of intervention is study design dependent. An A/B test can take place over a day, week, hour, month but both groups should be exposed to the same length of time.</li><br>
    </ul> 
    <hr style="border-top: 2px dashed black;">
    <ul><strong><u>Difference in change score.</u></strong>
        <li>This is a test that is used to determine if there is a difference in the dependent variable for individuals before and after a specific intervention.</li><br>
        <li>The study design would have two groups of participants/instances, and the dependent variable is measured at some baseline state before the intervention, and then measured at the end of the intervention.</li><br>
        <li>The change in score would be the difference between the baseline and end of intervention.</li><br>
        <li>Any difference between the two groups is a result of the intervention.</li><br>
    </ul>    
</div>

<div id="objective2" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Discuss the Assumptions for the independent-sample t-test.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Assumptions</u></strong><br>
        <p>In order to run this test, these assumptions must be considered. Failure to consider these assumptions can result in misleading and inaccurate conclusions.</p><br>
        <li>
            <ol>
                <li>You have one dependent variable that is measured on the continuous scale.</li><br>
                <li>You have one independent variable that consists of two categorical, independent groups.<br><strong>Examples:</strong> (Male, Female), (Home run, No Home Run), (College Degree, HS Degree), Heart Disease ("Yes", "No")</li><br>
                <li>The study has independence of observations. There is no relationship between the participants in either groups.  In the example above, people only live in one zip code and not both zip codes. Individuals are only included once.</li><br>
                <li>There should be no significant outliers in both groups. <em>Outliers are extreme data points within your sample that do no follow the pattern of the other data</em>.</li><br>
                <li>The dependent variable should be approximately normally distributed for each group of the independent variable.</li><br>
                <li>The variance of the dependent variable for each group of the independent variable is equal.  In other words, you have homogeneity of variance.</li><br>
            </ol>
        </li><br>
    </ul>
    <hr style="border-top: 2px dashed black;">
    <ul>
        <li><strong><u>Other test considerations include:</u></strong><br>
            <ul>
                <li>You have at least 6+ participants in each group to proceed with an independent-sample t-test. You can run the with less samples but it will be harder to infer to a larger population.</li><br>
                <li>If both groups have the same number of instances, then this is called a balanced design, if the groups have different number of instances, then this is called an unbalanced design.  The greater the unbalance, then the greater the negative effect on the validity of the test.</li> 
            </ul>
        </li><br>
    </ul>
</div>

<div id="objective3" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Mathematics of Independent-sample t-test.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The t-test statistic</u></strong>
       <li>The test statistics for the independent-sample t-test is calculated using the following formula:<br><br>
        \[t = \frac{\bar{X_{1}} -  \bar{X_{2}}}{\sqrt{\frac{s_{1}^{2}}{n_{1}} +  \frac{s_{2}^{2}}{n_{2}}}} \]
        </li><br>
       <li>where \(\bar{X_{1}},\ \bar{X_{2}}\) are the means for the dependent variables for the independent group 1 and group 2 respectively.</li><br>
       <li>and \(s_{1}^{2},\ s_{2}^{2}\) are the estimator of common variance for the two samples and individually can be calculated using the following formula:<br><br>
        \[s^{2} = \frac{\Sigma(x - \bar{X}_{1})^{2} + \Sigma(x - \bar{X}_{2})^{2}} {n_{1} + n_{2} - 2} \]
        </li><br>
        <li>Once the test statistic is calculated, the t-test table needs to be consulted to obtain the critical value for the Student's t-distribution corresponding to the \(\alpha \) level of significance, typically 0.05, 0.01, or 0.001.  The degree of freedom (df) must be used in the table to get the critical value, and is defined as:<br><br>
        \[df = n_{1} + n_{2} - 2 \]
        </li><br>
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>Critical Region</u></strong>
        <li>The critical region is the region of values that corresponds to the rejection of the null hypothesis at some chosen probability level (see plot below).</li><br>
        <li>The following steps are involved in the hypothesis testing process.</li><br>
            <ol>
                <li>Specify the null \(H_{0} \) and alternate hypothesis \(H_{A} \).</li><br>
                <li>Specify the level of significance \(\alpha \).</li><br>
                <li>Calculate the test statistic.</li><br>
                <li>Determine the critical value using the test statistic value, the degrees of freedom value, and the t-test table.</li><br>
                <li><mark style=" background-color: lightgreen;">If the the absolute value of the t-test statistic \(|t|\) is greater than the critical value, then the difference between the two independent groups is significantly different.</mark></li><br>
            </ol>
    </ul>
</div>

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

t = np.arange(-3, 3, 0.1)
T_pos = 2.048 # upper-tailed test where alpha = 0.05
T_neg = -2.048 # lower-tailed test where alpha = 0.05

plt.figure(figsize=(15,5)) # figure size
plt.plot(t, norm.pdf(t)) # plots the z curve using normal probability distribution
# fills in area where Z-test is 0.05 and above
plt.fill_between(t[t>T_pos], norm.pdf(t[t>T_pos]), color="red") # upper tail
plt.fill_between(t[t<T_neg], norm.pdf(t[t<T_neg]), color="red") # lower tail

# Plots area where Null Hypothesis is accepted
plt.text(-0.5, 0.02, r'Accept $H_{0}:\ p\ value > \alpha\ (0.05)$', fontsize=10)
plt.text(-0.3, 0.17, r'1 - $\alpha$ = 0.95', fontsize=10) # labels region where Null is accepted

# Draws dash line in region where null is accepted
plt.arrow(x=t[t>T_pos][0], y=0, dx=(t[t<T_neg][-1]-t[t>T_pos][0]), dy=0, shape='full', head_length= 0, head_width=0, length_includes_head=False, linestyle='--')

# Plots area where Null Hypothesis should be rejected

# vertical lines that indicate where to reject and accept null hypothesis
plt.vlines(t[t<T_neg][-1], 0, 0.10, colors="black", linestyle="dashed") # lower tail critical vertical line
plt.vlines(t[t>T_pos][0], 0, 0.10, colors="black", linestyle="dashed")  # upper tail critical vertical line

# Text that indicates where to reject
plt.text(T_pos, 0.12, r'Reject $H_{0}$:''\n'r'$p\ value < \alpha\ (0.05)$', fontsize=10)
plt.text(-3, 0.12, r'Reject $H_{0}$:''\n'r'$p\ value < \alpha\ (0.05)$', fontsize=10)
plt.text(T_pos, 0.17, r'$\alpha$ = 0.05, t-test = 2.048', fontsize=10)
plt.text(-3, 0.17, r'$\alpha$ = 0.05, t-test = -2.048', fontsize=10)

# Horizontal red arrowy that indicate direction to reject
plt.arrow(x=t[t<T_neg][-1], y=0.10, dx=-1, dy=0, shape='full', head_length= 0.05, head_width=0.01, length_includes_head=True, color="red", linestyle='--')
plt.arrow(x=t[t>T_pos][0], y=0.10, dx=1, dy=0, shape='full', head_length= 0.05, head_width=0.01, length_includes_head=True, color="red", linestyle='--')

# title for plot with details on test
plt.title('Critical Regions for Independent Sample t-test' '\n' r'$\alpha = 0.05,\ t \pm 2.048, \ df=28$')
plt.show()
```

<center><h3><font color="black">Calculate the effect size using Cohen's d.</font></h3></center><br>
    </div>
    <div style="font-size:16px; border:1px solid black; padding:5px">
    <ul><strong><u>Effect size</u></strong>
        <li><mark>The effect size is a method used to express the magnitude of the difference observed with the Independent-sample t-test</mark>.</li><br>
        <li>The Cohen's D Formula is a commonly used method to determine the magnitude (effect size) of the difference and is expressed:<br><br>
            \[d = \frac{|M_{1} - M_{2}|}{S_{pooled}} \]<br>
            where \(M_{1}, M_{2} \) are the sample means<br>
            <br>\(S_{pooled} \) denotes the pooled estimated population standard deviation<br>
            The pooled standard deviation is expressed as:<br><br>
            \[S_{p} = \sqrt{\frac{(N_{1}-1)*S_{1}^{2} + (N_{2}-1)*S_{2}^{2}}{N_{1} + N_{2} - 2}} \]
        </li><br>
        <li>Cohen's D can be calculated by using the group statistics table that is outputted by SPSS during this procedure.</li><br>
        <li>\[S_{p} = \sqrt{\frac{(20-1)*0.29190^{2} + (20-1)*0.39339^{2}}{20 + 20 - 2}} = 0.748 \]</li><br>
        <li>SPSS does produce Cohen's d test in the Independent Sample Effect Size table.  This is in the Point Estimate column for the Cohen's d row (illustrated below).</li><br>
        <li>The effect size can be determined by comparing the Cohen's d value with the following table:
        <table>
              <tr>

        </table>
        </li><br>
