<div id="objective1" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Introduction to One-Way ANOVA.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>One-Way ANOVA:</u></strong>
        <li>Is used to determine if there are any statistically significant difference between the means of two or more independent groups.</li><br>
        <li>One-way ANOVA compares the means of two or more independent groups in order to determine whether there is statistical evidence that the associated population means are significantly different.</li><br>
        <li>ANOVA is short for Analysis of Variance.</li><br>
        <li>ANOVA is also known as:
            <ul>
                <li>Between-subjects ANOVA</li>
                <li>One-factor ANOVA</li>
                <li>One-way Analysis of Variance</li>
            </ul>
        </li><br>
        <li>A one-way ANOVA is an omnibus test statistic that can only tell if you if there is a statistically significant difference but does not tell you which specific groups were significantly different.</li><br>
        <li>If you have more than three groups, a follow-up test would need to be conducted to identify which of these groups differ from each other.</li><br>
        <li>A one-way ANOVA is a parametric test.</li><br>
    </ul>
    <hr style="border-top: 2px dashed black;">  
    <ul><strong><u>Variables used in this test are known as</u></strong>
        <li>Dependent variable that is measured in the continuous level.</li><br>
        <li>Independent variable (grouping variable, or factor) that is divided into two or more mutually exclusive levels, or groups.</li><br>
    </ul>
    <hr style="border-top: 2px dashed black;">
        <ul><strong><u>Common Use Cases</u></strong>
        <li>Field studies</li><br>
        <li>Experiments</li><br>
        <li>Quasi-experiments</li><br>
        <li>Statistical differences among the mean of two or more groups.</li><br>
        <li>Statistical differences among the mean of two or more interventions.</li><br>
        <li>Statistical differences among the mean of two or more change scores.</li><br>
        <li>If the One-Way ANOVA only has two groups, then the test is very much an independent sample t-test.</li><br>          
    </ul>
    <hr style="border-top: 2px dashed black;">    
    <ul><strong><u>Example</u></strong>   
        <li>
            <ul>
                <li><strong>Case</strong> Determine if the there is a difference in the salary based on the type of analyst job position.</li><br>
                <li><strong>Dependent Variable</strong> Salary.</li><br>
                <li><strong>Independent Variable</strong> The type of analyst (Data Analyst, Financial Analyst, Research Analyst, Operations Research Analyst, Market Research Analyst).</li><br>
                <li><strong>Null Hypothesis \(H_{0} \):</strong> There is no difference in the mean salary between the analyst positions.</li><br>
                <li><strong>Alternative Hypothesis \(H_{A} \):</strong> The is a difference in the mean salary between the analyst positions.</li><br>
            </ul>
        </li><br>
    </ul>
    <hr style="border-top: 2px dashed black;">
    <ol><strong><u>The three common case uses for One-Way ANOVA.</u></strong>
        <li>Determine if there are differences between three or more independent groups.</li><br>
        <li>Determine if there are differences between three or more conditions/treatments with no pre-test measurement taken.</li><br>
        <li>Determine if there is differences in change scores.</li><br>
    </ol>
        <hr style="border-top: 2px dashed black;">
    <ol><strong><u>Balanced vs Unbalanced Designs.</u></strong>
        <li>If each group of the independent variable has the same number of instances, then this is a balanced design and is idea.</li><br>
        <li>If each group of the independent variable does not have the same number of instances, then this is a unbalanced design.</li><br>
        <li>The greater the unbalance the greater the negative effect of any violation of assumptions on the validity of the test.</li><br>
    </ol>
</div>

<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Assumptions</u></strong><br>
        <p>In order to run this test, these assumptions must be considered. Failure to consider these assumptions can result in misleading and inaccurate conclusions.</p><br>
            <ol>
                <li>You have one dependent variable that is measured on the continuous scale (ratio or interval).</li><br>
                <li>You have one independent variable that consists of two or more categorical, independent groups. If you have two independent variables, then you should consider a two-way ANOVA.</li><br>
                <li>You have independent of observations, which means there is no relationship between the instances/observations in each group of the independent variable or between the groups.  In other words, an instance cannot be present in more than one group.</li><br>
                <li>There should be no significant outliers in the groups of your independent variable in terms of the dependent variable.</li><br>
                <li>Your dependent variable should be approximately normally distributed for each group of the independent variable.</li><br>
                <li>You should have homogeneity of variance of the dependent variable.</li><br>
            </ol>
    </ul>
</div>

<div id="objective3" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Mathematics of One-Way ANOVA.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Description</u></strong>
        <li>A one-way layout consists of a single factor with several levels and multiple observations at each level.</li><br>
        <li>We can calculate the mean of the observation within each level of our factor (category)</li><br>
        <li>The residuals will tell us about the variation in each level.</li><br>
        <li>We can then look at the average of the mean of each level to obtain a grant mean.</li><br>
        <li>We also look at the deviation of the mean for each level from the grand mean to understand the the level effects.</li><br>
        <li>We then compare the variation within levels to the variation across all levels, which gives this method its name: analysis of variance.</li><br>        
    </ul>
    <hr style="border-top: 2px solid black;">
        <ul><strong><u>Model</u></strong>
        <li>The description can be modeled as follows:<br><br>
            \[y_{ij} = m + a_{i} + e_{ij} \]
            <ul style=" list-style-type: none;">The equation indicates
                <li>\(j\)th data value</li><br>
                <li>\(i\) the level</li><br>
                <li>The sum of the three components is the grand mean \(m\), the level effect \(a_{i}\), and the residual \(e_{ij}\).</li><br>
            </ul>
            </li><br>
        <li>The ANOVA table for the one-way case is given by:
            <ul style=" list-style-type: none;"><strong><u>Factor</u></strong>
                <li><strong>Sum of Squares:</strong> \(SS_{F} = J\Sigma(\bar{y_{i}} - \bar{y})^{2}\)</li>
                <li><strong>DoF:</strong> \(I-1\)</li>
                <li><strong>Mean Square:</strong> \(MSF = \frac{SS_{F}}{(I - 1)}\)</li>
                <li><strong>\(F_{0}\):</strong> \(\frac{MSF}{MSE}\)</li>
            </ul><br>
            <ul style=" list-style-type: none;"><strong><u>Residual</u></strong>
                <li><strong>Sum of Squares:</strong> \(SS_{E} = \Sigma\Sigma(y_{ij} - \bar{y})^{2}\)</li>
                <li><strong>DoF:</strong> \(I(J-1)\)</li>
                <li><strong>Mean Square:</strong> \(MSE = \frac{SS_{F}}{(I(J - 1))}\)</li>
            </ul><br>
            <ul style=" list-style-type: none;"><strong><u>Coor.Total</u></strong>
                <li><strong>Sum of Squares:</strong> \(SS_{T} = \Sigma\Sigma(y_{ij} - \bar{y})^{2}\)</li>
                <li><strong>DoF:</strong> \(IJ-1\)</li>
            </ul>
            </li><br>
        <li>Where<br><br>
            \[\bar{y_{i}} = \frac{1}{J}\Sigma_{j=1}^{J}y_{ij} \]
            </li><br>
            <li>and<br><br>
            \[\bar{y_{..}} = \frac{1}{IJ}\Sigma_{i=1}^{I}\Sigma_{j=1}^{J}y_{ij} \]
            </li><br>
            <li>The row "Corr. Total" row contains the corrected total sum of squares and the associated degrees of freedom (DoF)</li><br>
            <li>The level effect must sum to zero. This is taken care of by most analysis programs automatically.  By making the effect sum to zero, it applies a constraint to the model, and enables us to obtain a unique solution the the continuous linear model.</li><br>
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong><u>Critical Region and testing</u></strong>
        <li>The One-way ANOVA tests to see if the observed data support the hypothesis that the level of the factor are significantly different from each other. This is done by comparing the within-level variance to the between-level variance.</li><br>
        <li>If the observations within each level have the same variance, then we can calculate the variance within each level and pool these to obtain an estimate of the overall population variance. (This is the mean square of the residual.</li><br>
        <li>If there is no level effect then the mean square across levels would be the overall variance/</li><br>
        <li>If there is a level effect then the level mean square will be higher than the residual mean square.</li><br>
        <li><strong>\(F_{0}\):</strong> \(\frac{MSF}{MSE}\) provides this value.</li><br>
        <li>If the <strong>\(F_{0}\):</strong> value is significant at a given level of significance, then there is a level effect present in the data, and we reject the null hypothesis.</li><br>
        <li>The following steps are involved in the hypothesis testing process.</li><br>
            <ol>
                <li>Specify the null \(H_{0} \) and alternate hypothesis \(H_{A} \).</li><br>
                <li>Specify the level of significance \(\alpha \).</li><br>
                <li>Calculate the \(F_{0}\) test statistic.</li><br>
                <li>Determine the critical value using the test statistic value, the degrees of freedom value, and the F distribution table.</li><br>
<!--                 <li><mark style=" background-color: lightgreen;"></mark></li><br> -->
            </ol>
    </ul>
</div>

<div id="spss5" style="font-size:16px; border:1px solid black; padding:5px; background-color: #98FB98;">
<center><h3><font color="black">General Linear Model procedure for an effect size.</font></h3></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:5px">
<ul><strong><u>Effect Size</u></strong>
    <li>The one-way ANOVA informs you whether a difference between the group means is real but does not indicate the magnitude of that effect.</li><br>
    <li>There are different methods used to calculate the effect size, and each has their limitations.</li><br>
    <li>Two methods to measure the effect size for a one-way ANOVA is the the partial eta square \(\eta^{2}\) and the omega squared \(\omega^{2}\) method.</li><br>
    <li>The \(\omega^{2}\) method is calculated using the following formula:<br><br>
    \[\hat{\omega}^{2} = \frac{SS_{b} - (df_{b})MS_{w}}{SS_{t} + MS_{w}} \]<br>
        where \(SS_{b}\) is the sum of squares between groups<br>
        \(df_{b}\) is the degrees of freedom between groups<br>
        \(SS_{b}\) is the sum of the total sum of squares<br>
        \(MS_{w}\) is the mean square for within groups.
    </li><br>
    <li>These values can be found in the ANOVA ouput table</li>
    <li>The partial eta square \(\eta^{2}\) is provided by the test of between subjects effects table, in the last column at the "group" row.</li><br>
    <li>The preferred method is the omega \(\omega^{2}\)</li><br>
</ul>
    <ol><strong><u>Procedure</u></strong>
        <li><code>Analyze &#62; General Linear Model &#62; Univariate</code></li><br>
        <li><strong>Univariate dialogue box</strong>: <code>Alertness score to dependent variable list box &#62; groups to Fixed Factor box &#62; Click Options button</code></li><br>
        <li><strong>Univariate: Options dialogue box</strong>: <code>Select Estimates of effect size in display area &#62; Click Continue</code></li><br>
        <li><strong>Univariate dialogue box</strong>: <code>Click OK Button</code> will generate output.</li><br>
    </ol>
</div>
