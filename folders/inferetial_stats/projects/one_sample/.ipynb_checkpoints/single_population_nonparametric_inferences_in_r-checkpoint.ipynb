        <li>Are types of inferences made on a single population where the assumptions for parametric statistics are violated and do not apply.</li><br>
        <li>The inference is made about a population median.</li><br>
        <li>Test rely on less assumptions, but have smaller power to reject the $H_0$ when the population is normal.</li><br>
        <li>This post will discuss the hypothesis test and confidence inverval for nonparametric tests.</li><br>
        <li>Assumptions is that the data from from a random sample</li><br>
        <li>The random sample is representative of the entire population</li><br>
        <li>The R programming language will be used to carry out these calculations.</li><br>
    </ul>
    <hr>
    <ul><strong><u>How to idenfity if a nonparametric test is needed</u></strong>
        <li>A shapiro-wilk test suggest the sampled population is not normal.</li><br>
        <li>Histogram does not have a single peak and symetric tails.</li><br>
        <li>The data in normal q-q plot does not have a straight line along the diagonal expected and observe line (or z-quantiles versus the sample quantiles).</li><br>
    </ul>    
</div>

<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics</u></strong>
        <li>Shapiro-Wilk test test for normality</li><br>
        <li>
            <ul>Hypothesis test:<br>
                <li>Null Hypothesis. $H_o:$ Sample population is normal. p-value is large.</li><br>
                <li>Alternative Hypothesis $H_a:$ Sample population is not normal. p-value is small.</li><br>
            </ul>                
        </li><br> 
    </ul>
</div>

## Example 1

 - Old faithful geyser data in R MASS Library
 - Histogram for eruptions does not appear normally distributed
 - Carry out a shapiro wilk test

```r
library('MASS')
attach(faithful)
hist(eruptions)
```

## Shapiro Wilk Test

 - <code>shapiro.test(vector)</code>

```r
shapiro.test(eruptions)
```

## Example 1<font color="red"> p value is less than 0.05, reject $H_0$</font> Evidence to suggest the data is not normally distributed

<div id="objective4" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Exact test for a population proportion.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics</u></strong>
        <li>Hypothesis is similar to the z-test on population proportion although the only assumption is that the data comes form a random sample of units that are classified in a binary fashion.</li><br>
        <li>This is an ideal test to use when the “large enough sample” rule of thumb is not satisfied for the z-test for population proportion.</li><br>
        <li>The test in R is implemented with the <code>binom.test(x, n, p=P0, alt=TYPE, conf.level=CL)</code></li><br>
    </ul>
</div>

## Example 3

 - 12% of Employees have been arriving late to work, and its believed to be due to the parking garage malfunctions preventing people from coming in.
 - A new system has been implemented but we need to determine if the proportion of employees running late has decreased.
 - $\alpha = 0.05$
 - $H_0: p = 0.12$
 - $H_a: p \leq 0.12$
 - n = 250
 - df = none for proportions
 - 22 out of 250 late with new system

```r
binom.test(22, 250, p=0.12, alt='l', conf.level=0.95)
```

## Example 3<font color="red"> The p value is greater than 0.05 do not reject $H_0$</font>

<div id="objective5" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">One-Sample Sign Test for a Population Median.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The Basics</u></strong>
        <li>This inference test is used when testing a population median $\tilde{\mu}$</li><br>
        <li>
            <ul>Hypothesis test:<br>
                <li>Null Hypothesis. $H_o:\tilde{\mu} = \tilde{\mu_0} $</li><br>
                <li>Alternative Hypothesis 2 sided: $H_o:\tilde{\mu} \neq \tilde{\mu_0}$</li><br>
                <li>Alternative Hypothesis 1 sided lower tail: $H_o:\tilde{\mu} < \tilde{\mu_0} $</li><br>
                <li>Alternative Hypothesis 1 sided upper tail: $H_o:\tilde{\mu} > \tilde{\mu_0} $</li><br>
            </ul>                
        </li><br>
        <li>The model assumes the data are from a random sample</li><br>
        <li>The data is continuous numeric variable</li><br>
        <li>$\tilde{\mu}$ is $p = \frac{1}{2}$</li><br>
        <li>The test in R is implemented with the <code>binom.test(x, n, p=P0, alt=TYPE, conf.level=CL)</code></li><br>
    </ul>
</div>

## Example 3

 - Is the proportion of the data 0.5 or greater than 0.5
 - $\alpha = 0.05$
 - $H_0: p = 0.5$
 - $H_a: p > 0.5$

```r
data <- c(1.6, 2,2.7,1.3,2,1.3,0.3,0.9,1.2,1.2,0.2,1.3,5,0.8,7.4,3,0.6,1.8,2.5,0.3)
```

```r
# count data greater than 1
x <- sum(data>1); x
```

```r
# number of data points
n_data <- length(data); n_data
```

```r
binom.test(x, n_data, p=0.5, alt='g', conf.level=.95)
```

## Example 4<font color="red"> The p value is greater than 0.05 do not reject $H_0$, the median data is not greater than 0.5</font>
