<div id="objective1" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Provide an introduction to Pearson's Product-Moment Correlation.</font></h1></center>
    <hr style="border-top: 2px solid black;">
    <ul><h4></h4>
        <li>The Pearson product-moment correlation is used to determine the strength and direction of a linear relationship between <strong>two continuous variables</strong>.</li><br>        
        <li>The test generates a coefficient that is called the <strong>Pearson's correlation coefficient</strong>, denoted as \(\rho\).</li><br> 
        <li><strong>Pearson's correlation coefficient \(\rho\)</strong> measures the strength and direction of a linear relationship between <strong>two continuous variables</strong>.</li><br>
        <li>The value of the <strong>Pearson's correlation coefficient \(\rho\)</strong> ranges between <code>-1, 1</code>.</li><br>        
        <li>A <strong>Pearson's correlation coefficient \(\rho\)</strong> with a value of <code>1</code> is associated with a perfect positive linear relationship.</li><br>        
        <li>A <strong>Pearson's correlation coefficient \(\rho\)</strong> with a value of <code>-1</code> is associated with a perfect negative linear relationship.</li><br> 
        <li>A <strong>Pearson's correlation coefficient \(\rho\)</strong> with a value of <code>0</code> is associated with no linear relationship between two variables.</li><br> 
        <li>Pearson's correlation is used to determine the relationship between two quantitative variables, rather than the difference.</li><br> 
        <li>Pearson's correlation also provides information as to the degree which the two variables coincide with one another.</li><br>
        <li>Relatedness can be thought of as changes in one variable would correspond to changes in the other variable.</li><br> 
        <li>Pearson's correlation is one the most widely used test to measure the relationship between two sets of data. These other test include:
            <ul>
                <li>phi correlation coefficient</li>
                <li>point-biserial correlation</li>
                <li>Spearmanâ€™s rho</li>
                <li>partial correlation</li>
                <li>part correlation</li>
            </ul> 
        </li><br>
        <hr style="border-top: 2px solid black;">
        <li><strong>Example case uses:</strong>
            <ul>Determine the strength and direction of a linear relationship between 
                <li>the Median value of owner-occupied homes (continuous) and crime incidents (continuous) in a city</li>
                <li>the age of wine (continuous) and the levels of a given antioxidant (continuous)</li>
                <li>salaries (continuous) and length of employment in a given industry (continuous)</li>
            </ul> 
        </li><br>        
    </ul>
</div>

<div id="objective2" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Discuss the Assuptions of a Pearson's Correlation.</font></h1></center>
    <hr style="border-top: 2px solid black;">
    <ul><strong>What are assumptions?</strong>
        <li>Assumptions are characteristics about a given data.</li><br>        
        <li>A few examples of statistical assumptions include:
            <ul>
                <li><strong>Data types:</strong> the data has a specific attribute (integer, continuous, categorical, string, boolean).</li>
                <li><strong>Normality:</strong> the data has a normal distribution.</li>
                <li><strong>Homogeneity of variance:</strong> the data from multiple groups have the same variance.</li>
                <li><strong>Linearity:</strong> the data has a linear relationship.</li>
                <li><strong>Independence:</strong> the data has a specific attribute.</li>
            </ul>
        </li><br>
    </ul>
    <hr style="border-top: 2px solid black;">
    <ul><strong>Why test for assumptions?</strong>
        <li>Statistical tests are developed making characteristics of the data into consideration</li><br>
        <li>Failure to test for these assumptions can result in inaccurate conclusions.</li><br>
        <li>Part of the process of using Pearson's Correlation is checking to make sure the data you want to analyze passes the assumptions needed for this test.</li><br>        
        <li>There are <strong><font color="red">five assumptions</font></strong> that must be met to give you valid results.</li><br> 
        <li>It is common for data to violate some of these assumptions, and the aim of this section is to discuss how to proceed in those cases.</li><br>
    </ul>
    <hr style="border-top: 2px solid black;">
    <ol><strong>The Five Assumptions</strong><br><br>
        <li>Your two variables should be measured on a continuous scale. 
            <br><strong>Example:</strong> GRE scores (interval level), NFL player weights (ratio level).
            <br><u>This assumption is required to use this test</u>.
        </li><br>   
        <li>Your two continuous variables are paired. In other words, each instance in your dataset has a values for each variable.
            <br><strong>Example:</strong> To test for association between NFL players weight and height, we collect data for each instance (NFL players) and we data for the two variables we are interested in, 1) weight and 2) height.
            <br><u>This assumption is required to use this test</u>.
        </li><br>        
        <li>There needs to be a linear relationship between the two variables. This can be checked by using a scatter plot and visually inspecting the graph.</li><br> 
        <li>There should be no significant outliers. Outliers are extreme data points within your sample that do no follow the pattern of the other data.
        <br><u>Pearson's Correlation is sensitive to outliers, which can influence the value of \(\rho\)</u>.
        </li><br>
        <li>For inferential statistics (null hypothesis significance testing), you need to test for normality, or the assumption of bivariate normality.</li>
    </ol>
</div>

<div id="objective3" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Pearsons'Correlation Coefficient Formulas.</font></h1></center>
    <hr>
    <ol><h4>Types of correlation coefficient formulas:</h4><br>The formula used depends on where the data is drawn, population or a sample of the population.<br><br>
        <li>Pearson's Correlation Coefficient expressed in terms of <code>x</code> and <code>y</code> values:
            <ul style="list-style-type: none;">
                <li>\[r = {n(\Sigma xy) - (\Sigma x)(\Sigma y) ) \over \sqrt{[n\Sigma x^2 - (\Sigma x)^2][n\Sigma y^2 - (\Sigma y)^2]} } \]</li>
            </ul>
        </li><br>        
        <li>Sample Pearson's Correlation Coefficient between <code>x</code> and <code>y</code>:
            <ul style="list-style-type: none;">
                <li>\[r = {s_{xy} \over s_{x}  s_{y}} = {\Sigma(x_{i} - \bar x) (y_{i} - \bar y) \over \sqrt{\Sigma(x_{i} - \bar x)^2  \Sigma(y_{i} - \bar y)^2} } \]
</li>
                <li>Where <strong>r</strong> = correlation coefficient.</li>
                <li><strong>\(\mathbf{s_{xy}}\)</strong> = is the sample covariance.</li>
                <li><strong>\(\mathbf{s_{x}, s_{y}}\)</strong> = sample standard deviations.</li>
                <li><strong>\(\mathbf{x_{i}}\)</strong> = x-variable values in a sample.</li>
                <li><strong>\(\mathbf{y_{i}}\)</strong> = y-variable values in a sample.</li>
                <li><strong>\(\mathbf{\bar x, \bar y}\)</strong> = sample means.</li>
            </ul>
        </li><br> 
        <li>Population Pearson's Correlation Coefficient between <code>x</code> and <code>y</code>:
            <ul style="list-style-type: none;">
                <li>\[r = {\sigma_{xy} \over \sigma_{x}  \sigma_{y}} = {\Sigma(x_{i} - \bar x) (y_{i} - \bar y) \over \sqrt{\Sigma(x_{i} - \bar x)^2  \Sigma(y_{i} - \bar y)^2} } \]
</li>
                <li>Where <strong>r</strong> = correlation coefficient.</li>
                <li><strong>\(\mathbf{\sigma_{xy}}\)</strong> = is the population covariance.</li>
                <li><strong>\(\mathbf{\sigma_{x}, \sigma_{y}}\)</strong> = population standard deviations.</li>
                <li><strong>\(\mathbf{x_{i}}\)</strong> = x-variable values in a sample.</li>
                <li><strong>\(\mathbf{y_{i}}\)</strong> = y-variable values in a sample.</li>
                <li><strong>\(\mathbf{\bar x, \bar y}\)</strong> = population means.</li>   
            </ul>
        </li><br>
    </ol>
</div>

<div id="spss4" style="font-size:16px; border:1px solid black; padding:5px">
<center><h3><font color="blue">Step 4: Test for normality.</font></h3></center>
    <ol><strong>Normality test steps </strong>
        <li><code>Analyze > Descriptive Statistics > Explore</code></li>
        <li><code>Explore dialogue box: Move variables to dependent list box.</code></li>
        <li><code>Click Plots button > Boxplots select "None" Radial dial; Descriptive uncheck "stem-and-leaf", Check "Normality plots with tests > click Continue button</code></li>
        <li><code>Returned to Explore dialogue: In Display section, select "Plots" radial dial.</code></li>
        <li><code>Click OK button.</code></li>  
    </ol>
    <ul><strong>Interpretation of Results</strong>
        <li>Shapiro-Wilk test is a commonly used measure to determine if data is normally distributed</li>
        <li><code>Shapiro-Wilk Column > Sig.</code>Gives you the level of significance for this test</li>
        <li><code>Shapiro-Wilk Sig. &#60; 0.05 </code> the test is significant, the assumption of normality is violated.</li>
        <li><code>Shapiro-Wilk Sig. &ge; 0.05 </code> the test is not significant, the assumption of normality is not violated.</li>
        <li><font color="red">Since the p-value for both variables is <code>0.064</code> and <code>1.30</code>, Both variables were normally distributed, as assessed by Shapiro-Wilk's test (p > .05)</font>.</li>
    </ul>
    <ul><strong>Testing for normality using Normal Q-Q plots</strong>
        <li>A normal Q-Q plot is a good method to graphically assess for normality.</li>
        <li>If the data is normally distributed the circular dots that represent the data points for each variable, will be positioned approximately along a diagonal line in the Normal Q-Q plot.</li>
        <li>The data will not align perfectly to the line, but will hover around it.</li>
        <li>By inspected the plots for the Time seated data and the LDL data, we see that the data points wrap around the line, with minor skewness.</li>
        <li><font color="red">Based on the Normal Q-Q Plots, we can conclude that the engagement scores were normally distributed for both Time Seated and LDL, as assessed by visual inspection of Normal Q-Q Plots.</font>.</li>
    </ul>    
</div>

<div id="spss5" style="font-size:16px; border:1px solid black; padding:5px">
<center><h3><font color="blue">Carry out Bivariate Pearson's Correlation.</font></h3></center>
    <ol><strong>Procedure</strong>
        <li><code>Analyze > Correlate > Bivariate</code></li>
        <li>Bivariate Correlation Dialog box: 
            <ul>
                <li>Move both variables to variables box.</li>
                <li>Correlation coefficients box: make sure "Pearson" is checked off</li>
                <li>Test for significance box: make sure radial dial for "Two-tailed" is selected</li>  
                <li>Click Options button > Options dialogue Box: Leave Check boxes for "Statistics" unchecked, and for missing values box, radial dial "Exclude cases pairwise" selected.</li> 
            </ul>
        </li>
        <li><code>Click "Continue" button</code></li>
    </ol>
    <ul><strong>Interpretation of Results</strong>
        <li>The correlation coefficient can take values from +1 to -1.</li>
        <li>Step 1: look at your Pearson Correlation coefficient Time spent vs LDL: <code>0.371</code></li>
        <li>Step 2: Determine if the results are statistically significant, <code>p-value &#60;0.001 </code></li>
        <li><code>Sig (2-tailed) &#60; 0.05 </code> there is statistically significant linear relationship between time seated and LDL levels.</li>
        <li>IF <code>Sig (2-tailed) &ge; 0.05 </code> the test is not significant, and no evidence to suggest a linear relationship between variables.</li>
        <li><font color="red">There was a statistically significant, moderate positive correlation between daily time spent seated and LDL concentration, r(98) = .37, p &#8804; .0005</font>.</li>
    </ul>
</div>
