<h2>Get values needed to calculate the two-sample t statistic</h2>

```python
# import python dependencies
import pandas as pd
import numpy as np
```

```python
# import data
kc_data = pd.read_csv("kansascity_weights.csv")
tb_data = pd.read_csv("tambapay_weights.csv")
```

```python
# step 1: get the descriptive statistics needed (mean and std)
tampa_mean = round(tb_data.mean().values[0],3)
tampa_std = round(tb_data.std().values[0],3)
tampa_n = round(len(tb_data),4)
print(f'The Tampa Bay descriptive statistics are\nMean Weight: {tampa_mean}\nWeight Std: {tampa_std}\nNumber of Players: {tampa_n}')
```

```python
# get the descriptive statistics needed (mean and std)
kc_mean = round(kc_data.mean().values[0],3)
kc_std = round(kc_data.std().values[0],3)
kc_n = round(len(kc_data),4)
print(f'The Kansas City descriptive statistics are\nMean Weight: {kc_mean}\nWeight Std: {kc_std}\nNumber of Players: {kc_n}')
```

```python
# calculate the standard error (denominator )
standard_error = np.sqrt(np.square(tampa_std)/tampa_n + np.square(kc_std)/kc_n)
print(f'The standard error is: {standard_error.round(3)}')
```

<h2>Calculate the two-sample t statistic</h2>

 - Note the two sample t test is simplified so the numerator is only the mean of the sample populations as we are assuming that the mu1 and mu2 = 0
 - the k value is the degree of freedom for the smaller sample size minus 1

```python
t_stat = (tampa_mean -  kc_mean )/ np.sqrt(np.square(tampa_std)/tampa_n + np.square(kc_std)/kc_n)
print(f'The t statistic is: {t_stat.round(3)}')
```

<h2>Find out what the p-value is from the <a href="https://faculty.washington.edu/heagerty/Books/Biostatistics/TABLES/t-Tables/">t-Tables</a></h2>

 - small t values will require the use of an online p-value calculator
 - Recall this is a two-sided test
 - alpha at 0.05
 - we need to use the degrees of freedom (k) for the group with the smaller sample size and minus 1

```python
# the degree of fre
k_df = tampa_n - 1
print(f'The degrees of freedom for this example is: {k_df}')
```

```python
# the t statistic value is so low that we can't find the p-value using 
# a t distribution table. So we have to use an online calculate to get the p-value
# since p is > 0.05, indicates that we cannot reject the null hypotheis
p_value = 0.8451
```

<h2>Calculate the 95% Confidence Interval</h2>

 - critical values obtained from <a href="http://davidmlane.com/hyperstat/t-table.html">David M lane</a>

```python
t_critical_value = 1.9917 # taken from looking at the critical value for the t distribution table at df 75
ci_95_upper = (tampa_mean - kc_mean) + t_critical_value*np.sqrt(np.square(tampa_std)/tampa_n + np.square(kc_std)/kc_n)
ci_95_lower = (tampa_mean - kc_mean) - t_critical_value*np.sqrt(np.square(tampa_std)/tampa_n + np.square(kc_std)/kc_n)
print(f'The confidence interval for the difference in these two means is:({round(ci_95_lower,3)}, {round(ci_95_upper,3)})')
```
