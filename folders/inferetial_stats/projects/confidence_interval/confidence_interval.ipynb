<div style="font-size:16px; border:1px solid black; padding:10px">
    <center><h1>Prequel Information to Confidence Intervals: <br><font color="blue">the Sampling Distribution</font></h1></center>
<ul>
    <li>Before we get into confidence interval, it's important to understand <strong>Sampling Distributions.</strong></li><br>
    <li><mark><strong>Sampling Distribution:</strong> is the distribution of sample statistics with a mean approximately equal to the mean in the original distribution and a standard deviation known as the standard error.</mark></li><br>
    <li>Sample statistics are random variables because they vary from sample to sample.</li><br>    
    <li>Sample statistics include average, median, standard deviation, percentiles.</li><br>   
    <li>Sample statistics have a distribution called the sampling distribution.</li><br>
    <li>An important aspect of a sampling distribution is the standard error (SE).</li><br>
    <li>The standard error is the standard deviation of a sampling distribution.</li><br>
    <li>For a single categorical variable this is called <strong>the standard error of the proportion</strong>.</li><br>
    <li>For a single quantitative variable this is called <strong>the standard error of the mean</strong>.</li><br>
    <li><strong>Standard Error:</strong> Standard deviation of a sampling distribution</li><br>
    <li>The process of constructing a <strong>sampling distribution for a known population</strong>:
        <ol>
            <li>Take a simple random same of <em>n</em> from the population without replacement.</li>
            <li>Record your sample statistics of interest (mean, median, etc)</li>
            <li>Return observations back into the population</li>
            <li>Repeat many times</li>
        </ol>
    </li><br>
    <li>If you to this process long enough and if the sample size is large, the sampling distribution will be approximately normally with a mean equal to the population parameter.</li><br>
    <li>As sample size increases (see image below)
        <ul>
    <li>the variability of sampling distribution decreases.</li>
    <li>the shape of the sampling distribution becomes more similar to a normal distribution regardless of the shape of the population.</li>
        </ul>
    </li><br>
    <li><mark>In a perfect world, we would contrust sampling distributions with known population values, but that is not the case!  We often only have a sample of a population to work with, and we somehow have to make findings that generalize well to the entire population.  This is where confidence intervals come into play.</mark></li><br>
    </ul>
</div>

<center><h1><br><font color="blue">Why Confidence Intervals Matter</font></h1></center>    
    <ul>
        <li>Confidence intervals is a mathematical method to quantify the certainty (or uncertainty) of an estimate.  This is important because you rarely ever have the entire population of data to work with, so you need to calculate the probability that your value is within an acceptable range.</li><br>
        <li><mark><strong>Confidence Interval:</strong> A range computed using sample statistics to estimate an unknown population parameter with a stated level of confidence</mark></li><br>
        <li>Confidence intervals is a way of acknowledging uncertainty, and providing a range of potential values rather than a single point estimate.  It's a way to be "safe".</li><br>
    <li>In these cases we can use the sample data that we do have to construct a confidence interval to estimate the population parameter with a stated level of confidence.</li><br>
    <li>At the center of a confidence interval is the sample statistic, such as a sample mean or sample proportion. AKA point estimate.</li><br>
    <li><strong>Point Estimate:</strong> Sample statistic that serves as the best estimate for a population parameter.</li><br>
    <li>The width of the confidence interval is determined by the margin of error.</li><br>
    <li>The margin of error is the amount that is subtracted from and added to the point estimate to construct the confidence interval.</li><br>
    <li>This informaiton is visualized in the image below.</li><br>        
    <li><strong>Margin of Error:</strong> Half of the width of a confidence interval; equal to the multiplier times the standard error</li><br>
        <li>Confidence intervals are usually expressed as high percentages: <em>90% or 95%</em></li><br>
        <li>This percentage suggest that the 90% or 95% range contains the sampling distribution of the sample.</li><br>   
    <li>Confidence intervals are similar to frequency tables, histograms, boxplots, and standard errors. These tools help you understand the potential error in your sample estimate.</li><br>        
</ul>
</div>

<center><h1><br><font color="blue">Example Calculation</font></h1></center>    
    <ul>
        <li>Python will be used in the following example</li><br>
        <li>The goal is to calculate the 95% Confidence Interval for the Age of the 2021 Kansas City Football players.</li><br>
        <li>The estimated standard error will be the standard deviation since this is a standard error of the mean</li><br>
</ul>
</div>

```python
# import python dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# get data
age_df = pd.read_csv('kc_age.csv')
```

<h2>Histogram for Age of Active KC Football Players</h2>

```python
age_df.hist()
plt.show()
```

<h2>Calculate the Mean and Estimated Standard Error</h2> 

```python
# calculates the mean
mean_age = age_df.mean().values

# assigns the z value at 95%
z_val_95_perc = 1.96

# calculates the standard error standard deviation/square root of n
se_age = age_df.std().values/np.sqrt(len(age_df))
```

<h2>Calculate the Upper and Lower Values Confidence Inteval Values</h2> 

```python
ci_upper = (mean_age + z_val_95_perc * se_age).round(2)[0]
ci_lower = (mean_age - z_val_95_perc * se_age).round(2)[0]
```

```python
print(f'The 95% confidence interval for the age of active players\non the 2020-2021 Kansas City Chiefs Football team is {ci_lower} and {ci_upper}')
```
