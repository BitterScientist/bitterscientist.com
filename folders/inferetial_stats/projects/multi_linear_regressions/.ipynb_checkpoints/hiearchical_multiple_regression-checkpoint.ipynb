<div id="objective1" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Introduction to Hierarchical Multiple Regression (HMR).</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The basics</u></strong>
        <li>Also known sequential multiple regression.</li><br>
        <li>Hierarchical Multiple Regression (HMR) is a statistical model used to predict a dependent variable based on multiple independent variables.</li><br>
        <li>Hierarchical Multiple Regression (HMR) and Standard linear regression ask the same question: <em>how much extra variation in the dependent variable can be explained by the addition of one or more independent variables?</em></li><br>
        <li>Standard linear regression enters all the independent variables at the same time.</li><br>
        <li>Unlike standard multiple regression, HMR enables you to apply the independent variable into the regression model in the order of your choosing. The advantage is that this method allos you to 
            <ul>
        <li>a) control for the effects of the covariates</li>
        <li>b) when predicting the dependent variable, this method lets you take into account the possible causal effects of the independent variables in the order you may surmise are important.</li>        
                </ul>
        </li><br>
        <li>HRM is a framework for model comparison that allows you to build several regression models by adding variables to a previous model a step at at time.  In general, adding additional variables to the model may help improve the proportion of explained variance in the dependent variable.</li><br>
<!--         <li></li><br>         -->
    </ul>
<hr style="border-top: 1px dashed black;">
    <ul><strong><u>Hierarchical Multiple Regression Examples</u></strong>
        <li><strong>Example 1:</strong> Data obtained from the National Cancer Institution Health Information National Trends Survey data for HINTS5 Cycle 1 consists of 1000+ survey responses from men that answers questions regarding health information seeking behavior, their health status, and demographic details.  HMR will be used to predict digital health information seeking behavior (score) and the factors that may explain the variance in these predictions.
            <ul>
                <li><strong>Dependent variable (aka outcome, target, criterion):</strong> digital health information seeking behavior (score)</li><br>
                <li><strong>Independent variables (aka predictor, explanatory, regressor):</strong> education, age, income, employment, hispanic, non-hispanic, co-morbidities, fruit consumption, tablet or smart phone health tracking, digital information seeking.</li><br>
                <li>The HMR can help distinguish between the variation in digital health seeking behavior that can be explained by some and all of the independent variables, in a sequential manner.</li>
            </ul> 
</li><br>
        <li><strong>Example 2:</strong> HMR is used to understand whether the final exam score for an undergraduate stats class can be predicted based on lecture attendance, office hour attendance, number of hours of tutoring received per week, cumulative GPA at registration for class, and major.
            <ul>
                <li><strong>Dependent variable (aka outcome, target, criterion):</strong> the final exam score which is a continuous variable</li><br>
                <li><strong>Independent variables (aka predictor, explanatory, regressor):</strong> the lecture attendance (continuous), office hours attendance number of tutoring hours per week (continuous), cumulative GPA (continuous), and major (nominal data)</li><br>
                <li>The HMR can help distinguish between the variation in the exam performance that can be explained by some and all of the independent variables, in a sequential manner.</li>
            </ul>
        </li><br>
        <li><strong>Example 3:</strong> HMR is used to predict maximal aerobic capacity as indicator of fitness and health.
            <ul>
                <li><strong>Dependent variable (aka outcome, target, criterion):</strong> the VO_2 max is the maximal aerobic capacity and is continuous data.</li><br>
                <li><strong>Independent variables (aka predictor, explanatory, regressor):</strong> subject age, gender, weight, and heart rate 20 minutes after submaximal walking test. n = 100</li><br>
            </ul>
        </li>             
    </ul>
<hr style="border-top: 1px dashed black;">
    <ol><strong><u>Requirements for Hierarchical Multiple Regression</u></strong>
        <li>One dependent variable that is measured at the continuous level.</li><br>
        <li>Two or more independent variables measured at the continuous or nominal level. Ordinal features must be entered as a continuous or nominal variable.</li><br>
        <li>Independence of residuals (errors)</li><br>
        <li>Linear relationship between the predictor variables and the dependent variable</li><br>
        <li>No multicollinearity</li><br>
        <li>No significant outliers</li><br>
        <li>Residuals should be approximately normally distributed</li><br>        
    </ol>   
</div>

<div id="objective2" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Mathematical Intuition behind Hierarchical Multiple Regression.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>The basics</u></strong>
        <li>Hierarchical Multiple Regression begins with an initial set of variables in the regression equation. \[Model_1: ExamScore = Intercept + cGPA  \]</li><br>
        <li>To this initial model you add another set of independent variables to the regression equation, although no changes are made to the dependent variable (final exam scores in this example). \[Model_2: ExamScore = Intercept + cGPA + lecture_attendance + office_attendance  \]</li><br>
        <li>The additional set of independent variables is evaluated in terms of what they add to the prediction (exam scores) as assessed by the changes in the \(R^2 \).</li><br>
        <li>You continue to repeat this process until you finish testing all the blocks of independent variables in your model<br><br> \[Model_3: ExamScore = Intercept + cGPA + lecture_attendance + office_attendance + tutoring_hours + major  \]</li><br>
        <li>Each model will have an \(R^2 \) and our goal is to determine if our final model explains the dependent variable (final exam scores) over the previous models.</li><br>
        <li>The difference in the \(R^2 \) between models can be statistically tested for significance.</li><br>        
    </ul>
<hr style="border-top: 1px dashed black;">
    <ul><strong><u>Conceptual steps</u></strong>
        <li>Set up data so its primed for analysis</li><br>
        <li>Build sequential regression models by adding variables at each step</li><br>
        <li>Compute the \(R^2 \) using analysis of variance (ANOVA) and obtain coefficients using regression.</li><br>
        <li>Compare the sum of squares between the models from the ANOVA results.</li><br>
        <li>Compute the difference in the sum of squares at each step.</li><br>
        <li>Find the F-statistic and p-value for the sum of square differences</li><br>
        <li>Compute the \(R^2 \) from the sum of square differences <br><br> \[R^2 = \frac{SS_E}{SS_T} \]</li><br>        
    </ul>
<hr style="border-top: 1px dashed black;">
    <ul><strong><u>What is unique about HRM vs SMR</u></strong>
        <li>In most cases, an independent variable will contribute to the explained variation in the dependent variable, and some of that explanation will be unique to a specific independent variable.</li><br>
        <li>Standard multiple regression (SMR) does not allow for any of the shared variation to be attributed to a particular independent variable.</li><br>
        <li>Hierarchical Multiple Regression allows you to assign the shared variation to the variables in the preceding regression equation, and as a result you can calculate the added and unique variation a specific independent variable(s) explains the dependent variable.</li><br>       
    </ul>
<hr style="border-top: 1px dashed black;">
    <ul><strong><u>Common Study Designs that Use Hierarchical Multiple Regression</u></strong>
        <ul><strong>Taking covariates into account.</strong>
            <li>Used to understand the effect of an independent variable(s) on a dependent variable when taking potential confounders/nuisance variables into account.</li><br>
            <li>These potential confounders/nuisance variable are also called covariates.</li><br>
            <li>The confounders/nuisance/covariate variables are added to the equation first in the regression equation.</li><br>
            <li>The independent variables of interest are entered into the equation afterwards, and this allows the importance of these variables to be assessed after controlling for the confounders/nuisance/covariate variables.</li><br>
            <li>Using the statistics final exam example from above, if cGPA has been shown to have a strong relationship between exam performance and prior academic achievement, it would be important to account for this first and then focus on our independent variables of interest.</li><br>
            <li>Demographics is another example of a confounders/nuisance/covariate variable that will be considered first in our example, as the use of digital technology is a more interesting independent variation for predicting the digital health information seeking behavior score.</li><br>           
        </ul>
        <hr style="border-top: 0.5px dashed black;">        
        <ul><strong>Test the additional importance of one or more independent variable in predicting the dependent variable.</strong>
            <li>Used to understand the unique contribution of different independent variables in predicting the dependent variable.</li><br> 
            <li>This is sometimes done to predict the dependent variable with as few independent variables as possible, such as when the data collection is invasive and you want to minimize that intrusiveness of your study design.  Another reason is cost, if collecting additional data would be to high for the research budget.</li><br>
            <li>This study design is also useful in identifying if you can build a meaningful model with fewer independent variables.</li><br>           
        </ul>                
    </ul>    
</div>

<div id="objective3" class="alert alert-block alert-warning" style="font-size:16px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Demonstrate how to perform a Hierarchical Multiple Regression in SPSS.</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
    <ul><strong><u>Set-up Data</u></strong>
        <li>Need at least three variables: 1) dependent and 2-3) independent variables</li><br>
        <li>In this example we have VO2max for dependent variable, and age, weight, heart_rate, and gender as independent variables.</li><br>
        <li>Note: if you have polytomous nominal variables (more than 2 levels), then dummy variables must. be created.  Why? categorical variables do not have an intrinsic numeric order (unlike continuous which can be easily understood in terms of rates of change that is standard in regression).  Polytomous categorical variables such as ethnicity would be encoded with a number although those number (1-hispanic, 2-caucasion, 3-asian, 4-african american) would not have any numeric meaning but rather are labels.  Dummy variables creates a binary numeric feature that can be represented of the original data but also used in regression modeling.</li><br>
        <li>Continuous data must be identified as a scaled measure while nominal data need to have values assigned as a dictionary in the values column in SPSS variable view.</li><br>
        <li>If any of the variables is ordinal then they must be entered as either a scale or nominal variable in the variable tab.</li><br>
        <li>The role for each variable should be None, since the input selection that is default may perform some operations you do not intend to. Designating the variable as None gives you greater control.</li><br>        
    </ul>

        <li>In linear regression dialogue box, Transfer dependent variable and the first set of independent variables</li><br>
        <li>Click next in the block section inside linear regression dialogue box and move variables into respective fields. Click next and add the next set of variables and repeat as needed.</li><br>
        <li>Set up the statistics, regression and plot settings.</li><br>
        <li>The selected settings are to return not only the regression model but also diagnostic tests to evaluate if the data meets the requirements for the Hierarchical Multiple Regression.</li><br>       
    </ul>

<div style="font-size:14px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Assumption 3. Check for independence of observations, 4. linear relationships between dependent variables and each independent variables, 5. Homoscedasticity of Residuals</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
<ul>
    <li><strong>Assumption 3:</strong> The assumption of independence of observation in a multiple regression is designed to test for 1st order autocorrelation.  This means that the adjacent observations such as the errors are correlated and not independent. In SPSS the independence of observations can be checked using the Durbin-Watson statistic. The Durbin-Watson statistic can range between 0 to 4, but you want a value close to 2 for the assumption to be acceptable that there is independence of errors.  The Durbin-Watson value is 1.894, which is close to 2 and meets the assumption.</li><br>   
    <li><strong>Assumption 4:</strong> that there is a linear relationship between the dependent variable and each of the dependent variables are accessed by scatter plot using the studentized residuals to the predicted value and the partial regression plot for each independent variable and the VO2 max dependent variable.  The scatter plots each show linear relationship with no non-linear plots.</li><br>
    <li><strong>Assumption 5:</strong> that the data has homoscedasticity of residuals, which can be assessed by plotting the studentized residuals against the unstandardized predicted value, and we can see from that plot that the points evenly scatter between -2 and 2 for the entire range of values between 30 to 60 for the predicted values.  This suggest that the assumption of homoscedasticity is met.</li><br>
</ul>
</div>

<div style="font-size:14px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Assumption 6:  No Multicollinearity</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
<ul>
    <li>Multicollinearity refers to the situation when two or more independent variables are highly associates/correlated with each other.</li><br>   
    <li>The issues with multicollinearity is that makes it difficult to understand which independent variables contribute to the variance explained by the dependent variable and creates technical issues in calculating a multiple regression model.</li><br>
    <li>SPSS can be used to detects multicollinearity using two methods, by inspecting the correlation coefficients and the tolerance VIF values.</li><br>
    <li>The correlation table produced in SPSS presents Pearson correlation between variables, and you want to make sure none of values are greater than 0.7 (the exception are the diagonal cells as they represent the independent value compared to itself.  All other cells in the first set of rows for the Person Correlation are less than 0.7 besides heart rate, which is greater than .7.  Correlated independent variables may need to be reconsidered.</li><br>
    <li>The tolerance and VIF table presents Collinearity Statistics columns. The Tolerance and VIF are essentially reciprocals of each other and only one column needs to be consulted. If the tolerance value is less than 0.1, or the VIF greater than 10, then there is a collinearity problem.  None of the rows have this issue, so we can conclude that the assumption is met.</li><br>
</ul>
</div>

<div style="font-size:14px; border:1px solid black; padding:10px"><center><h1><br><font color="blue">Assumption 7:  No Significant Outliers, high leverage points or influential points</font></h1></center><br>
</div>
<div style="font-size:16px; border:1px solid black; padding:10px">
<br>
<ul>
    <li>Significant Outliers, high leverage points or influential points refer to unusual points that are different from the majority of data points and cab have an impact on the regression line.  There are several methods that can be used to identify three types of unusual data points. </li><br>   
    <li><strong>Significant outliers:</strong> are typically data points that are \(\pm3\ sd\).  A Casewide Diagnostic table highlights cases where points exceed these thresholds and provide their case number.  If no Casewide Diagnostic table is provided in the output, then no points were detected and no significant outliers are present.  This is the case for this example.</li><br>
    <li><strong>Studentized deleted residuals outliers:</strong> are used to determine if there are residuals that also exceed the \(\pm3\ sd\).  One way to do determine if there are any of these points is to sort the sdr_1 column that contains the residuals, by descending value. This will then let you inspect if there are extreme values, which in this case, is not present.</li><br>
    <li><strong>Leverage points:</strong> should be less than 0.2, and those between 0.2 and 0.5 are considered unusual and risky values to have in the model.  One way to identify these instances is to sort the LEV_1 column in descending order and checking if there are any values over 0.2.  No instance in this dataset fall over 0.2 threshold.</li><br>
    <li><strong>Influential points:</strong> the cooks COO_1 column stores the Cook's Distance values for each case, and if the measure if above 1, then that instance should be investigated.  An easy method to evaluate cook distance values is to sort in descending order the COO_1 column. In this example we do not have any values over 1.</li><br>
    <li>In aggregate, no evidence to suggest there are significant outliers, or unusual data points with leverage and cook values that exceed their respective thresholds.</li><br>
</ul>
</div>

<div style="font-size:16px; border:1px solid black; padding:10px">

<ul>

    <li>The histogram of the residuals has a clear center with symmetric tails while the normal Q-Q plot has all the residuals well aligned along the diagonal, which both provide strong evidence of normality.</li><br>
</ul>
</div>
